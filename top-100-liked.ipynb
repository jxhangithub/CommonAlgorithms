17. Letter Combinations of a Phone Number

Given a string containing digits from 2-9 inclusive, return all possible letter combinations that the number could represent. Return the answer in any order.

A mapping of digits to letters (just like on the telephone buttons) is given below. Note that 1 does not map to any letters.

```python
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        all = '23456789'
        d2c = dict()
        cnt = 0
        for d in all:
            if d == '7' or d == '9':
                length = 4
            else:
                length = 3
            tmp = []
            for i in range(length):
                tmp.append(chr(ord('a')+cnt))
                cnt += 1
            d2c[d] = ''.join(tmp)
        ans = []
        if not digits:
            return ans
        L = len(digits)

        def dfs(idx, path):
            if idx == L:
                ans.append(''.join(path))
                return
            for c in d2c[digits[idx]]:
                path.append(c)
                dfs(idx+1, path)
                path.pop()

        dfs(0, [])
        return ans



            
```

22. Generate Parentheses

```python
class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        ans = []
        lp = '('
        rp = ')'
        def dfs(left, right, path):
            if right == n:
                ans.append(''.join(path))
                return
            if left < n:
                path.append(lp)
                dfs(left+1, right, path)
                path.pop()
            if right < left:
                path.append(rp)
                dfs(left, right+1, path)
                path.pop()
        dfs(0, 0, [])
        return ans


```

39. Combination Sum

Given an array of distinct integers candidates and a target integer target, return a list of all unique combinations of candidates where the chosen numbers sum to target. You may return the combinations in any order.

The same number may be chosen from candidates an unlimited number of times. Two combinations are unique if the 
frequency
 of at least one of the chosen numbers is different.

The test cases are generated such that the number of unique combinations that sum up to target is less than 150 combinations for the given input.

```python
class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        candidates.sort()
        L = len(candidates)
        ans = []

        def dfs(idx, path, left):
            if left == 0:
                ans.append(list(path))
                return
            if left < candidates[idx]:
                return
            for i in range(idx, L):
                path.append(candidates[i])
                dfs(i, path, left-candidates[i])
                path.pop()
        
        dfs(0, [], target)

        return ans
```

46. Permutations

Given an array nums of distinct integers, return all the possible permutations. You can return the answer in any order.

```python
class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
        L = len(nums)
        ans = []

        def dfs(path, visit):
            if len(path) == L:
                ans.append(list(path))
                return
            for i in range(L):
                if i in visit:
                    continue
                path.append(nums[i])
                visit.add(i)
                dfs(path, visit)
                visit.discard(i)
                path.pop()

        dfs([], set())
        return ans
```

51. N-Queens

The n-queens puzzle is the problem of placing n queens on an n x n chessboard such that no two queens attack each other.

Given an integer n, return all distinct solutions to the n-queens puzzle. You may return the answer in any order.

Each solution contains a distinct board configuration of the n-queens' placement, where 'Q' and '.' both indicate a queen and an empty space, respectively.

```python
class Solution:
    def solveNQueens(self, n: int) -> List[List[str]]:
        cols = [0] * n
        diag = [0] * (2*n-1)
        anti = [0] * (2*n-1)
        ans = []

        def dfs(row, colPath):
            if row == n:
                ans.append(list(colPath))
                return
            for col in range(n):
                if cols[col] == 1 or diag[row-col+n-1] == 1 or anti[row+col]==1:
                    continue
                colPath.append(col)
                cols[col] = 1
                diag[row-col+n-1] = 1
                anti[row+col]=1
                dfs(row+1, colPath)
                cols[col] = 0
                diag[row-col+n-1] = 0
                anti[row+col]=0
                colPath.pop()
            
        dfs(0, [])

        def draw(colPath):
            res = []
            for col in colPath:
                tmp = []
                for i in range(n):
                    if i == col:
                        tmp.append('Q')
                    else:
                        tmp.append('.')
                res.append(''.join(tmp))
            return res


        boards = []
        for colPath in ans:
            boards.append(draw(colPath))

        return boards

            
```

79. Word Search

Given an m x n grid of characters board and a string word, return true if word exists in the grid.

The word can be constructed from letters of sequentially adjacent cells, where adjacent cells are horizontally or vertically neighboring. The same letter cell may not be used more than once.

```python
class Solution:
    def exist(self, board: List[List[str]], word: str) -> bool:
        R = len(board)
        C = len(board[0])
        L = len(word)

        def dfs(r, c, idx, visit): # current found: upto (r,c) match idx
            if idx == L-1:
                return True
            for dr, dc in (-1,0),(1,0),(0,1),(0,-1):
                nr = r+dr
                nc = c+dc
                if nr < 0 or nr>R-1 or nc<0 or nc>C-1 or (nr,nc) in visit:
                    continue
                if board[nr][nc] != word[idx+1]:
                    continue
                visit.add((nr,nc))
                if dfs(nr,nc,idx+1, visit):
                    return True
                visit.discard((nr,nc))
            return False
        
        for r in range(R):
            for c in range(C):
                if board[r][c] != word[0]:
                    continue
                if dfs(r, c, 0, set(list([(r,c)]))):
                    return True
        return False
            
            
```

```python
class Solution:
    def partition(self, s: str) -> List[List[str]]:
        L = len(s)
        dp = [[0] * L for _ in range(L+1)]
        for length in range(L+1):
            for i in range(L):
                if length in (0,1):
                    dp[length][i] = 1
                elif i+length<L+1 and dp[length-2][i+1] == 1 and s[i] == s[i+length-1]:
                    dp[length][i] = 1
        ans = []
        def dfs(idx, path):
            if idx == L:
                ans.append(list(path))
            for length in range(1, L-idx+1):
                if dp[length][idx] == 1:
                    path.append(s[idx:idx+length])
                    dfs(idx+length, path)
                    path.pop()

        dfs(0, [])
        return ans
        
```

4. Median of Two Sorted Arrays

Given two sorted arrays nums1 and nums2 of size m and n respectively, return the median of the two sorted arrays.

The overall run time complexity should be O(log (m+n)).

```python
class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:

        def getKth(arr1, arr2, k):
            if not arr1:
                return arr2[k]
            if not arr2:
                return arr1[k]
            L1 = len(arr1)
            L2 = len(arr2)
            m1 = L1 // 2
            m2 = L2 // 2
            v1 = arr1[m1]
            v2 = arr2[m2]
            if m1+m2 < k:
                if v1 < v2:
                    return getKth(arr1[m1+1:], arr2, k-m1-1)
                else:
                    return getKth(arr1, arr2[m2+1:], k-m2-1)
            else:
                if v1 < v2:
                    return getKth(arr1, arr2[:m2], k)
                else:
                    return getKth(arr1[:m1], arr2, k)
                
        len1 = len(nums1)
        len2 = len(nums2)
        L = len1+len2
        if L %2 == 1:
            return getKth(nums1, nums2, L // 2)
        else:
            first = getKth(nums1, nums2, L // 2 -1)
            second = getKth(nums1, nums2, L // 2 )
            return (first+second) / 2.0

            
```

33. Search in Rotated Sorted Array

There is an integer array nums sorted in ascending order (with distinct values).

Prior to being passed to your function, nums is possibly rotated at an unknown pivot index k (1 <= k < nums.length) such that the resulting array is [nums[k], nums[k+1], ..., nums[n-1], nums[0], nums[1], ..., nums[k-1]] (0-indexed). For example, [0,1,2,4,5,6,7] might be rotated at pivot index 3 and become [4,5,6,7,0,1,2].

Given the array nums after the possible rotation and an integer target, return the index of target if it is in nums, or -1 if it is not in nums.

You must write an algorithm with O(log n) runtime complexity.

```python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        L = len(nums)
        s,e = 0, L-1
        while s+1 < e:
            m = s+(e-s)//2
            v = nums[m]
            if v==target:
                return m
            if v > target:
                if v <= nums[e] or target >=nums[s]: # on one branch
                    e = m-1
                else: # one up, one down
                    s = m+1
            else:
                if target <= nums[e] or v >= nums[s]: # on one branch
                    s = m+1
                else: # one up, one down
                    e = m-1
        if nums[s] == target:
            return s
        if nums[e] == target:
            return e
        return -1
```

34. Find First and Last Position of Element in Sorted Array

If target is not found in the array, return [-1, -1].

You must write an algorithm with O(log n) runtime complexity.

```python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        if not nums:
            return [-1,-1]
        L = len(nums)

        def getFirst(tar):
            s,e = 0, L-1
            while s+1 < e:
                m=s+(e-s)//2
                v=nums[m]
                if v == tar:
                    e = m
                elif v > tar:
                    e= m-1
                else:
                    s = m+1
            if nums[s]==tar:
                return s
            if nums[e] == tar:
                return e
            return -1
        
        def getLast(tar):
            s,e = 0, L-1
            while s+1 < e:
                m=s+(e-s)//2
                v=nums[m]
                if v == tar:
                    s = m
                elif v > tar:
                    e= m-1
                else:
                    s = m+1
            
            if nums[e] == tar:
                return e
            if nums[s]==tar:
                return s
            return -1
        
        return [getFirst(target), getLast(target)]

```

74. Search a 2D Matrix

You are given an m x n integer matrix matrix with the following two properties:

Each row is sorted in non-decreasing order.
The first integer of each row is greater than the last integer of the previous row.
Given an integer target, return true if target is in matrix or false otherwise.

You must write a solution in O(log(m * n)) time complexity.

```python
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        R = len(matrix)
        C = len(matrix[0])
        start, end = 0, R*C-1
        while start + 1 < end:
            mid = start + (end-start) // 2
            value = matrix[mid // C][mid % C]
            if value == target:
                return True
            if value < target:
                start = mid+1
            else:
                end = mid-1
        if matrix[start // C][start % C] == target:
            return True
        if matrix[end // C][end % C] == target:
            return True
        return False
        
```

153. Find Minimum in Rotated Sorted Array

Suppose an array of length n sorted in ascending order is rotated between 1 and n times. For example, the array nums = [0,1,2,4,5,6,7] might become:

[4,5,6,7,0,1,2] if it was rotated 4 times.
[0,1,2,4,5,6,7] if it was rotated 7 times.
Notice that rotating an array [a[0], a[1], a[2], ..., a[n-1]] 1 time results in the array [a[n-1], a[0], a[1], a[2], ..., a[n-2]].

Given the sorted rotated array nums of unique elements, return the minimum element of this array.

You must write an algorithm that runs in O(log n) time.

```python
class Solution:
    def findMin(self, nums: List[int]) -> int:
        L = len(nums)
        start, end = 0, L-1
        while start + 1 < end:
            if nums[start] < nums[end]:
                return nums[start]
            mid = start + (end-start) // 2
            value = nums[mid]
            if value >= nums[start]:
                start = mid+1
            elif value <= nums[end]:
                end = mid

        first = nums[start]
        second = nums[end]
        if first < second:
            return first
        else:
            return second
            
```

124. Binary Tree Maximum Path Sum

A path in a binary tree is a sequence of nodes where each pair of adjacent nodes in the sequence has an edge connecting them. A node can only appear in the sequence at most once. Note that the path does not need to pass through the root.

The path sum of a path is the sum of the node's values in the path.

Given the root of a binary tree, return the maximum path sum of any non-empty path.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def maxPathSum(self, root: Optional[TreeNode]) -> int:

        def maxSum(root):
            if not root:
                return -math.inf,-math.inf
            leftSingle, leftMax = maxSum(root.left)
            rightSingle, rightMax = maxSum(root.right)
            single = root.val + max(leftSingle, rightSingle, 0)
            total = max(root.val, leftMax, rightMax, single, leftSingle, rightSingle, root.val + leftSingle + rightSingle)
            return single, total
        _, ans = maxSum(root)
        return ans
```

98. Validate Binary Search Tree

Given the root of a binary tree, determine if it is a valid binary search tree (BST).

A valid BST is defined as follows:

The left 
subtree
 of a node contains only nodes with keys less than the node's key.
The right subtree of a node contains only nodes with keys greater than the node's key.
Both the left and right subtrees must also be binary search trees.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isValidBST(self, root: Optional[TreeNode]) -> bool:

        def BST(root):
            if not root:
                return math.inf, -math.inf, True
            leftMin, leftMax, leftBST = BST(root.left)
            rightMin, rightMax, rightBST = BST(root.right)

            if leftBST and rightBST and root.val > leftMax and root.val < rightMin:
                return min(root.val, leftMin), max(root.val, rightMax), True
            else:
                return 0, 0, False
            
        _, _, ans = BST(root)
        return ans
```

101. Symmetric Tree

Given the root of a binary tree, check whether it is a mirror of itself (i.e., symmetric around its center).

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isSymmetric(self, root: Optional[TreeNode]) -> bool:
        if not root:
            return True
        q = [root.left, root.right]
        while q:
            r1 = q.pop()
            r2 = q.pop()
            if not r1 and not r2:
                continue
            elif not r1 or not r2:
                return False
            elif r1.val != r2.val:
                return False
            q.append(r1.left)
            q.append(r2.right)
            q.append(r1.right)
            q.append(r2.left)
        return True
```

Given the root of a binary search tree, and an integer k, return the kth smallest value (1-indexed) of all the values of the nodes in the tree.

```python
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        visit = set(nums)
        res = 0
        for i in nums:
            cnt = 1
            cur = i-1
            while cur in visit:
                visit.discard(cur)
                cur -= 1
                cnt += 1
            cur = i+1
            while cur in visit:
                visit.discard(cur)
                cur += 1
                cnt += 1
            res = max(res, cnt)
        return res

```

Design a data structure that follows the constraints of a Least Recently Used (LRU) cache.

Implement the LRUCache class:

LRUCache(int capacity) Initialize the LRU cache with positive size capacity.
int get(int key) Return the value of the key if the key exists, otherwise return -1.
void put(int key, int value) Update the value of the key if the key exists. Otherwise, add the key-value pair to the cache. If the number of keys exceeds the capacity from this operation, evict the least recently used key.
The functions get and put must each run in O(1) average time complexity.

```python
class Node:
    def __init__(self, key=None, value=None) -> None:
        self.key = key
        self.value = value
        self.prev = None
        self.next = None

class DLList:
    def __init__(self):
        self.head = Node()
        self.tail = Node()
        self.head.next = self.tail
        self.tail.prev = self.head

    def add2head(self,node):
        node.next = self.head.next
        node.prev = self.head
        self.head.next = node
        node.next.prev = node


    def remove(self,node):
        node.prev.next = node.next
        node.next.prev = node.prev
        node.prev = None
        node.next = None

    def move2head(self,node):
        self.remove(node)
        self.add2head(node)

    def removeTail(self):
        node = self.tail.prev
        self.remove(node)
        return node

    

class LRUCache:

    def __init__(self, capacity: int):
        self.capacity = capacity
        self.list = DLList()
        self.key2node = dict()

    def get(self, key: int) -> int:
        if key in self.key2node:
            node = self.key2node[key]
            self.list.move2head(node)
            return node.value
        else:
            return -1
        


    def put(self, key: int, value: int) -> None:
        if key in self.key2node:
            node = self.key2node[key]
            node.value = value
            self.list.move2head(node)
        else:
            if len(self.key2node) == self.capacity:
                node = self.list.removeTail()
                self.key2node.pop(node.key)
            node = Node(key, value)
            self.key2node[key] = node
            self.list.add2head(node)
        
```

```python
from typing import List
class Solution:
    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:
        R = len(matrix)
        C = len(matrix[0])
        res = []
        rowbegin = colbegin = 0
        rowend = R
        colend = C
        while rowbegin < rowend and colbegin < colend:
            for c in range(colbegin, colend):
                res.append(matrix[rowbegin][c])
            rowbegin += 1
            if not (rowbegin < rowend and colbegin < colend):
                break
            for r in range(rowbegin, rowend):
                res.append(matrix[r][colend-1])
            colend -= 1
            if not (rowbegin < rowend and colbegin < colend):
                break
            for c in range(colend-1, colbegin-1, -1):
                res.append(matrix[rowend-1][c])
            rowend -= 1
            if not (rowbegin < rowend and colbegin < colend):
                break
            for r in range(rowend-1, rowbegin-1, -1):
                res.append(matrix[r][colbegin])
            colbegin += 1
        return res 




```

73. Set Matrix Zeroes

Given an m x n integer matrix matrix, if an element is 0, set its entire row and column to 0's.

You must do it in place.

```python
class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        col0 = False
        row0 = False
        R = len(matrix)
        C = len(matrix[0])
        for r in range(R):
            if matrix[r][0] == 0:
                col0 = True
                break
        for c in range(C):
            if matrix[0][c] == 0:
                row0 = True
                break
        for r in range(1,R):
            for c in range(1,C):
                if matrix[r][c] == 0:
                    matrix[0][c] = 0
                    matrix[r][0] = 0

        for r in range(1,R):
            for c in range(1,C):
                if matrix[0][c] == 0 or matrix[r][0] == 0:
                    matrix[r][c] = 0

        if col0:
            for r in range(R):
                matrix[r][0] = 0
        if row0:
            for c in range(C):
                matrix[0][c] = 0
```

240. Search a 2D Matrix II

Write an efficient algorithm that searches for a value target in an m x n integer matrix matrix. This matrix has the following properties:

Integers in each row are sorted in ascending from left to right.
Integers in each column are sorted in ascending from top to bottom.


```python
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        R = len(matrix)
        C = len(matrix[0])
        row=0
        col=C-1
        while row < R and col >-1:
            val = matrix[row][col]
            if target == val:
                return True
            elif target > val:
                row += 1
            else:
                col -= 1
        return False
```

3. Longest Substring Without Repeating Characters

Given a string s, find the length of the longest 
substring
 without repeating characters.

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        chPos = dict()
        start = 0
        res = 0
        for i,c in enumerate(s):
            if c in chPos:
                res = max(res, i-start)
                newStart = chPos[c]+1
                for j in range(start, newStart):
                    chPos.pop(s[j])
                start = newStart
            chPos[c] = i
        res = max(res, len(s)-start)
        return res

```

76. Minimum Window Substring

Given two strings s and t of lengths m and n respectively, return the minimum window 
substring
 of s such that every character in t (including duplicates) is included in the window. If there is no such substring, return the empty string "".

The testcases will be generated such that the answer is unique.



```python
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        
        resStart = 0
        length = math.inf
        targetCnt = collections.Counter(t)
        windowCnt = collections.Counter()
        targetTotal = len(targetCnt)
        windowTotal = 0

        L = len(s)
        found = False
        start = 0
        for cur in range(L):
            ch = s[cur]
            windowCnt[ch] += 1
            if windowCnt[ch] == targetCnt[ch]:
                windowTotal += 1
                while windowTotal == targetTotal and start <= cur:
                    found = True
                    if length > cur-start+1:
                        length = cur-start+1
                        resStart = start
                    c = s[start]
                    windowCnt[c] -= 1
                    if windowCnt[c] < targetCnt[c]:
                        windowTotal -= 1
                    start += 1
            
        return s[resStart:resStart+length] if found else ''


```

239. Sliding Window Maximum

You are given an array of integers nums, there is a sliding window of size k which is moving from the very left of the array to the very right. You can only see the k numbers in the window. Each time the sliding window moves right by one position.

Return the max sliding window.



```python
# self
from collections import deque
class Solution:
    def maxSlidingWindow(self, nums: 'List[int]', k: 'int') -> 'List[int]':
        if not nums or not k:
            return []
        if k == 1:
            return nums 
        q = collections.deque()
        res = []
        for i, num in enumerate(nums):
            if i >= k:
                while q and q[0] <= i-k:
                    q.popleft()
            while q and nums[q[-1]] < num:
                q.pop()
            q.append(i)
            if i>=k-1:
                res.append(nums[q[0]])
        return res
```

438. Find All Anagrams in a String

Given two strings s and p, return an array of all the start indices of p's anagrams in s. You may return the answer in any order.

An Anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

```python
class Solution:
    def findAnagrams(self, s: str, p: str) -> List[int]:

        res = []
        Ls = len(s)
        Lp = len(p)
        if Lp > Ls:
            return res
        targetCnt = collections.Counter(p)
        windowCnt = collections.Counter()
        for i in range(Ls):
            if i >= Lp:
                windowCnt[s[i-Lp]] -= 1
            windowCnt[s[i]] += 1
            if i >= Lp-1:
                if windowCnt == targetCnt:
                    res.append(i-Lp+1)
        return res
        
        
```

20. Valid Parentheses

Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.

An input string is valid if:

Open brackets must be closed by the same type of brackets.
Open brackets must be closed in the correct order.
Every close bracket has a corresponding open bracket of the same type.

```python
class Solution:
    def isValid(self, s: str) -> bool:
        lefts = '{[('
        rights = '}])'
        r2l = dict()
        for l,r in zip(lefts, rights):
            r2l[r] = l
        stack = []
        for c in s:
            if c in lefts:
                stack.append(c)
            else:
                if stack and stack[-1] == r2l[c]:
                    stack.pop()
                else:
                    return False
        return not stack
        
```

84. Largest Rectangle in Histogram

Given an array of integers heights representing the histogram's bar height where the width of each bar is 1, return the area of the largest rectangle in the histogram.

```python
class Solution:
    def largestRectangleArea(self, heights: List[int]) -> int:
        stack = [-1] # initial marker 
        res = 0
        for i, h in enumerate(heights + [0]): # add 0 to remove all 
            while stack[-1] != -1 and heights[stack[-1]] >= h:
                # strictly increasing 
                hei = heights(stack.pop()) 
                # between left one and this hei, all bigger or equal to this height, if any smaller, it should be left one
                # between this hei and i, all bigger than this height, if any smaller than hei, we can not see hei here,
                width = i-stack[-1]-1
                res = max(res, hei*width)
            stack.append(i)
        return res
```

155. Min Stack

Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

Implement the MinStack class:

MinStack() initializes the stack object.
void push(int val) pushes the element val onto the stack.
void pop() removes the element on the top of the stack.
int top() gets the top element of the stack.
int getMin() retrieves the minimum element in the stack.
You must implement a solution with O(1) time complexity for each function.

```python
class MinStack:

    def __init__(self):
        self.stack = []
        self.minval = []

        

    def push(self, val: int) -> None:
        self.stack.append(val)
        if not self.minval or self.minval[-1] >= val:
            self.minval.append(val)
        
        

    def pop(self) -> None:
        if not self.stack:
            return
        val = self.stack.pop()
        if self.minval[-1] == val:
            self.minval.pop()
        

    def top(self) -> int:
        return self.stack[-1]
        

    def getMin(self) -> int:
        return self.minval[-1]
        


# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(val)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```

394. Decode String

Given an encoded string, return its decoded string.

The encoding rule is: k[encoded_string], where the encoded_string inside the square brackets is being repeated exactly k times. Note that k is guaranteed to be a positive integer.

You may assume that the input string is always valid; there are no extra white spaces, square brackets are well-formed, etc. Furthermore, you may assume that the original data does not contain any digits and that digits are only for those repeat numbers, k. For example, there will not be input like 3a or 2[4].

The test cases are generated so that the length of the output will never exceed 105.



```python
class Solution:
    def decodeString(self, s: str) -> str:
        strStack = [] # str before this num
        numStack = []
        curStr = ''
        curNum = 0
        
        for c in s:
            if c.isdigit():
                curNum *=10
                curNum += int(c)
            elif c.isalpha():
                curStr += c
            elif c == '[':
                strStack.append(curStr)
                numStack.append(curNum)
                curNum = 0
                curStr = ''
            elif c == ']':
                preStr = strStack.pop()
                preNum = numStack.pop()
                curStr = preStr + curStr * preNum

        return curStr


```

739. Daily Temperatures

Given an array of integers temperatures represents the daily temperatures, return an array answer such that answer[i] is the number of days you have to wait after the ith day to get a warmer temperature. If there is no future day for which this is possible, keep answer[i] == 0 instead.

```python
class Solution:
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        L = len(temperatures)
        res = [0]*L
        stack = [] # non increasing stack
        for i in range(L):
            while stack and temperatures[stack[-1]] < temperatures[i]:
                j = stack.pop()
                res[j] = i-j
            stack.append(i)
        return res

        
```

11. Container With Most Water

You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container.

```python
class Solution:
    def maxArea(self, height: List[int]) -> int:
        res = 0
        L = len(height)
        start = 0
        end = L-1
        while start < end:
            h1 = height[start]
            h2 = height[end]
            h = min(h1, h2)
            res = max(res, h * (end-start))
            if h1 == h: # h1 <= h2
                start += 1
            else:
                end -= 1
        return res
```

15. 3Sum

Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

Notice that the solution set must not contain duplicate triplets.

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        L = len(nums)
        res = []
        for i in range(L-2):
            if i>0 and nums[i]==nums[i-1]: # same first number only choose 1st one
                continue
            target = -nums[i]
            s, e = i+1, L-1
            while s < e:
                v1 = nums[s]
                v2 = nums[e]
                v = v1+v2
                if v == target:
                    res.append([-target,v1,v2])
                    while nums[s] == v1 and s<e: # same second number
                        s+=1
                elif v < target:
                    s+=1
                else:
                    e-=1
        return res
                
```
