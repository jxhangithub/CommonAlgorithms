```python
# 99. Recover Binary Search Tree

# You are given the root of a binary search tree (BST), where the values of exactly two nodes of the tree were swapped by mistake.
# Recover the tree without changing its structure.

class Solution:
    def recoverTree(self, root: TreeNode) -> None:
        """
        Do not return anything, modify root in-place instead.
        """
        
        stack = []
        x, y = None, None
        prev = None
        while stack or root:
            
            if root:
                stack.append(root)
                root = root.left
            else:
                
                node = stack.pop()
                if prev:
                    if prev.val > node.val:
                        y = node
                        if not x:
                            x = prev
                        else:
                            break # if swap positions are not adjacent, early stop 
                
                prev = node
                root = node.right
                
        x.val, y.val = y.val, x.val
        
        return
```

```python
# 286. walls and gates


class Solution:
    def wallsAndGates(self, rooms: List[List[int]]) -> None:
        """
        Do not return anything, modify rooms in-place instead.
        """
        if not rooms:
            return rooms
        H = len(rooms)
        W = len(rooms[0])
        queue = collections.deque()
        for i in range(H):
            for j in range(W):
                if rooms[i][j] == 0:
                    queue.append((i, j, 0))
                    
                    
        while queue:
            x, y, dst = queue.popleft()
            
            for dx, dy in (1, 0), (-1, 0), (0, -1), (0, 1):
                newX = x + dx
                newY = y + dy
                
                if 0 <= newX < H and 0 <= newY < W:
                    if rooms[newX][newY] == 2147483647:
                        rooms[newX][newY] = dst + 1
                        queue.append((newX, newY, dst + 1))
```

```python
### 363. Max Sum of Rectangle No Larger Than K

# Given a non-empty 2D matrix matrix and an integer k, find the max sum of a rectangle in the matrix such that its sum is no larger than k.

class Solution:
    def maxSumSubmatrix(self, matrix: List[List[int]], k: int) -> int:
        
        H = len(matrix)
        W = len(matrix[0])
        preSum = [[0] * (W + 1) for _ in range(H + 1)]
        
        for i in range(1, H + 1):
            for j in range(1, W + 1):
                preSum[i][j] = matrix[i - 1][j - 1] + preSum[i - 1][j] + preSum[i][j - 1] - preSum[i - 1][j - 1]
        answer = -float('inf')
        for lo in range(H + 1):
            for hi in range(lo + 1, H + 1):
                seen = []
                for i in range(W + 1):
                    temp = preSum[hi][i] - preSum[lo][i]  # row: [lo,hi-1], col: [0, i-1]
                    idx = bisect.bisect_left(seen, temp - k) # find the minimum number satisfies >= temp - k
                    if idx < len(seen):
                        answer = max(answer, temp - seen[idx])
                        if answer == k: return k
                    bisect.insort(seen, temp)
        return answer
```

```python
# 149. max points on a line 
    
# self
class Solution:
    def maxPoints(self, points: List[List[int]]) -> int:
        if not points:
            return 0
        n = len(points)
        result = 1
        for i in range(n):
            overlapCount = 0
            slopCnt = collections.defaultdict(int)
            start = points[i]
            for j in range(i + 1, n):
                end = points[j]
                if start == end:
                    overlapCount += 1
                    continue
                diff_y = end[1] - start[1]
                diff_x = end[0] - start[0]
                slope = diff_y / diff_x if diff_x != 0 else float('inf')
                slopCnt[slope] += 1  # if not integer, we can use[0, slope] for float and [1,0]for vertical line 
                result = max(result, slopCnt[slope] + 1 + overlapCount)
        return result
```
