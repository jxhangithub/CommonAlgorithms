### 1.2Sum (sorted array)
Given an array of integers that is already sorted in ascending order, find two numbers such that they add up to a specific target number.
The function twoSum should return indices of the two numbers such that they add up to the target,where index1 must be less than index2.

```python
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        
        left, right = 0, len(numbers) - 1
        
        while left < right: # cannot be left <= right, it would use one number twice
            
            if numbers[left] + numbers[right] < target: 
                left += 1
            elif numbers[left] + numbers[right] > target:
                right -= 1
            else:
                return [left + 1, right + 1]
            
        return [0, 0]
```

### 532. K-diff Pairs in an Array (unique pairs)
Given an array of integers and an integer k, you need to find the number of unique k-diff pairs in the array.
Here a k-diff pair is defined as an integer pair (i, j), where i and j are both numbers in the array and 
their absolute difference is k.### 532. K-diff Pairs in an Array
Given an array of integers and an integer k, you need to find the number of unique k-diff pairs in the array.
Here a k-diff pair is defined as an integer pair (i, j), where i and j are both numbers in the array and 
their absolute difference is k.

```python
# two pointer O(NlogN) due to sorting

class Solution:
    def findPairs(self, nums: List[int], k: int) -> int:
        
        if k < 0:
            return 0
        
        nums.sort()
        
        slow = 0
        fast = 1
        count = 0
        
        while fast < len(nums):
            
            if nums[fast] - nums[slow] < k:
                fast += 1
            elif nums[fast] - nums[slow] > k:
                slow += 1
            else: # nums[fast] - nums[slow] == k
                if fast == slow: # to meet the requirement of fast > slow
                    fast += 1
                else:
                    count += 1
                    fast += 1
                    # prevent duplicates
                    while fast < len(nums) and nums[fast] == nums[fast -1]:
                        fast += 1
        return count
    
# two pointer, more elegant solution
# O(NlogN + N)
class Solution:
    def findPairs(self, nums: List[int], k: int) -> int:
        
        nums.sort()
        
        slow = 0
        fast = 1
        count = 0
        while fast < len(nums):
            
            if slow >= fast:
                fast += 1
            elif nums[fast] - nums[slow] == k:
                count += 1
                slow += 1
                # to avoid duplicate pairs
                while slow < len(nums) and nums[slow] == nums[slow - 1]:
                    slow += 1
            elif nums[fast] - nums[slow] > k:
                slow += 1
            elif nums[fast] - nums[slow] < k:
                fast += 1
                
        return count
    
# use hashmap
# O(N)
class Solution:
    def findPairs(self, nums: List[int], k: int) -> int:
        
        counter = Counter(nums)
        ans = 0
        for x in counter:
            # we only track x + k
            # if we include x - k, there will be double counting
            if k > 0 and x + k in counter:
                ans += 1
            elif k == 0 and counter[x] > 1:
                ans += 1
        return ans
```

### 2006. Count Number of Pairs With Absolute Difference K (allow duplicates)

```python
# two pointer
class Solution:
    def countKDifference(self, nums: List[int], k: int) -> int:
        
        nums.sort()
        
        slow = 0
        fast = 1
        count = 0
        while fast < len(nums):
            
            if slow >= fast:
                fast += 1
            elif nums[fast] - nums[slow] == k:

                # here we need to count duplicates in slow and fast
                slow_count = 1
                slow += 1
                while slow < len(nums) and nums[slow] == nums[slow - 1]:
                    slow_count += 1
                    slow += 1
                    
                fast_count = 1
                fast += 1
                while fast < len(nums) and nums[fast] == nums[fast - 1]:
                    fast_count += 1
                    fast += 1
                
                count += slow_count * fast_count
                
            elif nums[fast] - nums[slow] > k:
                slow += 1
            elif nums[fast] - nums[slow] < k:
                fast += 1
                
        return count


# hashmap
class Solution:
    def countKDifference(self, nums: List[int], k: int) -> int:
        
        ans = 0
        count = Counter(nums)
        
        for x in count:
            
            if x + k in count:
                ans += count[x] * count[x + k]
                
        return ans
```

### 244. Shortest Word Distance II

Design a class which receives a list of words in the constructor, and implements a method that takes two words word1 and word2 and return the shortest distance between these two words in the list. Your method will be called repeatedly many times with different parameters. 

```python
class WordDistance:

    def __init__(self, words: List[str]):
        self.pos = collections.defaultdict(list)
        for i, w in enumerate(words):
            self.pos[w].append(i)
        
        self.memo = {}

    def shortest(self, word1: str, word2: str) -> int:
        if word1 > word2:
            word1, word2 = word2, word1
        if (word1, word2) in self.memo:
            return self.memo[(word1, word2)]
        
        list1 = self.pos[word1]
        list2 = self.pos[word2]
        p1, p2 = 0, 0
        minDist = float('inf')
        while p1 < len(list1) and p2 < len(list2):
            minDist = min(minDist, abs(list1[p1] - list2[p2]))
            if list1[p1] < list2[p2]:
                p1 += 1
            else:
                p2 += 1
        self.memo[(word1, word2)] = minDist
        return minDist
```

### 905. Sort Array By Parity
Given an array A of non-negative integers, return an array consisting of all the even elements of A, followed by all the odd elements of A.

You may return any answer array that satisfies this condition.

```python
class Solution:
    def sortArrayByParity(self, A: List[int]) -> List[int]:
        
        left, right = 0, len(A) - 1
        
        while left < right:
            
            if A[left] & 1 == 1 and A[right] & 1 == 0:
                A[left], A[right] = A[right], A[left]
            if A[left] & 1 == 0:
                left += 1
            if A[right] & 1 == 1:
                right -= 1
                
        return A
```

### 125. Valid Palindrome

```python
125. Valid Palindrome

Given a string, determine if it is a palindrome, considering only alphanumeric characters and ignoring cases.

Note: For the purpose of this problem, we define empty string as valid palindrome.

class Solution:
    def isPalindrome(self, s: str) -> bool:
        
        left, right = 0, len(s) - 1
        
        
        while left < right:
            
            while left < right and not s[left].isalnum():
                left += 1
            while left < right and not s[right].isalnum():
                right -= 1
            if s[left].lower() == s[right].lower():
                left, right = left + 1, right - 1
            else:
                return False
            
        return True
        
```

### 11. Container With Most Water
Given n non-negative integers a1, a2, ..., an , where each represents a point at coordinate (i, ai). n vertical lines are drawn such that the two endpoints of line i is at (i, ai) and (i, 0). Find two lines, which together with x-axis forms a container, such that the container contains the most water.

Note: You may not slant the container and n is at least 2.

```python
  
class Solution:
    def maxArea(self, height: List[int]) -> int:
        
        res = 0
        
        left, right = 0, len(height) - 1
        # The intuition is double loop with early stop and start condition
        while left < right:
            
            res = max(res, (right - left) * min(height[left], height[right]))
            
            if height[left] < height[right]:# no need to move right since moving right can only reduce the result
                left += 1
            else:
                right -= 1
                
        return res
```

### 360. Sort Transformed Array

Given a sorted array of integers nums and integer values a, b and c. Apply a quadratic function of the form f(x) = ax2 + bx + c to each element x in the array.

The returned array must be in sorted order.

Expected time complexity: O(n)

Example 1:

Input: nums = [-4,-2,2,4], a = 1, b = 3, c = 5
Output: [3,9,15,33]
Example 2:

Input: nums = [-4,-2,2,4], a = -1, b = 3, c = 5
Output: [-23,-5,1,7]

```python
class Solution:
    def sortTransformedArray(self, nums: List[int], a: int, b: int, c: int) -> List[int]:
        
        nums = [a * e * e + b * e +c for e in nums]
        
        left, right = 0, len(nums) - 1
        
        res = [0] * len(nums)
        
        
        # assume a >= 0
        # if a = 0 and we have a linear equation, the following also work
        # when a >=0, we have two reverse sorted array starting from each end
        if a >=0:
            idx = right
            while left <= right:

                if nums[left] > nums[right]:
                    res[idx] = nums[left]
                    left += 1
                else:
                    res[idx] = nums[right]
                    right -= 1

                idx -= 1
        else:
            # when a <0, we have two sorted arrays starting from each end
            idx = 0
            while left <= right:

                if nums[left] < nums[right]:
                    res[idx] = nums[left]
                    left += 1
                else:
                    res[idx] = nums[right]
                    right -= 1

                idx += 1
            
        return res
```

### 1498. Number of Subsequences That Satisfy the Given Sum Condition

You are given an array of integers nums and an integer target.

Return the number of non-empty subsequences of nums such that the sum of the minimum and maximum element on it is less or equal to target. Since the answer may be too large, return it modulo 109 + 7.

```python
class Solution:
    def numSubseq(self, nums: List[int], target: int) -> int:
        
        nums.sort()
        
        left, right = 0, len(nums) - 1
        ans = 0
        while left <= right:
            temp_sum = nums[left] + nums[right]
            if temp_sum <= target:
                ans += pow(2, (right - left), 1000000000 + 7)
                left += 1
            else:
                right -= 1
                
        return int(ans % (1e9 + 7))
```

### 915. Partition Array into Disjoint Intervals

Given an integer array nums, partition it into two (contiguous) subarrays left and right so that:

Every element in left is less than or equal to every element in right.
left and right are non-empty.
left has the smallest possible size.
Return the length of left after such a partitioning.

Test cases are generated such that partitioning exists.

```python
class Solution:
    def partitionDisjoint(self, nums: List[int]) -> int:
        
        left_max = [float('-inf')] * len(nums)
        
        for i in range(len(nums)):
            if i > 0:
                left_max[i] = max(left_max[i - 1], nums[i])
            else:
                left_max[i] = nums[i]
        
        right_min = [float('inf')] * len(nums)
        
        for i in range(len(nums) - 1, -1, -1):
            if i < len(nums) - 1:
                right_min[i] = min(right_min[i + 1], nums[i])
            else:
                right_min[i] = nums[i]
                
        
        for i in range(len(nums) - 1):
            if left_max[i] <= right_min[i + 1]:
                return i + 1
```

### 38. Count and Say

The count-and-say sequence is a sequence of digit strings defined by the recursive formula:

countAndSay(1) = "1"
countAndSay(n) is the way you would "say" the digit string from countAndSay(n-1), which is then converted into a different digit string.
To determine how you "say" a digit string, split it into the minimal number of groups so that each group is a contiguous section all of the same character. Then for each group, say the number of characters, then say the character. To convert the saying into a digit string, replace the counts with a number and concatenate every saying.

For example, the saying and conversion for digit string "3322251":

```python
class Solution:
    def countAndSay(self, n: int) -> str:

        # write your code here
        def generate(s):
            res = []
            slow, fast = 0, 0
            while slow < len(s) and fast < len(s):

                while fast < len(s) and s[fast] == s[slow]:
                    fast += 1
                
                res.append(str(fast - slow) + s[slow])

                slow = fast

            return ''.join(res)

        s = '1'
        for _ in range(n - 1):
            s = generate(s)

        return ''.join(s)       
```

### 1712. Ways to Split Array Into Three Subarrays

A split of an integer array is good if:

The array is split into three non-empty contiguous subarrays - named left, mid, right respectively from left to right.
The sum of the elements in left is less than or equal to the sum of the elements in mid, and the sum of the elements in mid is less than or equal to the sum of the elements in right.
Given nums, an array of non-negative integers, return the number of good ways to split nums. As the number may be too large, return it modulo 109 + 7.

```python
class Solution:
    def waysToSplit(self, nums: List[int]) -> int:
        
        prefix = [0]
        for num in nums:
            prefix.append(prefix[-1] + num)
            
        ans = j = k = 0
        
        for i in range(1, len(nums)):
            
            j = max(j, i + 1)
            
            # j is the left boundary, j is monotonic increasing as i increases
            while j < len(nums) and prefix[j] - prefix[i] < prefix[i]:
                j += 1
            
            k = max(k, j)
            # k is the right boundary, k is monotonic increasing as i increases
            while k < len(nums) and prefix[-1] - prefix[k] >= prefix[k] - prefix[i]:
                k += 1
            
            ans += k - j
            
        return int(ans % (1e9 + 7))
```

### 1868. Product of Two Run-Length Encoded Arrays

Run-length encoding is a compression algorithm that allows for an integer array nums with many segments of consecutive repeated numbers to be represented by a (generally smaller) 2D array encoded. Each encoded[i] = [vali, freqi] describes the ith segment of repeated numbers in nums where vali is the value that is repeated freqi times.

For example, nums = [1,1,1,2,2,2,2,2] is represented by the run-length encoded array encoded = [[1,3],[2,5]]. Another way to read this is "three 1's followed by five 2's".

```python
class Solution:
    def findRLEArray(self, encoded1: List[List[int]], encoded2: List[List[int]]) -> List[List[int]]:
        
        
        res = []
        
        p1, p2 = 0, 0
        while p1 < len(encoded1) and p2 < len(encoded2):
            
            val1, freq1 = encoded1[p1]
            val2, freq2 = encoded2[p2]
            
            min_freq = min(freq1, freq2)
            product_val = val1 * val2
            
            encoded1[p1][1] -= min_freq
            if encoded1[p1][1] == 0:
                p1 += 1
                
            encoded2[p2][1] -= min_freq
            if encoded2[p2][1] == 0:
                p2 += 1
            
            
            if not res or res[-1][0] != product_val:
                res.append([product_val, min_freq])
            else:
                res[-1][1] += min_freq
        
        return res
```

### 900. RLE Iterator

We can use run-length encoding (i.e., RLE) to encode a sequence of integers. In a run-length encoded array of even length encoding (0-indexed), for all even i, encoding[i] tells us the number of times that the non-negative integer value encoding[i + 1] is repeated in the sequence.

For example, the sequence arr = [8,8,8,5,5] can be encoded to be encoding = [3,8,2,5]. encoding = [3,8,0,9,2,5] and encoding = [2,8,1,8,2,5] are also valid RLE of arr.
Given a run-length encoded array, design an iterator that iterates through it.

Implement the RLEIterator class:

RLEIterator(int[] encoded) Initializes the object with the encoded array encoded.
int next(int n) Exhausts the next n elements and returns the last element exhausted in this way. If there is no element left to exhaust, return -1 instead.

```python
class RLEIterator:

    def __init__(self, encoding: List[int]):
        
        self.encoding = encoding
        self.idx = 0
        self.used = 0
        

    def next(self, n: int) -> int:
        
        out = None
        
        while self.idx < len(self.encoding):
            
            if self.used < self.encoding[self.idx]:
                taken = min(self.encoding[self.idx] - self.used, n)
                n -= taken
                out = self.encoding[self.idx + 1]
                self.used += taken
                
                if n == 0:
                    break
                else:
                    self.idx += 2
                    self.used = 0
            else:
                self.idx += 2
                self.used = 0

        if n == 0:
            return out
        else:
            return -1
```

```python

class Solution:
    def numSubarrayProductLessThanK(self, nums: List[int], k: int) -> int:
        if k <= 1:
            return 0
        cum_prod = 1
        res = 0
        slow = 0
        for fast in range(len(nums)):
            
            cum_prod *= nums[fast]
            
            # safe guard, cum_prod always >= 1
            # slow <= fast is not necessary
            while slow <= fast and cum_prod >= k:
                cum_prod //= nums[slow]
                slow += 1
            
            res += fast - slow + 1
            
        return res
```

### 438. Find All Anagrams in a String
Given a string s and a non-empty string p, find all the start indices of p's anagrams in s.

Strings consists of lowercase English letters only and the length of both strings s and p will not be larger than 20,100.

The order of output does not matter.


```python
class Solution:
    def findAnagrams(self, s: str, p: str) -> List[int]:
        
        p_count = collections.Counter(p)
        s_count = {}
        def valid(s_count, p_count):
            
            for k, v in p_count.items():
                if v > s_count.get(k, 0):
                    return False
            return True
        
        ans = []
        for i in range(len(s)):
            
            s_count[s[i]] = s_count.get(s[i], 0) + 1
            
            if i >= len(p):
                s_count[s[i - len(p)]] -= 1
            
            if i >= len(p) - 1:
                if valid(s_count, p_count):
                    ans.append(i - len(p) + 1)
        return ans
```

### 1248. Count Number of Nice Subarrays

Given an array of integers nums and an integer k. A subarray is called nice if there are k odd numbers on it.

Return the number of nice sub-arrays.

```python
class Solution:
    def numberOfSubarrays(self, nums: List[int], k: int) -> int:
        
        slow = 0
        count = 0
        oddCount = 0
        fast = 0
        while fast < len(nums):
            oddCount += nums[fast] & 1 == 1
            if oddCount == k:
                nextOddIdx = fast + 1
                while nextOddIdx < len(nums) and nums[nextOddIdx] & 1 == 0:
                    nextOddIdx += 1
            
                while oddCount == k:
                    count += (nextOddIdx - fast)
                    oddCount -= nums[slow] & 1 == 1
                    slow += 1
                fast = nextOddIdx
            else:
                fast += 1
        return count
    
    
class Solution:
    def numberOfSubarrays(self, nums: List[int], k: int) -> int:
        
        slow = 0
        count = 0
        oddCount = 0
        subCount = 0
        fast = 0
        for fast in range(len(nums)):
            if nums[fast] & 1 == 1:
                oddCount += 1
                subCount = 0
            
            
            while oddCount == k:
                subCount += 1
                oddCount -= nums[slow] & 1 == 1
                slow += 1
            count += subCount
            
        return count
    
# view the array as a binary array and convert to subarray sum k problem
# O(N) space
class Solution:
    def numberOfSubarrays(self, nums: List[int], k: int) -> int:
        
        presum_dict = {0:1}
        presum = 0
        ans = 0
        for num in nums:
            presum += (num % 2 == 1)
            ans += presum_dict.get(presum - k, 0)
            presum_dict[presum] = presum_dict.get(presum, 0) + 1
        return ans
```

### 1438. Longest Continuous Subarray With Absolute Diff Less Than or Equal to Limit
Given an array of integers nums and an integer limit, return the size of the longest continuous subarray such that the absolute difference between any two elements is less than or equal to limit.

In case there is no subarray satisfying the given condition return 0.

```python

# use monotone queues to maintain running max and min
class Solution:
    def longestSubarray(self, nums: List[int], limit: int) -> int:
        
        minQ = collections.deque()
        maxQ = collections.deque()
        slow = 0
        ans = 0
        for fast in range(len(nums)):
            
            while minQ and nums[minQ[-1]] > nums[fast]:
                minQ.pop()
            minQ.append(fast)
            while maxQ and nums[maxQ[-1]] < nums[fast]:
                maxQ.pop()
            maxQ.append(fast)
            
            while maxQ and minQ and nums[maxQ[0]] - nums[minQ[0]] > limit:
                while minQ and minQ[0] <= slow:
                    minQ.popleft()
                while maxQ and maxQ[0] <= slow:
                    maxQ.popleft()
                slow += 1
                
            ans = max(ans, fast - slow + 1)
            
        return ans

    
# tree solution
class Solution {
public:
    int longestSubarray(vector<int>& nums, int limit) {
        
        int answer = 1;
        int slow = 0;
        multiset<int> sorted;
        for(int fast = 0; fast < nums.size(); ++fast) {
            sorted.insert(nums[fast]);
            while (*sorted.rbegin() - *sorted.begin() > limit) {
                sorted.erase(sorted.find(nums[slow]));
                slow += 1;
            } 
            answer = max(answer, fast - slow + 1);
        }
        return answer;
    }
};
```

### 862. Shortest Subarray with Sum at Least K (contains both positive and negative values)
Return the length of the shortest, non-empty, contiguous subarray of A with sum at least K.

If there is no non-empty subarray with sum at least K, return -1.

```python
# use a monotine queue to maintain an increasing presum sequence
class Solution:
    def shortestSubarray(self, A: List[int], K: int) -> int:
        
        queue = collections.deque([(0, -1)])
        answer = float('inf')
        preSum = 0
        for i in range(len(A)):
            
            preSum += A[i]  
            # if there is a latest idx with a smaller presum, queue[-1][0] will not be used 
            while queue and preSum <= queue[-1][0]:
                queue.pop()
            queue.append((preSum, i))
            
            while preSum - queue[0][0] >= K:
                answer = min(answer, i - queue[0][1])
                # if queue[0][i] is used, it will not be used anymore, so pop it
                queue.popleft()
        return answer if answer != float('inf') else -1
```

### 795. Number of Subarrays with Bounded Maximum
We are given an array A of positive integers, and two positive integers L and R (L <= R).

Return the number of (contiguous, non-empty) subarrays such that the value of the maximum array element in that subarray is at least L and at most R.

```python
# use a monotine queue to maintain an decraseing sequence to enable retrive sliding max value
class Solution:
    def numSubarrayBoundedMax(self, A: List[int], L: int, R: int) -> int:
        if not A:
            return 0
        
        def atMost(R):
            res = 0
            queue = collections.deque()
            slow = 0
            for fast in range(len(A)):
                while queue and A[fast] >= A[queue[-1]]:
                    queue.pop()
                queue.append(fast)

                while queue and A[queue[0]] > R:
                    if slow == queue[0]:
                        queue.popleft()
                    slow += 1

                res += fast - slow + 1
            return res
        return atMost(R) - atMost(L - 1)
                

# a much simpler solution
class Solution:
    def numSubarrayBoundedMax(self, A: List[int], L: int, R: int) -> int:
        if not A:
            return 0
        
        def atMost(R):
            answer = 0
            
            count = 0
            for fast in range(len(A)):
                
                if A[fast] > R:
                    # reset to 0; restart counting
                    count = 0
                else:
                    count += 1
                    
                answer += count
            return answer
        return atMost(R) - atMost(L - 1)
            
                
```

### 142. Linked List Cycle II
Given a linked list, return the node where the cycle begins. If there is no cycle, return null.

To represent a cycle in the given linked list, we use an integer pos which represents the position (0-indexed) in the linked list where tail connects to. If pos is -1, then there is no cycle in the linked list.

Note: Do not modify the linked list.

```python
class Solution:
    def detectCycle(self, head: ListNode) -> ListNode:
        
        if not head:
            return head
        
        
        slow = head
        fast = head
        intersection = None
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                intersection = slow
                break
                
        if not intersection:
            return None
        
        p1 = head
        p2 = intersection
        
        while p1 != p2:
            p1 = p1.next
            p2 = p2.next
            
        return p1
        
```

### 287. Find the Duplicate Number
Given an array nums containing n + 1 integers where each integer is between 1 and n (inclusive), prove that at least one duplicate number must exist. Assume that there is only one duplicate number, find the duplicate one.



```python
class Solution:
    def findDuplicate(self, nums: List[int]) -> int:
        
        slow = nums[0]
        fast = nums[0]
        
        while True:
            slow = nums[slow]
            fast = nums[nums[fast]]
            if slow == fast:
                break
            
        p1 = nums[0]
        p2 = slow
        
        while p1 != p2:
            p1 = nums[p1]
            p2 = nums[p2]
            
        return p1
```

It is guaranteed that no two availability slots of the same person intersect with each other. That is, for any two time slots [start1, end1] and [start2, end2] of the same person, either start1 > end2 or start2 > end1.


```python
class Solution:
    def minAvailableDuration(self, slots1: List[List[int]], slots2: List[List[int]], duration: int) -> List[int]:
        
        slots1.sort(key=lambda x: (x[0], x[1]))
        slots2.sort(key=lambda x: (x[0], x[1]))
        
        first, second = 0, 0
        
        while first < len(slots1) and second < len(slots2):
            
            p1 = slots1[first]
            p2 = slots2[second]
            overLap = min(p1[1], p2[1]) - max(p1[0], p2[0])
            
            if overLap >= duration:
                return [max(p1[0], p2[0]), max(p1[0], p2[0]) + duration]
            
            if p1[1] <= p2[1]:
                first += 1
            else:
                second += 1
                
                
        return []
                
```
