```python
# 1101. The Earliest Moment When Everyone Become Friends

# followup是如果允许friend之间cancel relationship 怎么做。(no compression, just use ordered connect like smaller is parent, bigger is child)

class UnionFind:
    def __init__(self, size):
        self.group = [i for i in range(size)]
        self.weight = [0]*size
        
    def find(self, x):
        if self.group[x] != x:
            self.group[x] = self.find(self.group[x])
        return self.group[x]
    
    def union(self, x, y):
        groupx = self.find(x)
        groupy = self.find(y)
        if groupx == groupy:
            return False
        if self.weight[groupx] >= self.weight[groupy]:
            self.group[groupy] = groupx
            self.weight[groupx] += self.group[groupy]
        else:
            self.group[groupx] = groupy 
            self.weight[groupy] += self.group[groupx]
        return True
    

class Solution:
    def earliestAcq(self, logs: List[List[int]], n: int) -> int:
        if not logs or not n:
            return -1
        logLen = len(logs)
        uf = UnionFind(n)
        cnt = n
        for t, x, y in sorted(logs):
            if uf.union(x,y):
                cnt -= 1
                if cnt == 1:
                    return t
        return -1
```

```python
class Solution:  # topological sort -> one valid path
    def topSort(self, graph):
        # write your code here
        node_to_indegree = self.get_indegree(graph)
        
        order = []
        queue = collections.deque([x for x, v in node_to_indegree.items() if v == 0])

        while queue:
            node = queue.popleft()
            order.append(node)
            for nb in node.neighbors:
                node_to_indegree[nb] -= 1
                if node_to_indegree[nb] == 0:
                    queue.append(nb)
        return order

    def get_indegree(self, graph):
        node_to_indegree = {x: 0 for x in graph}
        
        for node in graph:
            for nb in node.neighbors:
                node_to_indegree[nb] += 1
        return node_to_indegree
```

```python
# 314. Binary Tree Vertical Order Traversal


# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None
from collections import defaultdict
class Solution:
    def verticalOrder(self, root: TreeNode) -> List[List[int]]:
        if root is None:
            return []

        columnTable = defaultdict(list)
        min_column = max_column = 0
        queue = deque([(root, 0)])

        while queue:
            node, column = queue.popleft()

            if node is not None:
                columnTable[column].append(node.val)
                min_column = min(min_column, column)
                max_column = max(max_column, column)

                queue.append((node.left, column - 1))
                queue.append((node.right, column + 1))

        return [columnTable[x] for x in range(min_column, max_column + 1)]
```

```python
# 213. House Robber II
class Solution:
    def rob(self, nums: List[int]) -> int:
        if len(nums) == 0 or nums is None:
            return 0

        if len(nums) == 1:
            return nums[0]

        return max(self.rob_simple(nums[:-1]), self.rob_simple(nums[1:]))

    def rob_simple(self, nums: List[int]) -> int:
        t1 = 0
        t2 = 0
        for current in nums:
            temp = t1  
            t1 = max(current + t2, t1)    # previous max vs current + prepre max(no pre house)
            t2 = temp  # last time max

        return t1
```

```python

# 1197. Minimum Knight Moves 

# family tree。 找两个人最近的祖先。比较trick的地方就是，一个的祖先是颗倒挂树，会指数增长。三哥要求优化。

class Solution:
    def minKnightMoves(self, x: int, y: int) -> int:
        x, y = abs(x), abs(y)
        
        if (x, y) == (0, 0):
            return 0

        queue = [[(0, 0)], [(x, y)]]
        visited = [set([(0, 0)]), set([(x, y)])]
               
        Delta = [(2, 1), (1, 2), (2, -1), (1, -2), (-1, -2), (-2, -1), (-2, 1), (-1, 2)]
        
        step = 0
        
        while True:

            if len(queue[0]) <= len(queue[1]):
                idx = 0
            else:
                idx  = 1
            
            step += 1
            
            newQ = []
            for i, j in queue[idx]:
                for dx, dy in Delta:
                    newX, newY = i + dx, j + dy

                    if (newX, newY) in visited[1 - idx]: # if in the counterparty's territory
                        return step

                    if (newX, newY) not in visited[idx] and -2 <= newX <= x + 2 and -2 <= newY <= y + 2:
                        newQ.append((newX, newY))
                        visited[idx].add((newX, newY))
            queue[idx] = newQ
    
            
```

```python
# 1631. Path With Minimum Effort

# variation of Dijsktra algorithm
# in the min_heap, the key is the min effort to reach the node
# the effort however, it is not computed additively

class Solution:
    def minimumEffortPath(self, heights: List[List[int]]) -> int:
        
        res = {}  # this is just a visit set
        
        h = [(0, 0, 0)]
        
        while h:
            effort, x, y = heapq.heappop(h)
            # minimum entry for (max_effort of a route to current point)
            
            if (x, y) in res:
                continue
            
            # critical, avoid (x, y) being repeatedly visite
            res[(x, y)] = effort  # no need this value if only one target, if one to all we need this 
            
            if (x, y) == (len(heights) - 1, len(heights[0]) - 1):
                return effort
            
            for dx, dy in (-1, 0), (1, 0), (0, -1), (0, 1):
                new_x = x + dx
                new_y = y + dy
                
                if 0 <= new_x < len(heights) and 0 <= new_y < len(heights[0]):
                    effort2 = abs(heights[new_x][new_y] - heights[x][y])
                    # 4 potential directions come to new_x, new_y, 
                    # they are coming in order, but add the abs here may change the order
                    # so we need to finalize the result later after all in queue. 
                    if (new_x, new_y) not in res:
                        heapq.heappush(h, (max(effort, effort2),new_x, new_y))   
                        
            
```

```python

# 1293. Shortest Path in a Grid with Obstacles Elimination


# state is (position, remaining_actions)
class Solution:
    def shortestPath(self, grid: List[List[int]], k: int) -> int:
        
        H = len(grid)
        W = len(grid[0])
        
        queue = collections.deque([(0, 0, 0, k)])
        visited = set([(0, 0, k)])
        while queue:
            
            x, y, dist, remain = queue.popleft()
            if (x, y) == (H - 1, W - 1):
                return dist
            
            for dx, dy in (-1, 0), (1, 0), (0, -1), (0, 1):
                
                newX = x + dx
                newY = y + dy
                
                if 0 <= newX < H and 0 <= newY < W:
                    
                    
                    if grid[newX][newY] == 0 and (newX, newY, remain) not in visited:
                        
                        queue.append((newX, newY, dist + 1, remain))
                        visited.add((newX, newY, remain))
                    elif grid[newX][newY] == 1 and remain > 0 and (newX, newY, remain - 1) not in visited:
                        queue.append((newX, newY, dist + 1, remain - 1))
                        visited.add((newX, newY, remain - 1))
        
        return -1
```

```python
# 在一个2d空间中，有一些Vampire和mirror，input是这些vampire和镜子的row, column 
# 坐标和镜子方向（东南西北），设定是vampire在同一行或同一列面对自己的镜子中看到自己会embarrassed。
# 要求return 所有embarrassed vampire的坐标以及embarrassed的方向。
# 比如如果vampire在左边的镜子中看到自己，embarrassed方向就是西边。vampire对同类来说是透明的。
# 如果面向vampire的镜子被另一面相反方向的镜子挡住vampire就不会看到自己。


def mirrorVampire(grid):  # mn, ms, mw, me for mirror
    result = []
    H = len(grid)
    W = len(grid[0])
    for i in range(H):
        for j in range(W):
            if grid[i][j] == 'mn':
                for k in range(i-1, -1, -1):
                    if grid[k][j] == 'mn' or grid[k][j] == 'ms':
                        break
                    elif grid[k][j] == 'v':
                        result.append((k, j, 's'))
            elif grid[i][j] == 'ms':
                for k in range(i+1, H):
                    if grid[k][j] == 'mn' or grid[k][j] == 'ms':
                        break
                    elif grid[k][j] == 'v':
                        result.append((k, j, 'n'))
            elif grid[i][j] == 'me':
                for k in range(j+1, W):
                    if grid[i][k] == 'me' or grid[i][k] == 'mw':
                        break
                    elif grid[i][k] == 'v':
                        result.append((i,k, 'w'))
            elif grid[i][j] == 'mw':
                for k in range(j-1, -1, -1):
                    if grid[i][k] == 'me' or grid[i][k] == 'mw':
                        break
                    elif grid[i][k] == 'v':
                        result.append((i,k, 'e'))
    return result 

grid = [
    ['v', 'v', 'mw'],
    ['v', 'ms', 'mw'],
    ['me', 'mn', 'v']
]

print(mirrorVampire(grid))
            
                    
                        
        
```
