### 590. N-ary Tree Postorder Traversal

```python
590. N-ary Tree Postorder Traversal

Given an n-ary tree, return the postorder traversal of its nodes' values.

Nary-Tree input serialization is represented in their level order traversal, each group of children is separated by 
the null value (See examples).

"""
# Definition for a Node.
class Node:
    def __init__(self, val=None, children=None):
        self.val = val
        self.children = children
"""

# recursive
class Solution:
    def postorder(self, root: 'Node') -> List[int]:
        
        if not root:
            return []
        
        res = []
        for c in root.children:
            res.extend(self.postorder(c))
        res.append(root.val)
        
        return res
    
# iterative first version, mimicing the recursive process
class Solution:
    def postorder(self, root: 'Node') -> List[int]:
        
        stack = []
        res = []
        nextIdx = 0        
        
        while stack or root:
            
            if not root:
                root, nextIdx = stack.pop()
            
            if nextIdx < len(root.children): 
                stack.append((root, nextIdx + 1))
                root = root.children[nextIdx]
                nextIdx = 0
            else:# if it is a leaf or all childrens are done
                res.append(root.val)
                root = None
        
        return res
    
# iterative second version, using additional flag
class Solution:
    def postorder(self, root: 'Node') -> List[int]:
        
        stack = [(root, False)]
        res = []
        
        while stack:
            
            node, visited = stack.pop()
            if not node:
                continue
                
            if visited:
                res.append(node.val)
            else:
                stack.append((node, True))
                for c in reversed(node.children):
                    stack.append((c, False))
                    
        return res
```

### 102. Binary Tree Level Order Traversal

Given a binary tree, return the level order traversal of its nodes' values. (ie, from left to right, level by level).

For example:
Given binary tree [3,9,20,null,null,15,7],

```python
# BFS 
class Solution:
    def levelOrder(self, root: TreeNode) -> List[List[int]]:
        
        if not root:
            return []

        res = []        
        queue = [root]
        while queue:
            levelOrder = []            
            nextQueue = []
            for node in queue:
                levelOrder.append(node.val)
                if node.left:
                    nextQueue.append(node.left)
                if node.right:
                    nextQueue.append(node.right)
                    
            res.append(levelOrder)
            queue = nextQueue
            
        return res
    
DFS solution
class Solution:
    def levelOrder(self, root: TreeNode) -> List[List[int]]:
        
        if not root:
            return []

        res = []        
        stack = [(root, 0)]
        while stack:
            node, level = stack.pop()
            if len(res) == level:
                res.append([])
            res[level].append(node.val)
            # right first
            if node.right:
                stack.append((node.right, level + 1))
            
            if node.left:
                stack.append((node.left, level + 1))
```

Given a binary tree, return the zigzag level order traversal of its nodes' values. (ie, from left to right, then right to left for the next level and alternate between).

For example:
Given binary tree [3,9,20,null,null,15,7],

    /  \
   15   7

```python
class Solution:
    def zigzagLevelOrder(self, root: TreeNode) -> List[List[int]]:
        
        if not root:
            return []
        
        queue = [root]
        res = []
        
        while queue:
            
            newQ = []
            level = []
            
            for node in queue:
                level.append(node.val)
                if node.left:
                    newQ.append(node.left)
                if node.right:
                    newQ.append(node.right)
                    
            if len(res) % 2 == 0:
                res.append(level)
            else:
                res.append(level[::-1])
            queue = newQ
            
        return res
            
```

### 314. Binary Tree Vertical Order Traversal

Given a binary tree, return the vertical order traversal of its nodes' values. (ie, from top to bottom, column by column).

If two nodes are in the same row and column, the order should be from left to right.


```python
# BFS plus hashmap
class Solution:
    def verticalOrder(self, root: TreeNode) -> List[List[int]]:
        
        if not root:
            return []
        
        nodes = collections.defaultdict(list)
        
        queue = collections.deque([(0, root)])
        
        minCol = 0
        maxCol = 0
        while queue:
            
            col,  node = queue.popleft()
            nodes[col].append(node.val)
            if node.left:
                queue.append((col - 1, node.left))
                minCol = min(minCol, col - 1)            
            
            if node.right:
                queue.append((col + 1, node.right))
                maxCol = max(maxCol, col + 1)

        
        res = []
        for i in range(minCol, maxCol + 1):
            res.append(nodes[i])
        
        return res
            
            
            
                
```

### diagonal traversal of binary tree

```python
Example Input
Input 1:

            1
          /   \
         4      2
        / \      \
       8   5      3
          / \    /
         9   7  6
        
[1, 2, 3, 4, 5, 7, 6, 8, 9]


# BFS solution, similar to level by level traversal
# Note that the same slope level can be access by going right child
class Solution:
    # @param A : root node of tree
    # @return a list of integers
    def solve(self, root):

        if not root:
            return []
            
        queue = collections.deque([root])

        results = []
        while queue:
            
            node = queue.popleft()
            
            while node:
                results.append(node.val)
                if node.left:
                    queue.append(node.left)
                node = node.right
            
            
        return results

# DFS solution using additional hashmap to record slope level
class Solution:
    # @param A : root node of tree
    # @return a list of integers
    def solve(self, root):

        if not root:
            return []
            
        stack = [(root, 0)]
        maxLeftCount = 0
        resultsDict = collections.defaultdict(list)
        while stack:
            
            node, leftCount = stack.pop()
            resultsDict[leftCount].append(node.val)
            if leftCount > maxLeftCount:
                maxLeftCount = leftCount
            if node.right:
                stack.append((node.right, leftCount))
            if node.left:
                stack.append((node.left, leftCount + 1))
        
        results = []
        for k in range(maxLeftCount + 1):
            results += resultsDict[k]
        return results
```

### 637. Average of Levels in Binary Tree
Given a non-empty binary tree, return the average value of the nodes on each level in the form of an array.

```python
/**
 * Definition for a binary tree node.
 * struct TreeNode {
 *     int val;
 *     TreeNode *left;
 *     TreeNode *right;
 *     TreeNode() : val(0), left(nullptr), right(nullptr) {}
 *     TreeNode(int x) : val(x), left(nullptr), right(nullptr) {}
 *     TreeNode(int x, TreeNode *left, TreeNode *right) : val(x), left(left), right(right) {}
 * };
 */
class Solution {
public:
    vector<double> averageOfLevels(TreeNode* root) {
        if (root == nullptr) {
            return {};
        }
        vector<double> result;
        queue<TreeNode*> q;
        q.push(root);
        while (!q.empty()) {
            double tempSum = 0.0;
            int tempCount = 0;
            int size = q.size();
            for (int i = 0; i < size; ++i) {
                auto node = q.front(); q.pop();
                tempSum += node->val;
                tempCount += 1;
                if (node->left) q.push(node->left);
                if (node->right) q.push(node->right);
            }
            result.push_back(tempSum / tempCount);
        }
        return result;
        
    }
};
```

### 958. Check Completeness of a Binary Tree
Given a binary tree, determine if it is a complete binary tree.

Definition of a complete binary tree from Wikipedia:
In a complete binary tree every level, except possibly the last, is completely filled, and all nodes in the last level are as far left as possible. It can have between 1 and 2h nodes inclusive at the last level h.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isCompleteTree(self, root: TreeNode) -> bool:
        
        if not root:
            return None
        queue = collections.deque([root])

        while queue:
            node = queue.popleft()
            if not node:
                continue
            for child in node.left, node.right:
                if child and queue and queue[-1] is None:
                    return False
                queue.append(child)
        return True
```

### 545. Boundary of Binary Tree

Given a binary tree, return the values of its boundary in anti-clockwise direction starting from root. Boundary includes left boundary, leaves, and right boundary in order without duplicate nodes.  (The values of the nodes may still be duplicates.)

Left boundary is defined as the path from root to the left-most node. Right boundary is defined as the path from root to the right-most node. If the root doesn't have left subtree or right subtree, then the root itself is left boundary or right boundary. Note this definition only applies to the input binary tree, and not applies to any subtrees.

The left-most node is defined as a leaf node you could reach when you always firstly travel to the left subtree if exists. If not, travel to the right subtree. Repeat until you reach a leaf node.

The right-most node is also defined by the same way with left and right exchanged.

```python
class Solution {
public:
    vector<int> boundaryOfBinaryTree(TreeNode* root) {
        if (!root)
            return {};
        
        vector<TreeNode*> leftBD{root};
        TreeNode* curr = root->left;
        while (curr) {
            leftBD.push_back(curr);
            curr = curr->left? curr->left:curr->right;
        }
        vector<TreeNode*> rightBD{root};
        curr = root->right;
        while (curr) {
            rightBD.push_back(curr);
            curr = curr->right? curr->right:curr->left;
        }
        vector<TreeNode*> leaves;
        curr = root;
        stack<TreeNode*> s;
        s.push(root);
        while (!s.empty()) {
            auto node = s.top(); s.pop();
            if (!node->left && !node->right)
                leaves.push_back(node);
            if (node->right)
                s.push(node->right);
            if (node->left)
                s.push(node->left);
        }
        vector<int> res;
        unordered_set<TreeNode*> seen;
        for (auto &e: leftBD)
            visit(res, seen, e);
        for (auto &e: leaves)
            visit(res, seen, e);
        for (auto e = rightBD.rbegin(); e != rightBD.rend(); ++e)
            visit(res, seen, *e);
        
        
        return res;
    }
    
    void visit(vector<int>& res, unordered_set<TreeNode*> &seen, TreeNode* node) {
        if (seen.find(node) == seen.end()) {
            res.push_back(node->val);
            seen.insert(node);
        }
    }
};
```

### 104. Maximum Depth of Binary Tree

```python
104. Maximum Depth of Binary Tree
Given a binary tree, find its maximum depth.

The maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.

Note: A leaf is a node with no children.

Example:

Given binary tree [3,9,20,null,null,15,7],

    3
   / \
  9  20
    /  \
   15   7
return its depth = 3.

# maximum depth of a binary tree
def maxDepth(self, root):
    # write your code here
    if root is None:
        return 0

    return max(self.maxDepth(root.left), self.maxDepth(root.right)) + 1


# DFS solution
class Solution:
    def maxDepth(self, root: TreeNode) -> int:

        if not root:
            return 0
        
        stack = [(root, 1)]
        maxDepth = 0
        while stack:
            
            node, d = stack.pop()
            maxDepth = max(maxDepth, d)
            
            if node.left:
                stack.append((node.left, d + 1))
            if node.right:
                stack.append((node.right, d + 1))
        return maxDepth
    
# extension to N-ary tree
class Solution:
    def maxDepth(self, root: 'Node') -> int:

        if not root:
            return 0
        
        stack = [(root, 1)]
        maxDepth = 0
        while stack:
            
            node, d = stack.pop()
            if node:
                maxDepth = max(maxDepth, d)
                for c in node.children:
                    stack.append((c, d + 1))
        return maxDepth 
    

# alternative BFS method
class Solution:
    def maxDepth(self, root: TreeNode) -> int:
        
        if not root:
            return 0
        
        queue = [root]
        depth = 0
        while queue:    
            newQ = []
            for i in range(len(queue)):
                node = queue.pop()
                if node.left:
                    newQ.append(node.left)
                if node.right:
                    newQ.append(node.right)
            depth += 1        
            queue = newQ
        return depth
```

### 951. Flip Equivalent Binary Trees

```python
For a binary tree T, we can define a flip operation as follows: choose any node, and swap the left and right child subtrees.

A binary tree X is flip equivalent to a binary tree Y if and only if we can make X equal to Y after some number of flip operations.

Write a function that determines whether two binary trees are flip equivalent.  The trees are given by root nodes root1 and root2.

class Solution:
    def flipEquiv(self, root1: TreeNode, root2: TreeNode) -> bool:
        
        if not root1 and not root2:
            return True
        elif not root1 and root2:
            return False
        elif not root2 and root1:
            return False
        else:
            if root1.val != root2.val:
                return False
            
            if self.flipEquiv(root1.left, root2.left) and self.flipEquiv(root1.right, root2.right):
                return True
            if self.flipEquiv(root1.left, root2.right) and self.flipEquiv(root1.right, root2.left):
                return True
            
            return False 
        
# time complexity analysis
In worst case, the recursion corresponds to a perfect quaternary (means 4-nary) tree, which has 4 ^ d = N ^ 2 nodes, 
and we have to traverse all nodes. d = logN is the depth of the binary tree.

One worst case for input:
two perfect binary trees: root1 & root2.

Root1's nodes are all 0s;
Root2's nodes are all 0s, with the exception that left and right bottoms are both 1s.

However, when all node values are unique, the time complexity is O(N)
at least 1 out of flipEquiv(r1.left, r2.left) and flipEquiv(r1.left, r2.right) will terminate; Similiarly, at least 1 out of flipEquiv(r1.right, r2.right) andflipEquiv(r1.right, r2.leftt) will terminate.

Obviously at most 2 out of the 4 recursive calls could go all the way down to the bottom.

# stack solution, assuming values are unique
class Solution:
    def flipEquiv(self, root1: TreeNode, root2: TreeNode) -> bool:
        stack1 = [root1]
        stack2 = [root2]
        
        while stack1 and stack2:
            
            node1, node2 = stack1.pop(), stack2.pop()
            if not node1 and not node2:
                continue
            
            if not node1 or not node2 or node1.val != node2.val:
                return False
            L = node1.left.val if node1.left else -1
            R = node1.right.val if node1.right else -1
            if L < R:
                node1.left, node1.right = node1.right, node1.left
            L = node2.left.val if node2.left else -1
            R = node2.right.val if node2.right else -1
            if L < R:
                node2.left, node2.right = node2.right, node2.left
            
            stack1.extend([node1.left, node1.right])
            stack2.extend([node2.left, node2.right])
        return True
```

Example:

Input: [1,2,3,4,5]

```python

# dfs group elements by heights
class Solution:
    def findLeaves(self, root: TreeNode) -> List[List[int]]:
        
        results = []
        self.dfs(results, root)
        return results
    
    
    def dfs(self, results, node):
        
        if not node:
            return 0
        height = max(self.dfs(results, node.left), self.dfs(results, node.right)) + 1
        if height > len(results):
            results.append([])
        results[height - 1].append(node.val)
        return height
    
## dfs iterative post order
class Solution:
    def findLeaves(self, root: TreeNode) -> List[List[int]]:
        
        if not root:
            return []
        
        results = []
        heights = {}
        stack = [root]
        while stack:
            node = stack[-1]
            if node.right and node.right not in heights:
                stack.append(node.right)
            elif node.left and node.left not in heights:
                stack.append(node.left)
            else:
                stack.pop()
                leftHeight = heights.get(node.left, 0)
                rightHeight = heights.get(node.right, 0)
                height = max(leftHeight, rightHeight) + 1
                heights[node] = height
                if height > len(results):
                    results.append([])
                results[height - 1].append(node.val)
        return results

## Topological sort method, over skill
class Solution:
    def findLeaves(self, root: TreeNode) -> List[List[int]]:
        
        if not root:
            return []
        
        G = collections.defaultdict(list)
        indegree = {}
        stack = [root]
        while stack:
            node = stack.pop()
            indegree[node] = 0
            if node.left:
                G[node.left].append(node)
                stack.append(node.left)
                indegree[node] += 1
            if node.right:
                G[node.right].append(node)
                stack.append(node.right)
                indegree[node] += 1
        
        queue = [node for node, val in indegree.items() if val == 0]
        results = []
        while queue:
            
            tempResults = []
            newQ = []
            for node in queue:
                tempResults.append(node.val)
                for nb in G[node]:
                    indegree[nb] -= 1
                    if indegree[nb] == 0:
                        newQ.append(nb)
            results.append(tempResults)
            queue = newQ
            
        return results
```

### 1339. Maximum Product of Splitted Binary Tree

Given a binary tree root. Split the binary tree into two subtrees by removing 1 edge such that the product of the sums of the subtrees are maximized.

Since the answer may be too large, return it modulo 10^9 + 7.

```python
class Solution:
    def maxProduct(self, root: TreeNode) -> int:
        
        sumDict = {None: 0}
        stack = [root]
        while stack:
            node = stack[-1]
            if node.left and node.left not in sumDict:
                stack.append(node.left)
            elif node.right and node.right not in sumDict:
                stack.append(node.right)
            else:
                stack.pop()
                sumDict[node] = sumDict[node.left] + sumDict[node.right] + node.val

        maxAns = 0
        totalSum = sumDict[root]
        for node in sumDict:
            maxAns = max(maxAns, sumDict[node] * (totalSum - sumDict[node]))
        return maxAns % int(1e9 + 7)
                
            
```

### 437. Path Sum III (count down path with target sum)

```python
You are given a binary tree in which each node contains an integer value.

Find the number of paths that sum to a given value.

The path does not need to start or end at the root or a leaf, but it must go downwards (traveling only from parent nodes to child nodes).

The tree has no more than 1,000 nodes and the values are in the range -1,000,000 to 1,000,000.

# double dfs O(N^2) worst case, best case O(nlog n)
class Solution:
    def pathSum(self, root: TreeNode, sum: int) -> int:
        
        self.count = 0
        stack = [root]
        while stack:
            node = stack.pop()
            if not node:
                continue
            self.dfs(node, sum)
            stack.append(node.left)
            stack.append(node.right)
        return self.count
        
    def dfs(self, node, target):
        
        if not node:
            return
        
        # find one path
        if node.val == target:
            self.count += 1
        target -= node.val
        # continue the search
        
        self.dfs(node.left, target)
        self.dfs(node.right, target)
        

O(n) solution using hash map
class Solution:
    def pathSum(self, root: TreeNode, sum: int) -> int:
        
        self.count = 0
        cache = {0: 1}
        self.dfs(root, sum, 0, cache)
        return self.count
        
    def dfs(self, node, target, currPathSum, cache):
        
        if not node:
            return
        
        currPathSum += node.val
        complement = currPathSum - target 
        #cache which saves all the path sum (from root to current node) and their frequency.
        self.count += cache.get(complement, 0)
        
        cache[currPathSum] = cache.get(currPathSum, 0) + 1
        self.dfs(node.left, target, currPathSum, cache)
        self.dfs(node.right, target, currPathSum, cache)
        # this path is not available anymore
        cache[currPathSum] -= 1
```

### 1372. Longest ZigZag Path in a Binary Tree
Given a binary tree root, a ZigZag path for a binary tree is defined as follow:

Choose any node in the binary tree and a direction (right or left).
If the current direction is right then move to the right child of the current node otherwise move to the left child.
Change the direction from right to left or right to left.
Repeat the second and third step until you can't move in the tree.
Zigzag length is defined as the number of nodes visited - 1. (A single node has a length of 0).

Return the longest ZigZag path contained in that tree.


```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

# recursive bottom up
class Solution:
    def longestZigZag(self, root: TreeNode) -> int:
        
        self.ans = 0
        
        def help(node):
            
            if not node:
                return 0, 0
            
            if node.left:
                leftLeft, leftRight = help(node.left)
                left = 1 + leftRight
            else:
                left = 1
            
            if node.right:
                rightLeft, rightRight = help(node.right)
                right = 1 + rightLeft
            else:
                right = 1
                
            self.ans = max(self.ans, right, left)
            
            return left, right
        
        help(root)
        
        return self.ans - 1 if self.ans != 0 else 0
    
# iterative top down
class Solution:
    def longestZigZag(self, root: TreeNode) -> int:
        
        if not root:
            return 0
        
        stack = [(root, 1, True), (root, 1, False)]
        maxVal = 1
        while stack:
                        # right Flag indicates we are searching for the right children
            node, count, rightFlag = stack.pop()
            
            maxVal = max(maxVal, count)
            if node.left:
                if rightFlag:
                    # if we have left child but we are searching for the right, we restart and search the left
                    # there is no harm to append another stack.append((node.left, 1, True)), but it is duplicate
                    # since stack = [(root, 1, True), (root, 1, False)]
                    stack.append((node.left, 1, False))
                else:
                    stack.append((node.left, count + 1, True))
                    
            if node.right:
                if rightFlag:
                    stack.append((node.right, count + 1, False))
                else:
                    stack.append((node.right, 1, True))
                    
            
        return maxVal - 1
```

```python
Consider all the leaves of a binary tree.  From left to right order, the values of those leaves form a leaf value sequence.
Two binary trees are considered leaf-similar if their leaf value sequence is the same.

Return true if and only if the two given trees with head nodes root1 and root2 are leaf-similar.

class Solution:
    def leafSimilar(self, root1: TreeNode, root2: TreeNode) -> bool:
        
        leafSeq1 = self.getLeafSeq(root1)
        leafSeq2 = self.getLeafSeq(root2)
        
        return leafSeq1 == leafSeq2
    
    # preOrder to get leaf sequence from left to right
    def getLeafSeq(self, root):
        if not root:
            return []
        results = []
        stack = [root]
        while stack:
            node = stack.pop()
            if not (node.left or node.right):
                results.append(node.val)
            if node.right:
                stack.append(node.right)            
            if node.left:
                stack.append(node.left)

        return results
```

```python
404. Sum of Left Leaves

Find the sum of all left leaves in a given binary tree.

Example:

    3
   / \
  9  20
    /  \
   15   7

There are two left leaves in the binary tree, with values 9 and 15 respectively. Return 24.

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def sumOfLeftLeaves(self, root: TreeNode) -> int:
        
        if not root:
            return 0
        # assuming root is not a left leaf in any case
        stack = [(root, root)]
        res = 0
        while stack:
            node, parent = stack.pop()
            if parent.left == node and not (node.left or node.right):
                res += node.val
            
            if node.left:
                stack.append((node.left, node))
            if node.right:
                stack.append((node.right, node))
                
                
        return res
```

### 1315. Sum of Nodes with Even-Valued Grandparent

```python
Given a binary tree, return the sum of values of nodes with even-valued grandparent.  (A grandparent of a node is the parent of its parent, if it exists.)

If there are no nodes with an even-valued grandparent, return 0.

class Solution:
    def sumEvenGrandparent(self, root: TreeNode) -> int:
        
        if not root:
            return 0
        
        stack = [(root, None, None)]
        result = 0
        while stack:
            node, parent, grandParent = stack.pop()
            if grandParent and grandParent.val & 1 == 0:
                result += node.val
                
            if node.left:
                stack.append((node.left, node, parent))
            if node.right:
                stack.append((node.right, node, parent))
                
        return result
```

```python
#
# Assuming two nodes exist in the same subtree, traversing from the top, if the root matches the one of node, if the root is the LCA
# if not, we search left subtree and right subtree. 
# note that the function return once find at least one match

# therefore, we find two notes in the left and right respectively, then current node will be the LCA
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        if not root:
            return None
        
        if root == p or root == q:
            return root
        
        leftLCA = self.lowestCommonAncestor(root.left, p, q)
        rightLCA = self.lowestCommonAncestor(root.right, p, q)
        
        if leftLCA and rightLCA:
            return root
        if leftLCA:
            return leftLCA
        elif rightLCA:
            return rightLCA
   
```

### 426. Convert Binary Search Tree to Sorted Doubly Linked List

```python
426. Convert Binary Search Tree to Sorted Doubly Linked List

Convert a Binary Search Tree to a sorted Circular Doubly-Linked List in place.

You can think of the left and right pointers as synonymous to the predecessor and successor 
pointers in a doubly-linked list. For a circular doubly linked list, the predecessor of the first
element is the last element, and the successor of the last element is the first element.

We want to do the transformation in place. After the transformation, the left pointer of the tree 
node should point to its predecessor, and the right pointer should point to its successor. You should 
return the pointer to the smallest element of the linked list.

"""
# Definition for a Node.
class Node:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
"""

class Solution:
    def treeToDoublyList(self, root: 'Node') -> 'Node':
        
        if not root:
            return None
        
        stack = []
        curr = root
        
        prev = None
        head = None
        while curr or stack:
            
            if curr:
                stack.append(curr)
                curr = curr.left
            else:
                
                node = stack.pop()
                if prev:
                    prev.right = node
                    node.left = prev
                else:
                    head = node
                
                prev = node
                curr = node.right
                
        prev.right = head
        head.left = prev
        
        return head
        
```

You always start to construct the left child node of the parent first if it exists.


```python
class Solution:
    def str2tree(self, s: str) -> TreeNode:
        
        if not s:
            return None
        
        def help(startIdx):        
            endIdx = startIdx
            while endIdx < len(s) and (s[endIdx] == '-' or s[endIdx].isdigit()):
                endIdx += 1
            root = TreeNode(int(s[startIdx:endIdx]))
            
            if endIdx < len(s) and s[endIdx] == '(':
                root.left, endIdx = help(endIdx + 1)
            # for the same parent node, if enconter ( again, create the right node
            
            if endIdx < len(s) and s[endIdx] == '(':
                root.right, endIdx = help(endIdx + 1)
                
            return root, endIdx + 1 
        
        root, _ = help(0)
        return root
    
# a slow but clean solution
class Solution:
    def str2tree(self, s: str) -> TreeNode:
        idx = s.find('(')
        if idx < 0:
            return TreeNode(int(s)) if s else None
        
        bal = 0
        for jx in range(idx, len(s)):
            if s[jx] == '(': bal += 1
            if s[jx] == ')': bal -= 1
            if bal == 0: break
        
        root = TreeNode(int(s[:idx]))
        root.left = self.str2tree(s[idx + 1:jx])
        root.right = self.str2tree(s[jx + 2:-1])
        return root
```

### 971. Flip Binary Tree To Match Preorder Traversal

Given a binary tree with N nodes, each node has a different value from {1, ..., N}.

A node in this binary tree can be flipped by swapping the left child and the right child of that node.

Consider the sequence of N values reported by a preorder traversal starting from the root.  Call such a sequence of N values the voyage of the tree.

(Recall that a preorder traversal of a node means we report the current node's value, then preorder-traverse the left child, then preorder-traverse the right child.)

Our goal is to flip the least number of nodes in the tree so that the voyage of the tree matches the voyage we are given.

If we can do so, then return a list of the values of all nodes flipped.  You may return the answer in any order.

If we cannot do so, then return the list [-1].

```python
class Solution:
    def flipMatchVoyage(self, root: TreeNode, voyage: List[int]) -> List[int]:
        
        voyage.reverse()
        flip = []
        def help(root, voyage):
            if not root:
                return True
            if root.val != voyage[-1]:
                return False
            voyage.pop()
            if root.left and voyage[-1] != root.left.val:
                flip.append(root.val)
                return help(root.right, voyage) and help(root.left, voyage)  
            else:
                return help(root.left, voyage) and help(root.right, voyage)
        if not help(root, voyage):
            return [-1]
        else:
            return flip
```

### 310. Minimum Height Trees

For an undirected graph with tree characteristics, we can choose any node as the root. The result graph is then a rooted tree. Among all possible rooted trees, those with minimum height are called minimum height trees (MHTs). Given such a graph, write a function to find all the MHTs and return a list of their root labels.

Format
The graph contains n nodes which are labeled from 0 to n - 1. You will be given the number n and a list of undirected edges (each edge is a pair of labels).

You can assume that no duplicate edges will appear in edges. Since all edges are undirected, [0, 1] is the same as [1, 0] and thus will not appear together in edges.

```python
class Solution:
    def findMinHeightTrees(self, n: int, edges: List[List[int]]) -> List[int]:
        
        if n == 0:
            return []
        if n == 1:
            return [0]
        
        G = collections.defaultdict(list)
        
        for e in edges:
            G[e[0]].append(e[1])
            G[e[1]].append(e[0])
            
        # find the leaves
        layer = [i for i in G if len(G[i]) == 1]
        while layer:
            
            nextLayer = []
            for node in layer:
                for nb in G[node]:
                    G[nb].remove(node) # remove its children such that it can becomes leaves again
                    if len(G[nb]) == 1: # become leaves
                        nextLayer.append(nb)
                                    
            if not nextLayer:
                return layer
            else:
                layer = nextLayer
```

The width of one level is defined as the length between the end-nodes (the leftmost and right most non-null nodes in the level, where the null nodes between the end-nodes are also counted into the length calculation.

```python
# level traversal. Every node is ia binary number.
class Solution:
    def widthOfBinaryTree(self, root: TreeNode) -> int:
        
        if not root:
            return 0
        queue = collections.deque([(root, 0)])
        minVal = float('inf')
        answer = 0
        while queue:
            for _ in range(len(queue)):
                node, num = queue.popleft()
                minVal = min(minVal, num)
                answer = max(answer, num - minVal + 1)
                if node.left:
                    queue.append((node.left, num << 1))
                if node.right:
                    queue.append((node.right, (num << 1) + 1))        
            minVal = float('inf')
        return answer

class Solution:
    def widthOfBinaryTree(self, root: TreeNode) -> int:
        
        if not root:
            return 0
        levelMin = [0]
        stack = [(root, 0, 0)]
        answer = 0
        while stack:
            node, level, num = stack.pop()
            
            if level >= len(levelMin):
                levelMin.append(num)
            else:
                levelMin[level] = min(levelMin[level], num)
                answer = max(answer, num - levelMin[level] + 1)
            if node.right:# must right first to update levelMin
                stack.append((node.right, level + 1, (num << 1) + 1))    
            if node.left:
                stack.append((node.left, level + 1, num << 1))
            
        return answer
```
