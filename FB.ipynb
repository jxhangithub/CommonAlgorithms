426. Convert Binary Search Tree to Sorted Doubly Linked List

Convert a Binary Search Tree to a sorted Circular Doubly-Linked List in place.

You can think of the left and right pointers as synonymous to the predecessor and successor pointers in a doubly-linked list. For a circular doubly linked list, the predecessor of the first element is the last element, and the successor of the last element is the first element.

We want to do the transformation in place. After the transformation, the left pointer of the tree node should point to its predecessor, and the right pointer should point to its successor. You should return the pointer to the smallest element of the linked list.

```python
"""
# Definition for a Node.
class Node:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
"""

class Solution:
    def treeToDoublyList(self, root: 'Optional[Node]') -> 'Optional[Node]':

        def helper(root):
            if not root:
                return None, None
            leftleft, leftright = helper(root.left)
            rightleft, rightright = helper(root.right)
            root.left = leftright
            root.right = rightleft
            if leftright:
                leftright.right = root
            if rightleft:
                rightleft.left = root
            left = leftleft if leftleft else root
            right = rightright if rightright else root
            return left, right
        if not root:
            return root
        head, tail = helper(root)
        head.left = tail
        tail.right = head
        return head



        
```

15. 3Sum

Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

Notice that the solution set must not contain duplicate triplets.

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()
        L = len(nums)
        res = []
        for i in range(L-2):
            if i>0 and nums[i]==nums[i-1]:
                continue
            target = -nums[i]
            s, e = i+1, L-1
            while s < e:
                v1 = nums[s]
                v2 = nums[e]
                v = v1+v2
                if v == target:
                    res.append([-target,v1,v2])
                    while nums[s] == v1 and s<e:
                        s+=1
                elif v < target:
                    s+=1
                else:
                    e-=1
        return res
                
```

In Unix-style file system context, a single period '.' signifies the current directory, a double period ".." denotes moving up one directory level, and multiple slashes such as "//" are interpreted as a single slash. The simplified canonical path should adhere to the following rules:

It must start with a single slash '/'.
Directories within the path should be separated by only one slash '/'.

```python
class Node:
    def __init__(self, key=None, value=None) -> None:
        self.key = key
        self.value = value
        self.prev = None
        self.next = None

class DLList:
    def __init__(self):
        self.head = Node()
        self.tail = Node()
        self.head.next = self.tail
        self.tail.prev = self.head

    def add2head(self,node):
        node.next = self.head.next
        node.prev = self.head
        self.head.next = node
        node.next.prev = node


    def remove(self,node):
        node.prev.next = node.next
        node.next.prev = node.prev
        node.prev = None
        node.next = None

    def move2head(self,node):
        self.remove(node)
        self.add2head(node)

    def removeTail(self):
        node = self.tail.prev
        self.remove(node)
        return node

    

class LRUCache:

    def __init__(self, capacity: int):
        self.capacity = capacity
        self.list = DLList()
        self.key2node = dict()

    def get(self, key: int) -> int:
        if key in self.key2node:
            node = self.key2node[key]
            self.list.move2head(node)
            return node.value
        else:
            return -1
        


    def put(self, key: int, value: int) -> None:
        if key in self.key2node:  
            # if override node directly, we will keep the old node in list
            # list size is bigger than dict size
            # when remove this old node, it will remove the wrong key value for the new node in dict, 
            # and may kepp or remove new node as well
            node = self.key2node[key]
            node.value = value
            self.list.move2head(node)
        else:
            if len(self.key2node) == self.capacity:
                node = self.list.removeTail()
                self.key2node.pop(node.key)
            node = Node(key, value)
            self.key2node[key] = node
            self.list.add2head(node)
        
```

```python
class Solution:
    def numIslands(self, grid: List[List[int]]) -> int:
        R = len(grid)
        C = len(grid[0])
        parent = [ r*C+c for r in range(R) for c in range(C)]

        def find(x): # find root
            if x != parent[x]: # not root
                parent[x] = find(parent[x]) # until find root
            return parent[x]  # return root

        def union(x,y):
            parent[find(x)] = find(y) # parent of one root is set as another root

        res = 0
        for r in range(R):
            for c in range(C):
                if grid[r][c] == '1':
                    res += 1  # one additional island
                    for dr, dc in (-1,0), (1,0), (0,-1), (0,1):
                        nr = r+dr
                        nc = c+dc
                        if nr<0 or nr>R-1 or nc<0 or nc>C-1 or grid[nr][nc] == '0':
                            continue
                        if find(r*C+c) != find(nr*C+nc): # this connected two has not union yet
                            union(r*C+c, nr*C+nc)
                            res -= 1 # 2 island become 1
        return res
```

If two nodes are in the same row and column, the order should be from left to right.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

class Solution:
    def verticalOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        tmp = collections.defaultdict(list)
        q = collections.deque([(root, 0)])
        while q:
            node, col = q.popleft()
            if not node:
                continue
            tmp[col].append(node.val)
            q.append((node.left, col-1))
            q.append((node.right, col+1))
        res = []
        for col in sorted(tmp.keys()):
            res.append(tmp[col])
        return res
```

```python
class Solution:
    def validWordAbbreviation(self, word: str, abbr: str) -> bool:
        Lw = len(word)
        Lb = len(abbr)
        iw = ib = 0
        while iw < Lw and ib < Lb:
            if abbr[ib].isdigit():
                num = int(abbr[ib])
                if num == 0: # start with 0 is not allowed
                    return False
                ib += 1
                while ib < Lb and abbr[ib].isdigit():
                    num = num*10 + int(abbr[ib])
                    ib += 1
                iw += num
            elif abbr[ib].isalpha():
                if abbr[ib] != word[iw]:
                    return False
                ib += 1
                iw += 1
        return ib == Lb and iw == Lw
```

10. Regular Expression Matching

Given an input string s and a pattern p, implement regular expression matching with support for '.' and '*' where:

'.' Matches any single character.​​​​
'*' Matches zero or more of the preceding element.
The matching should cover the entire input string (not partial).

```python
class Solution:
    def isMatch(self, s: str, p: str) -> bool:
        cache = {}
        # from current i, j match to end of string

        def dp(si, pi) -> bool:
            if (si, pi) in cache:
                return cache[(si, pi)]
            if pi == len(p):
                return si == len(s)
            curMatch = si < len(s) and pi < len(p) and p[pi] in (s[si], '.')
            fullMatch = curMatch and dp(si+1, pi+1)
            if pi+1 < len(p) and p[pi+1] == '*':
                zeroMatch = dp(si, pi+2)
                multiMatch = curMatch and dp(si+1, pi)
                fullMatch = fullMatch or zeroMatch or multiMatch
                #           match one,   match 0,     match multiple
            cache[(si, pi)] = fullMatch
            return fullMatch
        return dp(0,0)
```

```python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        L = len(nums)
        s,e = 0, L-1
        while s+1 < e:
            m = s+(e-s)//2
            v = nums[m]
            if v==target:
                return m
            if v > target:
                if v <= nums[e] or target >=nums[s]: # on one branch
                    e = m-1
                else: # two branch
                    s = m+1
            else:
                if target <= nums[e] or v >= nums[s]: # on one branch
                    s = m+1
                else: # two branch
                    e = m-1
        if nums[s] == target:
            return s
        if nums[e] == target:
            return e
        return -1
```

You must write an algorithm that runs in O(log n) time.

```python
class Solution:
    def findPeakElement(self, nums: List[int]) -> int:
        L = len(nums)
        start, end = 0, L-1
        while start + 1 < end:
            mid = start + (end-start) // 2
            prev = nums[mid-1]
            cur = nums[mid]
            next = nums[mid+1]
            if cur > prev and cur > next:
                return mid
            elif cur < prev:
                end = mid -1
            elif cur < next:
                start = mid+1
        if nums[start] >= nums[end]:
            return start
        if nums[start] <= nums[end]:
            return end
        
```

A trie (pronounced as "try") or prefix tree is a tree data structure used to efficiently store and retrieve keys in a dataset of strings. There are various applications of this data structure, such as autocomplete and spellchecker.

Implement the Trie class:

Trie() Initializes the trie object.
void insert(String word) Inserts the string word into the trie.
boolean search(String word) Returns true if the string word is in the trie (i.e., was inserted before), and false otherwise.
boolean startsWith(String prefix) Returns true if there is a previously inserted string word that has the prefix prefix, and false otherwise.

```python
class Node:
    def __init__(self) -> None:
        self.children = dict()
        self.isWord = False


class Trie:

    def __init__(self):
        self.root = Node()
        

    def insert(self, word: str) -> None:
        cur = self.root
        for ch in word:
            if ch not in cur.children:
                cur.children[ch] = Node()
            cur = cur.children[ch]
        cur.isWord = True
        
    def _searchPrefixNode(self, prefix: str) -> Node:
        cur = self.root
        for ch in prefix:
            if ch not in cur.children:
                return None
            cur = cur.children[ch]
        return cur

        
    def search(self, word: str) -> bool:
        cur = self._searchPrefixNode(word)
        return cur and cur.isWord
        
        
    def startsWith(self, prefix: str) -> bool:
        cur = self._searchPrefixNode(prefix)
        return cur != None
        
```

235. Lowest Common Ancestor of a Binary Search Tree

Given a binary search tree (BST), find the lowest common ancestor (LCA) node of two given nodes in the BST.

According to the definition of LCA on Wikipedia: “The lowest common ancestor is defined between two nodes p and q as the lowest node in T that has both p and q as descendants (where we allow a node to be a descendant of itself).”

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        if not root:
            return root
        pv = p.val
        qv = q.val
        start = min(pv, qv)
        end = max(pv, qv)
        while root:
            if root.val > end:
                root = root.left
            elif root.val < start:
                root = root.right
            elif start <= root.val <= end:
                return root
        return None
```

```python
class Solution:
    def maximumSwap(self, num: int) -> int:
        digits = []
        tmp = num
        while tmp > 0:
            digits.append(tmp %10)
            tmp //= 10
        digits = list(reversed(digits))
        L = len(digits)
        idx = None
        for i in range(L-1):
            if digits[i] < digits[i+1]:
                idx = i
                break
        if idx == None: # if decreasing all the way, already biggest
            return num
            
        

        idy = None
        for i in range(idx+1, L): # find the right most, largest digit
            if idy == None or digits[i] >= digits[idy]:
                idy = i


        for i in range(idx+1): # find the left most, samller than largest digit
            if digits[i] < digits[idy]:
                idx = i
                break

        digits[idx], digits[idy] = digits[idy], digits[idx]
        cur = 0
        for d in digits:
            cur *= 10
            cur += d
        return cur


```

865. Smallest Subtree with all the Deepest Nodes

Given the root of a binary tree, the depth of each node is the shortest distance to the root.

Return the smallest subtree such that it contains all the deepest nodes in the original tree.

A node is called the deepest if it has the largest depth possible among any node in the entire tree.

The subtree of a node is a tree consisting of that node, plus the set of all descendants of that node.

```python
class Solution(object):
    def subtreeWithAllDeepest(self, root):
        n2d = dict()
        def markDepth(node, depth):
            if not node:
                return
            n2d[node] = depth
            markDepth(node.left, depth+1)
            markDepth(node.right, depth+1)

        markDepth(root, 0)
        maxDepth = max(n2d.values())

        def deepest(node):
            if not node or n2d[node] == maxDepth:
                return node
            L = deepest(node.left)
            R = deepest(node.right)
            return node if L and R else L if L else R

        return deepest(root)
```

977. Squares of a Sorted Array

Given an integer array nums sorted in non-decreasing order, return an array of the squares of each number sorted in non-decreasing order.

```python
class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        L = len(nums)
        res = [0]*L
        left = 0
        right = L-1
        idx = L-1
        while left <= right:
            vl = nums[left]
            vr = nums[right]
            if abs(vl) <= abs(vr):
                res[idx] = vr * vr
                idx -= 1
                right -= 1
            else:
                res[idx] = vl * vl
                idx -= 1
                left += 1
        return res
        
```

```python
class Solution:
    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        idx = m+n-1
        i1 = m-1
        i2 = n-1
        while i1 > -1 and i2 > -1:
            if nums1[i1] >= nums2[i2]:
                nums1[idx] = nums1[i1]
                idx -= 1
                i1 -= 1
            else:
                nums1[idx] = nums2[i2]
                idx -= 1
                i2 -= 1
        while i2 > -1:
            nums1[idx] = nums2[i2]
            idx -= 1
            i2 -= 1
```

236. Lowest Common Ancestor of a Binary Tree

Given a binary tree, find the lowest common ancestor (LCA) of two given nodes in the tree.

According to the definition of LCA on Wikipedia: “The lowest common ancestor is defined between two nodes p and q as the lowest node in T that has both p and q as descendants (where we allow a node to be a descendant of itself).”

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':

        def lca(root):
            if not root:
                return False, False, None
            leftp, leftq, leftlca = lca(root.left)
            rightp, rightq, rightlca = lca(root.right)
            if leftlca:
                return True, True, leftlca
            if rightlca:
                return True, True, rightlca
            rootp = (root.val == p.val)
            rootq = (root.val == q.val)
            ifp = rootp or leftp or rightp
            ifq = rootq or leftq or rightq
            return ifp, ifq, (root if ifp and ifq else None)
        _,_,res = lca(root)
        return res
```

13. Roman to Integer

Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.

Symbol       Value
I             1
V             5
X             10
L             50
C             100
D             500
M             1000
For example, 2 is written as II in Roman numeral, just two ones added together. 12 is written as XII, which is simply X + II. The number 27 is written as XXVII, which is XX + V + II.

Roman numerals are usually written largest to smallest from left to right. However, the numeral for four is not IIII. Instead, the number four is written as IV. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as IX. There are six instances where subtraction is used:

I can be placed before V (5) and X (10) to make 4 and 9. 
X can be placed before L (50) and C (100) to make 40 and 90. 
C can be placed before D (500) and M (1000) to make 400 and 900.
Given a roman numeral, convert it to an integer.

```python
class Solution:
    def romanToInt(self, s: str) -> int:
        romans = 'IVXLCDM'
        numbers = [1,5,10,50,100,500,1000]
        r2n = {}
        for r,n in zip(romans, numbers):
            r2n[r] = n
        tmp = []
        for c in s:
            tmp.append(r2n[c])
        res = tmp[0]
        for i in range(1,len(tmp)):
            if tmp[i] > tmp[i-1]:
                res -= 2*tmp[i-1]
            res += tmp[i]
        return res
```

20. Valid Parentheses

Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.

An input string is valid if:

Open brackets must be closed by the same type of brackets.
Open brackets must be closed in the correct order.
Every close bracket has a corresponding open bracket of the same type.

```python
class Solution:
    def isValid(self, s: str) -> bool:
        lefts = '{[('
        rights = '}])'
        r2l = dict()
        for l,r in zip(lefts, rights):
            r2l[r] = l
        stack = []
        for c in s:
            if c in lefts:
                stack.append(c)
            else:
                if stack and stack[-1] == r2l[c]:
                    stack.pop()
                else:
                    return False
        return not stack
        
```

25. Reverse Nodes in k-Group

Given the head of a linked list, reverse the nodes of the list k at a time, and return the modified list.

k is a positive integer and is less than or equal to the length of the linked list. If the number of nodes is not a multiple of k then left-out nodes, in the end, should remain as it is.

You may not alter the values in the list's nodes, only nodes themselves may be changed.

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        if k==1:
            return head
        dummy = ListNode()
        dummy.next = head
        cur = dummy
        first = head
        while first:
            kth = first
            for i in range(k-1):
                if kth:
                    kth = kth.next
            if not kth:
                break
            start = kth.next
            kth.next = None
            pre = None
            q = first
            while q != None:
                next = q.next
                q.next = pre
                pre = q
                q = next
            cur.next = kth
            cur = first
            cur.next = start
            first = start
        return dummy.next
```

28. Find the Index of the First Occurrence in a String

Given two strings needle and haystack, return the index of the first occurrence of needle in haystack, or -1 if needle is not part of haystack.

```python
class Solution: # KMP algo
    def strStr(self, haystack: str, needle: str) -> int:
        LH = len(haystack)
        LN = len(needle)
        if LN > LH:
            return -1
        failBackTo = [0]*LN  
        # based on needle, find pattern if not match current char, 
        # move back to previous already matched prefix, 
        # usually not back to 0 index
        idxMatch = 0
        for i in range(1, LN):
            while idxMatch > 0 and needle[idxMatch] != needle[i]:
                idxMatch = failBackTo[idxMatch-1]
            if needle[idxMatch] == needle[i]:
                idxMatch += 1
            failBackTo[i] = idxMatch
        # idx: 0 1 2 3 4 5 6 
        # arr: A B C D A B D
        # pre: 0 0 0 0 1 2 0

        idxMatch = 0
        for i in range(LH):
            while idxMatch > 0 and haystack[i] != needle[idxMatch]:
                idxMatch = failBackTo[idxMatch-1]
            if haystack[i] == needle[idxMatch]:
                idxMatch += 1
            if idxMatch == LN:
                return i-LN+1
        return -1
```

43. Multiply Strings

Given two non-negative integers num1 and num2 represented as strings, return the product of num1 and num2, also represented as a string.

Note: You must not use any built-in BigInteger library or convert the inputs to integer directly.

```python
class Solution:
    def multiply(self, num1: str, num2: str) -> str:
        if num1 == '0' or num2 == '0':
            return '0'
        arr1 = list(map(int, reversed(num1)))
        arr2 = list(map(int, reversed(num2)))
        res = []
        for i, vi in enumerate(arr1):
            carry = 0
            for j, vj in enumerate(arr2):
                carry += vi * vj
                if i+j == len(res):
                    res.append(carry %10)
                else:
                    carry += res[i+j] 
                    res[i+j] = carry %10
                carry //= 10
            k = i+len(arr2)-1
            while carry:
                k += 1
                if k == len(res):
                    res.append(carry %10)
                else:
                    carry += res[k]
                    res[k] = carry %10
                carry //=10 
        return ''.join(map(str,reversed(res)))


        
```

44. Wildcard Matching

Given an input string (s) and a pattern (p), implement wildcard pattern matching with support for '?' and '*' where:

'?' Matches any single character.
'*' Matches any sequence of characters (including the empty sequence).
The matching should cover the entire input string (not partial).

```python
class Solution:
    def isMatch(self, s: str, p: str) -> bool:
        Ls = len(s)
        Lp = len(p)
        dp = [[False] * (Ls+1) for _ in range(Lp+1)] # index in dp is length from 0 index for each string
        # dp[0][0] = True # zero length match zero length
        # # dp[1][0] = p[0] == '*'
        
        for idp in range(Lp+1):
            for ids in range(Ls+1):
                if idp == 0 and ids == 0:
                    dp[idp][ids] = True
                else:
                    oneMatch = zeroMatch = moreMatch = False
                    if idp>0 and ids>0:
                        oneMatch = p[idp-1] in (s[ids-1], '?', '*') and dp[idp-1][ids-1]
                        moreMatch = p[idp-1] =='*' and dp[idp][ids-1]
                    if idp>0:
                        zeroMatch = p[idp-1] =='*' and dp[idp-1][ids]
                    
                    dp[idp][ids] = oneMatch or zeroMatch or moreMatch
        # print(dp)
        return dp[-1][-1]
```
