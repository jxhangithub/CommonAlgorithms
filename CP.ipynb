The atomic element always starts with an uppercase character, then zero or more lowercase letters, representing the name.

One or more digits representing that element's count may follow if the count is greater than 1. If the count is 1, no digits will follow.

For example, "H2O" and "H2O2" are possible, but "H1O2" is impossible.
Two formulas are concatenated together to produce another formula.

For example, "H2O2He3Mg4" is also a formula.
A formula placed in parentheses, and a count (optionally added) is also a formula.

For example, "(H2O2)" and "(H2O2)3" are formulas.
Return the count of all elements as a string in the following form: the first name (in sorted order), followed by its count (if that count is more than 1), followed by the second name (in sorted order), followed by its count (if that count is more than 1), and so on.

The test cases are generated so that all the values in the output fit in a 32-bit integer.

 

Example 1:

Input: formula = "H2O"
Output: "H2O"
Explanation: The count of elements are {'H': 2, 'O': 1}.
Example 2:

Input: formula = "Mg(OH)2"
Output: "H2MgO2"
Explanation: The count of elements are {'H': 2, 'Mg': 1, 'O': 2}.
Example 3:

Input: formula = "K4(ON(SO3)2)2"
Output: "K4N2O14S4"
Explanation: The count of elements are {'K': 4, 'N': 2, 'O': 14, 'S': 4}.

```python
class Solution:
    def countOfAtoms(self, s: str) -> str:
        
        def parse(idx):
            currCount = collections.Counter()
            while idx < len(s):
                
                if s[idx].isalpha():
                    end = idx + 1
                    while end < len(s) and s[end].islower() and s[end].isalpha():
                        end += 1
                    last = s[idx:end]
                    idx = end
                    count = 1
                    if idx < len(s) and s[idx].isdigit():
                        end = idx + 1
                        while end < len(s) and s[end].isdigit():
                            end += 1
                        count = int(s[idx:end])
                        idx = end
                    currCount[last] += count    
                elif s[idx] == '(':
                    idx, tempCount = parse(idx + 1)
                    for k in tempCount:
                        currCount[k] += tempCount[k]
                    
                elif s[idx] == ')':
                    # parse the digits follow it
                    end = idx + 1
                    if end < len(s) and s[end].isdigit():
                        digitEnd = end + 1
                        while digitEnd < len(s) and s[digitEnd].isdigit():
                            digitEnd += 1
                        count = int(s[end:digitEnd])
                        for k in currCount:
                            currCount[k] *= count 
                        end = digitEnd
                    return end, currCount
            return idx, currCount
                    
                    
        _, counter = parse(0)                    
        sortedKeys = sorted(counter.keys())
        result = []
        for k in sortedKeys:
            result.append(k)
            if counter[k] > 1:
                result.append(str(counter[k]))
        return ''.join(result)
```

### 1254. Number of Closed Islands

Given a 2D grid consists of 0s (land) and 1s (water).  An island is a maximal 4-directionally connected group of 0s and a closed island is an island totally (all left, top, right, bottom) surrounded by 1s.

Return the number of closed islands.

```python
class Solution:
    def closedIsland(self, grid: List[List[int]]) -> int:
        
        count = 0
        
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == 0:
                    count += self.bfs(grid, i, j)
        return count
    
    def bfs(self, grid, i, j):
        
        queue = collections.deque([(i, j)])
        grid[i][j] = 2
        close_flag = 1
        
        while queue:
            x, y = queue.popleft()
            if x == 0 or x == len(grid) - 1 or y == 0 or y == len(grid[0]) - 1:
                close_flag = 0
            
            for dx, dy in (-1, 0), (1, 0), (0, -1), (0, 1):
                new_x = x + dx
                new_y = y + dy
                if 0 <= new_x < len(grid) and 0 <= new_y < len(grid[0]):
                    if grid[new_x][new_y] == 0:
                        grid[new_x][new_y] = 2
                        queue.append((new_x, new_y))
        
        return close_flag
```

270. Closest Binary Search Tree Value

Given the root of a binary search tree and a target value, return the value in the BST that is closest to the target. If there are multiple answers, print the smallest.

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def closestValue(self, root: Optional[TreeNode], target: float) -> int:
        result = float('inf')
        while root:
            if abs(result - target) > abs(root.val -target) or (root.val < result and abs(result - target) == abs(root.val -target)):
                result = root.val
            if target < root.val:
                root = root.left
            elif target == root.val:
                return root.val
            else:
                root = root.right
        return result
```

56. Merge Intervals

Given an array of intervals where intervals[i] = [starti, endi], merge all overlapping intervals, and return an array of the non-overlapping intervals that cover all the intervals in the input.

```python
class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        events = []
        for start, end in intervals:
            events.append((start, -1))
            events.append((end, 1))
        cnt = 0
        start = None
        events.sort()
        res = []
        for e, c in sorted(events):
            if start == None:
                start = e
            cnt += c
            if cnt == 0:
                res.append([start, e])
                start = None
        return res
```

```python
class Solution:
    def uniquePathsWithObstacles(self, grid: List[List[int]]) -> int:
        if not grid or not grid[0]:
            return 0
        R = len(grid)
        C = len(grid[0])
        
        dp = [[0] * (C+1) for _ in range(R+1)]
        dp[1][1] = 0 if grid[0][0] == 1 else 1
        for i in range(1, R+1):
            for j in range(1, C+1):
                dp[i][j] += 0 if grid[i-1][j-1] == 1 else dp[i-1][j] + dp[i][j-1]
        return dp[-1][-1]
```

724. Find Pivot Index

Given an array of integers nums, calculate the pivot index of this array.

The pivot index is the index where the sum of all the numbers strictly to the left of the index is equal to the sum of all the numbers strictly to the index's right.

If the index is on the left edge of the array, then the left sum is 0 because there are no elements to the left. This also applies to the right edge of the array.

Return the leftmost pivot index. If no such index exists, return -1.

```python
class Solution:
    def pivotIndex(self, nums: List[int]) -> int:
        preSum = []
        for i in nums:
            preSum.append(i if not preSum else preSum[-1] + i) 
        for idx in range(len(nums)):
            left = 0 if idx==0 else preSum[idx-1]
            right = 0 if idx == len(nums)-1 else preSum[len(nums)-1] - preSum[idx]
            if left == right:
                return idx
        return -1

```

```python
"""
Intuition
Split a binary search tree (BST) into two trees based on a given target value. 
The main idea is to recursively traverse the BST and at each node, 
decide whether it should belong to the left split 
(containing all values less than or equal to the target) or the right split 
(containing all values greater than the target). This task leverages the properties of a BST,
 where the left child of a node is always less than the node and the right child is always greater.

Approach
The solution involves a recursive function BST that takes a node and the target value as inputs. 
At each step, the function compares the node's value with the target:

If the node's value is greater than the target, the function recursively processes the left child. 
The returned left subtree is part of the left split, 
and the right subtree needs to be reattached to the current node to maintain the BST property for the right split.

If the node's value is less than or equal to the target, the function recursively processes the right child.
 The returned right subtree is part of the right split, 
 and the left subtree needs to be reattached to the current node for the left split.

The recursion stops when it encounters a null node, returning a pair of null nodes in such cases.

Complexity
Time complexity:
The time complexity is O(n). In the worst-case scenario, the tree might be skewed, resembling a linked list. 
In such cases, the function would need to visit each node exactly once, 
leading to a complexity linearly proportional to the number of nodes in the tree.

Space complexity:
The space complexity is O(h), where h is the height of the BST. 
This complexity arises from the recursion stack used during the traversal of the tree. 
In the best case, where the tree is balanced, the height h would logn, 
resulting in a space complexity of O(logn). However, in the worst case, 
particularly when the BST is skewed (resembling a linear chain), the height of the tree becomes n,
 leading to a space complexity of O(n).
"""
# Code
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def splitBST(self, root: Optional[TreeNode], target: int) -> List[Optional[TreeNode]]:
        def BST(root, target):
            # base case
            if not root:
                return None, None

            if root.val > target:
                left, right = BST(root.left, target)
                # 我的理解是，如果root.val > target，就进去root的左边找出比 target 小和大的左右子树 left 和 right
                # 因为如果root.val > target，那么其实自动就分成了 root 和 root.left 两个部分了

                root.left = right
                # 找到的right，即刚才进去root.left之后，在这里面发现的应该属于root这部分的(即: > target部分的树串)
                # 把它接到root.left

                return left, root
                # 返回分好的两颗子树，root.left(<=target and >target)

            else:
                left, right = BST(root.right, target)
                root.right = left
                return root, right
            

        leftTree, rightTree = BST(root, target)
        ans = [leftTree, rightTree]
        return ans
```
