726. Number of Atoms

Given a string formula representing a chemical formula, return the count of each atom.

The atomic element always starts with an uppercase character, then zero or more lowercase letters, representing the name.

One or more digits representing that element's count may follow if the count is greater than 1. If the count is 1, no digits will follow.

For example, "H2O" and "H2O2" are possible, but "H1O2" is impossible.
Two formulas are concatenated together to produce another formula.

For example, "H2O2He3Mg4" is also a formula.
A formula placed in parentheses, and a count (optionally added) is also a formula.

For example, "(H2O2)" and "(H2O2)3" are formulas.
Return the count of all elements as a string in the following form: the first name (in sorted order), followed by its count (if that count is more than 1), followed by the second name (in sorted order), followed by its count (if that count is more than 1), and so on.

The test cases are generated so that all the values in the output fit in a 32-bit integer.

 

Example 1:

Input: formula = "H2O"
Output: "H2O"
Explanation: The count of elements are {'H': 2, 'O': 1}.
Example 2:

Input: formula = "Mg(OH)2"
Output: "H2MgO2"
Explanation: The count of elements are {'H': 2, 'Mg': 1, 'O': 2}.
Example 3:

Input: formula = "K4(ON(SO3)2)2"
Output: "K4N2O14S4"
Explanation: The count of elements are {'K': 4, 'N': 2, 'O': 14, 'S': 4}.

```python
class Solution:
    def countOfAtoms(self, s: str) -> str:
        
        def parse(idx):
            currCount = collections.Counter()
            while idx < len(s):
                
                if s[idx].isalpha():
                    end = idx + 1
                    while end < len(s) and s[end].islower() and s[end].isalpha():
                        end += 1
                    last = s[idx:end]
                    idx = end
                    count = 1
                    if idx < len(s) and s[idx].isdigit():
                        end = idx + 1
                        while end < len(s) and s[end].isdigit():
                            end += 1
                        count = int(s[idx:end])
                        idx = end
                    currCount[last] += count    
                elif s[idx] == '(':
                    idx, tempCount = parse(idx + 1)
                    for k in tempCount:
                        currCount[k] += tempCount[k]
                    
                elif s[idx] == ')':
                    # parse the digits follow it
                    end = idx + 1
                    if end < len(s) and s[end].isdigit():
                        digitEnd = end + 1
                        while digitEnd < len(s) and s[digitEnd].isdigit():
                            digitEnd += 1
                        count = int(s[end:digitEnd])
                        for k in currCount:
                            currCount[k] *= count 
                        end = digitEnd
                    return end, currCount
            return idx, currCount
                    
                    
        _, counter = parse(0)                    
        sortedKeys = sorted(counter.keys())
        result = []
        for k in sortedKeys:
            result.append(k)
            if counter[k] > 1:
                result.append(str(counter[k]))
        return ''.join(result)
```

### 1254. Number of Closed Islands

Given a 2D grid consists of 0s (land) and 1s (water).  An island is a maximal 4-directionally connected group of 0s and a closed island is an island totally (all left, top, right, bottom) surrounded by 1s.

Return the number of closed islands.

```python
class Solution:
    def closedIsland(self, grid: List[List[int]]) -> int:
        
        count = 0
        
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == 0:
                    count += self.bfs(grid, i, j)
        return count
    
    def bfs(self, grid, i, j):
        
        queue = collections.deque([(i, j)])
        grid[i][j] = 2
        close_flag = 1
        
        while queue:
            x, y = queue.popleft()
            if x == 0 or x == len(grid) - 1 or y == 0 or y == len(grid[0]) - 1:
                close_flag = 0
            
            for dx, dy in (-1, 0), (1, 0), (0, -1), (0, 1):
                new_x = x + dx
                new_y = y + dy
                if 0 <= new_x < len(grid) and 0 <= new_y < len(grid[0]):
                    if grid[new_x][new_y] == 0:
                        grid[new_x][new_y] = 2
                        queue.append((new_x, new_y))
        
        return close_flag
```

```python
from typing import List
class Solution:
    def closedIsland(self, grid: List[List[int]]) -> int:
        if len(grid) == 0 or len(grid[0]) == 0:
            return 0
        R = len(grid)
        C = len(grid[0])
        res = 0
        visited = set()

        def bfs(r: int, c: int) -> int:
            if (r, c) in visited or grid[r][c] == 1:
                return 0
            q = collections.deque()
            q.append((r,c))
            curVi = set()
            isQulify = True
            while q:
                x,y = q.popleft()
                curVi.add((x,y))
                for dx, dy in (0,1), (0,-1), (-1,0), (1,0):
                    nx = x+dx
                    ny = y+dy
                    if nx < 0 or nx > R-1 or ny < 0 or ny > C-1:
                        isQulify = False
                    elif (nx, ny) in curVi:
                        continue
                    elif grid[nx][ny] == 0:
                        q.append((nx,ny))
            visited.update(curVi)
            return 1 if isQulify else 0

        for r in range(R):
            for c in range(C):
                res += bfs(r, c)

        return res 
```

```python
from typing import List
class Solution:
    def closedIsland(self, grid: List[List[int]]) -> int:
        if len(grid) == 0 or len(grid[0]) == 0:
            return 0
        R = len(grid)
        C = len(grid[0])
        res = 0
        visited = set()
        stepVi = set()
        isQualify = [True]

        def dfs(r: int, c: int) -> int:
            if (r, c) in visited or grid[r][c] == 1:
                isQualify[0] = False
                return
            stepVi.add((r,c))
            for dx, dy in (0,1), (0,-1), (-1,0), (1,0):
                nx = r+dx
                ny = c+dy
                if nx < 0 or nx > R-1 or ny < 0 or ny > C-1:
                    isQualify[0] = False
                elif (nx, ny) in stepVi:
                    continue
                elif grid[nx][ny] == 0:
                    dfs(nx,ny)

        for r in range(R):
            for c in range(C):
                isQualify[0] = True
                stepVi.clear()
                dfs(r, c)
                visited.update(stepVi)
                res += 1 if isQualify[0] else 0

        return res 
```
