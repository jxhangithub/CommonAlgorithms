```python
def comFeatures(features):
    if not features:
        return [[]]
    result = [[]] 
    for fea in features:   # lenFea
        newresult = []
        for res in result:   # v1 * v2 * .... Vi-1
            for val in fea:    # lenVal
                resNew = res[:]
                resNew.append(val)
                newresult.append(resNew)
        # print(newresult)
        result = newresult

    return ['|'.join(e) for e in result]

print(comFeatures([['food', "sports"],["female","male"]]))

# 

# time: v1 * v2 * ... * vn -> O(lenVal ^ lenFea * lenVal * lenFea) 

# v + v^2 +...+ v^n = (1-v^n) / 1-v -> V^F
    
# space O(lenFea*lenVal), lenFea * lenVal size output 

# print(comFeatures([['food', "sports"]]))
```

```python
# k length subarray sum max 


def maxKsum(array, k):
    if not array: 
        return 0 # empty
    n = len(array)
    sum = 0. # ksum 
    max_sum = 0
    for i in range(n):
        sum += array[i]
        if i >= k:
            sum -= array[i-k]
            max_sum = max(sum, max_sum)
    return max_sum 



print(maxKsum([1,3,2,4],2))

# time: O(N), N_array length
# space: O(1), no additonal memeory 



# two non-overlapping k length subarray sum max 


def max2ksub(array, k):
    if not array:
        return 0
    n = len(array)
    if 2*k > n:
        return 0
    presum = [0] * (n+1). # optioze more 
    for i in range(n):
        presum[i+1]=presum[i]+array[i] # 0 - i-1
    # print(presum)
    ksum = [0]*(n-k+1)
    for i in range(n-k+1):
        ksum[i] = presum[i+k] - presum[i]
    # print(ksum)
    sum2 = 0
    max_sum2 = 0
    for i in range(n-k+1):
        for j in range (i+k,n-k+1):
            sum2 = ksum[i] + ksum[j]
            # print(ksum[i], ksum[j], sum2)
            max_sum2 = max(max_sum2, sum2)
    return max_sum2 


# print(max2ksub([1,3,2,5,6,8,4],2))

print(max2ksub([1,3,3,1],2))

# time O(N^2)
# space O(N). N - array length 


```
