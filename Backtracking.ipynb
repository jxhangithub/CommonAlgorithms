### General template

```python
The honest working code is given below. We include a global finished flag to allow for premature termination, which could be set in any application-specific routine.
bool finished = FALSE;               /* found all solutions yet? */

backtrack(int a[], int k, data input)
{
        int c[MAXCANDIDATES];        /* candidates for next position */
        int ncandidates;             /* next position candidate count */
        int i;                       /* counter */

        if (is_a_solution(a,k,input))
                process_solution(a,k,input);
        else {
                k = k+1;
                construct_candidates(a,k,input,c,&ncandidates);
                for (i=0; i<ncandidates; i++) {
                        a[k] = c[i];
                        backtrack(a,k,input);
                        if (finished) return;   /* terminate early */
                }
        }
}
```

### Subset

```python
# 78. Subsets
# https://leetcode.com/problems/subsets/
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        
        def help(i, temp):
            res.append(list(temp)) # step1. process solution Store the result at every node.
            if i == len(nums): # step 2 termination condition
                return
            for j in range(i, len(nums)): # step 3 construct more candidates
                temp.append(nums[j])
                help(j + 1, temp)
                temp.pop()     
        res = []    
        help(0, [])
        return res

class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        
        def help(i, temp):  
            if i == len(nums):
                res.append(list(temp)) # only store the result at the end
                return
            if i > len(nums):
                return
            temp.append(nums[i])
            help(i + 1, temp)
            temp.pop()
            help(i + 1, temp)
        res = []
        help(0, [])
        return res
```

### 77. Combinations

```python
77. Combinations
# Given two integers n and k, return all possible combinations of k numbers out of 1 ... n.

class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:      
        def help(temp, i):
            # step 1 process solution
            if len(temp) == k:
                res.append(list(temp))
                return
            # step 2 early termination condition
            if i > n:
                return
            
            # step 3 construct more solutions
            # one way to grow a tree, select and skip the i element
            temp.append(i) # since we are 1 based
            help(temp, i + 1)
            temp.pop()
            help(temp, i + 1)
        res = []
        help([], 1)
        return res
    
    
class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
        def help(temp, i):
            if len(temp) == k:
                res.append(list(temp))
                return
            
            if i == n + 1:
                return
            
            # anther way to grow the tree, skip different elements    
            for j in range(i, n + 1):
                temp.append(j) # since we are 1 based
                help(temp, j + 1)
                temp.pop()
        res = []
        help([], 1)
        return res    
```

### 46. Permutations

```python

#46. Permutations
#https://leetcode.com/problems/permutations/

# a naive solution
class Solution:
    """
    @param: nums: A list of integers.
    @return: A list of permutations.
    """
    def permute(self, nums):
        # write your code here
        
        # level can be removed since len(temp) == level
        def help(level, temp):
            
            # step1: process solution
            if len(temp) == len(nums):
                if len(set(temp)) == len(nums):
                    res.append(temp[:])
                return    
            # step2: early termination
            if level == len(nums):
                return
            # step3: construct other candidates
            for j in range(len(nums)):
                
                temp.append(nums[j])
                help(level + 1, temp)
                temp.pop()
                
        res = []
        help(0, [])
        return res





def permute(self, nums):
    def backtrack(start, end):
        if start == end:
            ans.append(nums[:])
        for i in range(start, end):
            nums[start], nums[i] = nums[i], nums[start]
            backtrack(start+1, end)
            nums[start], nums[i] = nums[i], nums[start]

    ans = []
    backtrack(0, len(nums))
    return ans   


class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
      
        def help(temp, used):
            if len(temp) == len(nums):
                res.append(list(temp))
                return               
            for j in range(len(nums)):
                if used[j] == 0:
                    temp.append(nums[j])
                    used[j] = 1
                    help(temp, used)
                    used[j] = 0
                    temp.pop()
                    
        used = [0 for _ in range(len(nums))]
        res = []
        help([], used)
        
        return res
```

```python
6. Permutations II
Given a list of numbers with duplicate number in it. Find all unique permutations.

# A dictionary solution
class Solution:
    """
    @param: :  A list of integers
    @return: A list of unique permutations
    """

    def permuteUnique(self, nums):
        # write your code here
        
        def help(count, temp):
            
            if len(temp) == len(nums):
                res.append(temp[:])
            else:
            
                for x in count: # here we must iterate over dict keys instead of nums
                    if count[x] > 0:
                        count[x] -= 1 
                        temp.append(x)
                        help(count, temp)
                        temp.pop()
                        count[x] += 1
                    
        
        nums.sort()
        res = []
        count = collections.Counter(nums)
        help(count, [])
        
        return res

class Solution:
    """
    @param: :  A list of integers
    @return: A list of unique permutations
    """

    def permuteUnique(self, nums):
        # write your code here
        
        def help(visited, temp):
            
            if len(temp) == len(nums):
                res.append(temp[:])
            else:
            
                for i in range(len(nums)):
                    if visited[i]:
                        continue
                    # a' a" b
                    # => a' a" b => √
                    # => a" a' b => x
                    # 不能跳过一个a选下一个a
                    if i > 0 and nums[i] == nums[i - 1] and not visited[i - 1]:
                        continue
                    
                    visited[i] = True
                    temp.append(nums[i])
                    help(visited, temp)
                    temp.pop()
                    visited[i] = False
                    
        
        nums.sort()
        
        res = []
        visited = [False] * len(nums)
        help(visited, [])
        
        return res
```

### 90. Subsets II (uniqueness)

### 491. Increasing Subsequences
Given an integer array, your task is to find all the different possible increasing subsequences of the given array, and the length of an increasing subsequence should be at least 2.

```python
class Solution:
    def findSubsequences(self, nums: List[int]) -> List[List[int]]:
        
        def help(i, temp):
            if len(temp) >= 2:
                    res.append(list(temp))
            
            used = set()
            for j in range(i, len(nums)):
                # note that using j > i, nums[j] == nums[j - 1] to skip is not sufficient because we cannot sort nums
                if nums[j] in used:
                    continue
                
                if not temp or nums[j] >= temp[-1]:
                    used.add(nums[j])
                    temp.append(nums[j])
                    help(j + 1, temp)
                    temp.pop()
        # note that we cannot sort
        res = []
        help(0, [])
        return res
```

### 17. Letter Combinations of a Phone Number

```python
17. Letter Combinations of a Phone Number
Given a string containing digits from 2-9 inclusive, return all possible letter combinations that the number could represent.

A mapping of digit to letters (just like on the telephone buttons) is given below. Note that 1 does not map to any letters.

class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        
        if not digits:
            return []
        
        d2L = {}
        d2L['2'] = 'abc'
        d2L['3'] = 'def'
        d2L['4'] = 'ghi'
        d2L['5'] = 'jkl'
        d2L['6'] = 'mno'
        d2L['7'] = 'pqrs'
        d2L['8'] = 'tuv'
        d2L['9'] = 'wxyz'
        
        res = []
        
        def help(res, i, temp):
            
            if i == len(digits):
                res.append(''.join(temp))
                return
                
            
            for c in d2L[digits[i]]:
                temp.append(c)
                help(res, i + 1, temp)
                temp.pop()
        
        help(res, 0, [])
        
        return res
```

### 39. Combination Sum (no duplicate, unlimited choice and unique final result)
Given a set of candidate numbers (candidates)  and a target number (target), 
find all unique combinations in candidates where the candidate numbers sums to target.

The same repeated number may be chosen from candidates unlimited number of times.
The solution set must not contain duplicate combinations.

```python
class Solution:
   
    def combinationSum(self, num, target):
        # write your code here
        def help(temp, i, target):
            if target == 0:
                res.append(temp[:])
            elif i < len(num) and target >= num[i]: # eary pruning
                for j in range(i, len(num)):
                    # critical: do not use the same element in the same level of backtracking tree
                    # for example, if the num is 1, 1, 2, 3, 4, in the depth direction, we will check 1, 2, 3, 4 and target - 1; 
                    # then 2, 3, 4 and target - 1 -2.
                    # therefore, in the horizon direction, we should avoid check 1, 2, 3, 4 and target -1
                    if j > i and num[j] == num[j - 1]:
                        continue
                    temp.append(num[j])
                    help(temp, j, target - num[j])
                    temp.pop()
        
        num.sort()      
        res = []
        help([], 0, target)
        return res
    
    
```

### 40. Combination Sum II (positive, contain duplicate and unique final result)
#40. Combination Sum II
Given a set of candidate numbers (candidates)  and a target number (target), 
find all unique combinations in candidates where the candidate numbers sums to target.

The solution set must not contain duplicate combinations.
All numbers (including target) will be positive integers.


```python
class Solution:
   
    def combinationSum(self, num, target):
        # write your code here
        def help(temp, i, target):
            if target == 0:
                res.append(temp[:])
            elif i < len(num) and target >= num[i]: # eary pruning
                for j in range(i, len(num)):
                    if j > i and num[j] == num[j - 1]:
                        continue
                    temp.append(num[j])
                    help(temp, j + 1, target - num[j]) # this is the difference from combinationSum I, use J + 1 since every element get chosen onece
                    temp.pop()
        
        num.sort()      
        res = []
        help([], 0, target)
        return res  
    
# use dictionary method
class Solution:
    def combinationSum2(self, num, target):
        # write your code here
        def help(temp, i, target, count):
            if target == 0:
                res.append(temp[:])
            elif target >= keys[i]: #pruning
                for j in range(i, len(keys)):
                    num = keys[j]
                    if count[num] > 0:
                        count[num] -= 1
                        temp.append(num)
                        help(temp, j, target - num, count)
                        temp.pop()
                        count[num] += 1
        
        keys = list(set(num))
        keys.sort() # sort the keys to enable early pruning
        count = collections.Counter(num)
        res = []
        help([], 0, target, count)
        return res
```

### 51. N-Queens

```python
51. N-Queens
Given an integer n, return all distinct solutions to the n-queens puzzle.

Each solution contains a distinct board configuration of the 
n-queens' placement, where 'Q' and '.' both indicate a queen and an empty space respectively.
class Solution:
    def solveNQueens(self, n: int) -> List[List[str]]:
        # write your code here
        results = []
        
        
        self.dfs(n, 0, [], results )
        
        return results
        
        
    def dfs(self, n, rowIdx, queens, results):
    
        if rowIdx == n:
            results.append(self.queenPos2Board(queens, n))
        else:
            for colIdx in range(n):
                # only add valid solutions
                if self.isValid(rowIdx, colIdx, queens):
                    queens.append((rowIdx, colIdx))
                    self.dfs(n, rowIdx + 1, queens, results)
                    queens.pop()
                    
    
    def isValid(self, i, j, queens):
        
        for q in queens:
            if i == q[0] or j == q[1]:
                return False
            if abs(q[0] - i) == abs(q[1] - j):
                return False
                
        return True
    
    
        
    def queenPos2Board(self, pos, n):
        grid = []
        for i in range(n):
            string = ['.' for _ in range(n)]
            string[pos[i][1]] = 'Q'
            grid.append(''.join(string))


        return grid
        
        
```

### 698. Partition to K Equal Sum Subsets

```python
Given an array of integers nums and a positive integer k, find whether it's possible to divide this array into k non-empty subsets whose sums are all equal.

 

Example 1:

Input: nums = [4, 3, 2, 3, 5, 2, 1], k = 4
Output: True
Explanation: It's possible to divide it into 4 subsets (5), (1, 4), (2,3), (2,3) with equal sums.
    
class Solution:
    def canPartitionKSubsets(self, nums: List[int], k: int) -> bool:
        totalSum = sum(nums) 
        subSumTarget = totalSum // k
        if k * subSumTarget != totalSum:
            return False
        
        
        def help(subSumTarget, sumGroup, k, idx):
            
            if len(nums) == idx:
                return True
            else:
                
                for i in range(k):
                    if sumGroup[i] + nums[idx] <= subSumTarget:
                        sumGroup[i] += nums[idx]
                        if help(subSumTarget, sumGroup, k, idx + 1):
                            return True
                        sumGroup[i] -= nums[idx]
                        if sumGroup[i] == 0: # because order does not matter, only need to search once for each element
                            break
                return False
        # additional steps to reduce complexity
        nums.sort()
        if nums[-1] > subSumTarget: return False
        while nums and nums[-1] == subSumTarget:
            nums.pop()
            k -= 1
        nums.reverse() # consider large elements first reduce the branching at the beginning
        return help(subSumTarget, [0] * k, k, 0)    
```

### 679. 24 Game

```python
class Solution:
    def judgePoint24(self, operands: List[int]) -> bool:
                
        if len(operands) == 1:
            if abs(operands[0] - 24) < 1e-6:
                return True
            else:
                return False
        else:
            operandsCopy = [e for e in operands]
            for i in range(len(operands)):
                for j in range(len(operands)):
                    if i != j:
                        operandsCopy.remove(operands[i])
                        operandsCopy.remove(operands[j])
                        for operator in ["+", "-", "*", "."]:
                            if operator == '.' and operands[j] == 0:
                                continue
                            if operator == '+':
                                temp = operands[i] + operands[j]
                            elif operator == '-':
                                temp = operands[i] - operands[j]
                            elif operator == '*':
                                temp = operands[i] * operands[j]   
                            elif operator == '.':
                                temp = operands[i] / operands[j]
                            operandsCopy.append(temp)
                            if self.judgePoint24(operandsCopy):
                                return True
                            operandsCopy.pop()
                        operandsCopy.append(operands[i])
                        operandsCopy.append(operands[j])
                            
            return False
```

### 329. Longest Increasing Path in a Matrix

```python
Given an integer matrix, find the length of the longest increasing path.

From each cell, you can either move to four directions: left, right, up or down.
    You may NOT move diagonally or move outside of the boundary (i.e. wrap-around is not allowed).

class Solution:
    def longestIncreasingPath(self, matrix: List[List[int]]) -> int:
        
        if not matrix:
            return 0
        
        memo = {}
        H = len(matrix)
        W = len(matrix[0])
        self.maxVal = 0
        
        def help(i, j):
            
            if (i, j) in memo:
                return memo[(i, j)]
            
            memo[(i, j)] = 1
            
            for dx, dy in (-1, 0), (1, 0), (0, -1), (0, 1):
                newX = i + dx
                newY = j + dy
                if 0 <= newX < H and 0 <= newY < W and matrix[newX][newY] > matrix[i][j]:
                    memo[(i, j)] = max(memo[i, j], 1 + help(newX, newY))
            
            self.maxVal = max(self.maxVal, memo[(i, j)])
            return memo[(i, j)]
        
        for i in range(H):
            for j in range(W):
                help(i, j)
        
        return self.maxVal
```

```python
# recursive dfs with memo
class Solution:
    def isInterleave(self, s1: str, s2: str, s3: str) -> bool:
        
        memo = {}
        
        def dfs(i, j, k):
            # note that k = i + j
            if i == len(s1) and j == len(s2):
                return True
            else:
                
                if (i, j) in memo:
                    return memo[(i, j)]
                
                if i < len(s1) and s1[i] == s3[k] and dfs(i + 1, j, k + 1):
                    memo[(i, j)] = True
                    return True
                if j < len(s2) and s2[j] == s3[k] and dfs(i, j + 1, k + 1):
                    memo[(i, j)] = True
                    return True
                memo[(i, j)] = False
                return False

        if len(s1) + len(s2) != len(s3):
            return False
        
        return dfs(0, 0, 0)

# iteartive dfs
# note that visited serves as memo, if a node in visited, then any path starting from it will fail, so no need to explore
class Solution:
    def isInterleave(self, s1: str, s2: str, s3: str) -> bool:
        
        if len(s1) + len(s2) != len(s3):
            return False
        stack = [(0, 0, 0)]
        visited = set([(0, 0, 0)])
        
        while stack:            
            i, j, k = stack.pop()
            # note that k = i + j
            if i == len(s1) and j == len(s2):
                return True
                
            if i < len(s1) and s1[i] == s3[k] and (i + 1, j, k + 1) not in visited:
                visited.add((i + 1, j, k + 1))
                stack.append((i + 1, j, k + 1))
            if j < len(s2) and s2[j] == s3[k] and (i, j + 1, k + 1) not in visited:
                visited.add((i, j + 1, k + 1))
                stack.append((i, j + 1, k + 1))

        return False
```

### 741. Cherry Pickup

```python
class Solution:
    def cherryPickup(self, grid: List[List[int]]) -> int:
        
        
        dp = {}
        H = len(grid)
        W = len(grid[0])        
        def dfs(i, j, m):
            if (i, j, m) in dp:
                return dp[i, j, m]
            n = i + j - m
            if not (0 <= i < H and 0 <= j < W and 0 <= m < H and 0 <= n < W):
                dp[i, j, m] = -float('inf')
                return dp[i, j, m]
            if grid[i][j] == -1 or grid[m][n] == -1:
                dp[i, j, m] = -float('inf')
                return dp[i, j, m]
            
            if (i, j) == (H - 1, W - 1) and (m, n) == (H - 1, W - 1):
                return grid[i][j]
            
            dp[i, j, m] = -float('inf')
            dp[i, j, m] = max(dp[i, j, m], dfs(i + 1, j, m + 1))
            dp[i, j, m] = max(dp[i, j, m], dfs(i, j + 1, m + 1))
            dp[i, j, m] = max(dp[i, j, m], dfs(i + 1, j, m))
            dp[i, j, m] = max(dp[i, j, m], dfs(i, j + 1, m))
            if (i, j) != (m, n):
                dp[i, j, m] += grid[i][j] + grid[m][n]
            else:
                dp[i, j, m] += grid[i][j]
            return dp[i, j, m]
        result = dfs(0, 0, 0)
        return result if result != -float('inf') else 0
```

### 87. Scramble String

```python
87. Scramble String
Hard

435

657

Add to List

Share
Given a string s1, we may represent it as a binary tree by partitioning it to two non-empty substrings recursively.

Below is one possible representation of s1 = "great":

    great
   /    \
  gr    eat
 / \    /  \
g   r  e   at
           / \
          a   t
To scramble the string, we may choose any non-leaf node and swap its two children.

class Solution:
    def isScramble(self, s1: str, s2: str) -> bool:
        
        if s1 == s2:
            return True

        if len(s1) != len(s2):
            return False
        
        countS1 = collections.Counter(s1)
        countS2 = collections.Counter(s2)
        
        if countS1 != countS2:
            return False

        Len = len(s1)
        for k in range(1, Len):
            
            
            if self.isScramble(s1[:k], s2[:k]) and self.isScramble(s1[k:], s2[k:]):
                return True
            
            if self.isScramble(s1[:k], s2[Len - k:]) and self.isScramble(s1[k:], s2[:Len - k]):
                return True
            
        return False
            
```

### 403. Frog Jump
A frog is crossing a river. The river is divided into x units and at each unit there may or may not exist a stone. The frog can jump on a stone, but it must not jump into the water.

Given a list of stones' positions (in units) in sorted ascending order, determine if the frog is able to cross the river by landing on the last stone. Initially, the frog is on the first stone and assume the first jump must be 1 unit.

If the frog's last jump was k units, then its next jump must be either k - 1, k, or k + 1 units. Note that the frog can only jump in the forward direction.

```python
class Solution:
    def canCross(self, stones: List[int]) -> bool:
        
        self.memo ={}
        return self.dfs(stones[0], 0, stones, [1])
    
    
    def dfs(self, pos, idx, stones, jumps):
        
        if idx == len(stones) - 1:
            return True
        else:
            
            state = tuple([idx] + jumps)
            if state in self.memo:
                return self.memo[state]
            
            # the start idx can be further optimized by binary search
            for j in range(idx + 1, len(stones)):
                
                jumpSize = stones[j] - pos
                if jumpSize > max(jumps):
                    break
                if jumpSize in jumps:
                    newJumps = [jumpSize - 1, jumpSize, jumpSize + 1]
                    if self.dfs(stones[j], j, stones, newJumps):
                        self.memo[state] = True
                        return True
            self.memo[state] = False
            return False
        
```

```python
332. Reconstruct Itinerary
Medium

1342

825

Add to List

Share
Given a list of airline tickets represented by pairs of departure and arrival airports [from, to], reconstruct the itinerary in order. All of the tickets belong to a man who departs from JFK. Thus, the itinerary must begin with JFK.

Note:

If there are multiple valid itineraries, you should return the itinerary that has the smallest lexical order when read as a single string. For example, the itinerary ["JFK", "LGA"] has a smaller lexical order than ["JFK", "LGB"].
All airports are represented by three capital letters (IATA code).
You may assume all tickets form at least one valid itinerary.


class Solution:
    def findItinerary(self, tickets: List[List[str]]) -> List[str]:
        
        G = collections.defaultdict(list)
        
        numTrips = len(tickets)
        
        for t in tickets:
            
            G[t[0]].append(t[1])
            
        for k in G:
            G[k].sort()
            
            
            
        
        def help(stop, stopNum, trip):
            
            if stopNum == numTrips:
                
                self.res = list(trip)
                
                return True
            
            else:
                
                dest = list(G[stop])
                
                while dest:
                    
                    node = dest.pop(0)
                    
                    G[stop].pop(0) # G also serve as the marker marking if the ticket is used in the path
                    
                    trip.append(node)
                    
                    if help(node, stopNum + 1, trip):
                        return True
                    
                    trip.pop()
                    
                    G[stop].append(node)
                    
                return False
            
        help("JFK", 0, ["JFK"])
        
        return self.res
                    
            
            
        
        
        
```
