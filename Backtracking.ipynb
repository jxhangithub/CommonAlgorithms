```python
22. Generate Parentheses

Given n pairs of parentheses, write a function to generate all combinations of well-formed parentheses.

class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        
        if n <= 0:
            return []
        
        
        res = []
        
        
        def help(res, i, leftCount, rightCount, temp):
            
            if i == 2 * n:
                res.append(''.join(temp))
                return
            
            if leftCount < n:
                temp.append('(')
                leftCount += 1
                help(res, i + 1, leftCount, rightCount, temp)
                temp.pop()
                leftCount -= 1
            
            if rightCount < leftCount:
                temp.append(')')
                rightCount += 1
                help(res, i + 1, leftCount, rightCount, temp)
                temp.pop()
                rightCount -= 1
                
                
        help(res, 0, 0, 0, [])
        
        return res
```

```python
416. Partition Equal Subset Sum

Given a non-empty array containing only positive integers, find if the array 
can be partitioned into two subsets such that the sum of elements in both subsets is equal.

class Solution:
    def canPartition(self, nums: List[int]) -> bool:
        
        totalSum = sum(nums)
        
        if totalSum % 2 == 1:
            return False
        targetSum = totalSum // 2
        
        memo = {}
        
        def help(i, tempSum):
            
            if i == len(nums):
                if tempSum == targetSum:
                    return True
                else:
                    return False
                
            if (i, tempSum) in memo:
                return memo[(i, tempSum)]
            
            if tempSum == targetSum:
                memo[(i, tempSum)] = True
                return True
            
            
            if help(i + 1, tempSum) or help(i + 1, tempSum + nums[i]):
                memo[(i, tempSum)] = True
                return True
            
            memo[(i, tempSum)] = False
            return False
        
        return help(0, 0)
        
```

```python
494. Target Sum
You are given a list of non-negative integers, a1, a2, ..., an, and a target, S. Now you have 2 symbols + and -. For each integer, you should choose one from + and - as its new symbol.

Find out how many ways to assign symbols to make sum of integers equal to target S.

class Solution:
    def findTargetSumWays(self, nums: List[int], S: int) -> int:
        
        
        memo = {}
        
        def help(i, target):
            
            if i == len(nums):
                if target == 0:
                    return 1
                else:
                    return 0
            else:
                if (i, target) in memo:
                    return memo[(i, target)]
                
                ways = help(i + 1, target + nums[i]) + help(i + 1, target - nums[i])
                
                memo[(i, target)] = ways
                
                return ways
                
                
        return help(0, S)
                
```

```python
Give a dictionary of words and a sentence with all whitespace removed, return the number of
sentences you can form by inserting whitespaces to the sentence so that each word can be found in the dictionary.

## ignore cases!!
Input:
"CatMat"
["Cat", "Mat", "Ca", "tM", "at", "C", "Dog", "og", "Do"]
Output: 3
Explanation:
we can form 3 sentences, as follows:
"CatMat" = "Cat" + "Mat"
"CatMat" = "Ca" + "tM" + "at"
"CatMat" = "C" + "at" + "Mat"

class Solution:
    """
    @param: : A string
    @param: : A set of word
    @return: the number of possible sentences.
    """

    def wordBreak3(self, s, dict):
        # Write your code here
        
        dp = [0] * (len(s) + 1)
        
        dp[0] = 1
        
        s = s.lower()
        dict = [w.lower() for w in dict]
        
        for i in range(1, len(s) + 1):
            for j in range(0, i):
                if s[j : i] in dict:
                    dp[i] += dp[j]
                    
                    
        
        return dp[-1]
```
