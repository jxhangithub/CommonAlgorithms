### Implement substring matching (KMP)
KMP algorithm preprocesses pat[] and constructs an auxiliary lps[] of size m (same as size of pattern) which is used to skip characters while matching.
name lps indicates longest proper prefix which is also suffix.. A proper prefix is prefix with whole string not allowed. For example, prefixes of “ABC” are “”, “A”, “AB” and “ABC”. Proper prefixes are “”, “A” and “AB”. Suffixes of the string are “”, “C”, “BC” and “ABC”.
We search for lps in sub-patterns. More clearly we focus on sub-strings of patterns that are either prefix and suffix.
For each sub-pattern pat[0..i] where i = 0 to m-1, lps[i] stores length of the maximum matching proper prefix which is also a suffix of the sub-pattern pat[0..i].

lps[i] = the longest proper prefix of pat[0..i] which is also a suffix of pat[0..i].

```python
def KMPSearch(pat, txt):
    
    M = len(pat)
    N = len(txt)
    
    # lps[i] = the longest proper prefix of pat[0..i] which is also a suffix of pat[0..i]. 
    lps = computeLPS(pat)
    
    i = 0
    j = 0
    while i < N:
        if pat[j] == txt[i]:
            i += 1
            j += 1
        if j == M:
            print("Found pattern at index ", i - j)
            j = lps[j - 1]
        elif i < N and pat[j] != txt[i]:
            if j > 0:
                j = lps[j - 1]
            else:# if at j = 0 we do not match, just move to next i
                i += 1
                
    
def computeLPS(pat):
    lps = [0] * len(pat)
    
    i = 1
    L = 0 # length of the previous longest prefix suffix 
    # the loop calculates lps[i] for i = 1 to M-1 
    while i < len(pat):
        if pat[i] == pat[L]:
            L += 1
            lps[i] = L
            i += 1
        else:
            
            if L > 0:
                L = lps[L - 1]
            else:
                lps[i] = 0
                i += 1
    print(lps)
    return lps
                
txt = "ABABDABACDABABCABAB"
pat = "ABABCABAB"
pat = "AAACAAAA"
KMPSearch(pat, txt) 
```

### 214. Shortest Palindrome
Given a string s, you are allowed to convert it to a palindrome by adding characters in front of it. Find and return the shortest palindrome you can find by performing this transformation.

```python
# brute force solution
class Solution:
    def shortestPalindrome(self, s: str) -> str:
        
        longest = 1
        for i in range(len(s), 0, -1):
            t = s[:i]
            if t == t[::-1]:
                longest = i
                break
        complement = s[longest:] 
        return complement[::-1] + s

# KMP solution
class Solution:
    def shortestPalindrome(self, s: str) -> str:
        sold = s
        s = s + '#' + s[::-1]
        
        i = 1
        L = 0
        lps = [0] * len(s)
        while i < len(s):
            if s[i] == s[L]:
                L += 1
                lps[i] = L
                i += 1
            else:
                if L > 0:
                    L = lps[L - 1]
                else:
                    i += 1
        longest = lps[-1]
        complement = sold[longest:] 
        return complement[::-1] + sold
```

### 796. Rotate String

We are given two strings, A and B.

A shift on A consists of taking string A and moving the leftmost character to the rightmost position. For example, if A = 'abcde', then it will be 'bcdea' after one shift on A. Return True if and only if A can become B after some number of shifts on A.

```python
# Consider a string S="helloworld". Now, given another string T="lloworldhe", can we figure out if T is a rotated version of S? Yes, we can! We check if S is a substring of T+T.
class Solution:
    def rotateString(self, A: str, B: str) -> bool:
        if len(A) != len(B):
            return False
        temp = B + B
        return temp.find(A) != -1
    
# use KMP to search for substring search B in A + A    
class Solution:
    def rotateString(self, A: str, B: str) -> bool:
        if len(A) != len(B):
            return False
        
        if not A:
            return True
        
        lps = [0] * len(A)
        i = 1
        L = 0
        while i < len(A):
            if A[i] == A[L]:
                L += 1
                lps[i] = L
                i += 1
            else:
                if L > 0:
                    L = lps[L - 1]
                else:
                    i += 1
        
        temp = B + B
        i, j = 0, 0
        while i < len(temp):
            if A[j] == temp[i]:
                i += 1
                j += 1
                if j == len(A):
                    return True
            else:
                if j > 0:
                    j = lps[j - 1]
                else:
                    i += 1
                    
        return False
```

### 28. Implement strStr()
Implement strStr().

Return the index of the first occurrence of needle in haystack, or -1 if needle is not part of haystack.

```python
class Solution:
    def strStr(self, text: str, pat: str) -> int:
        
        if not pat:
            return 0
        
        if len(text) < len(pat):
            return -1
        
        base = 32
        mod = int(1e9 + 7)
        
        pat_hash = 0
        txt_hash = 0
        
        for i in range(len(pat)):
            pat_hash = (pat_hash * base + ord(pat[i])) % mod
            txt_hash = (txt_hash * base + ord(text[i])) % mod
        
        # when hash matches, we double check
        if pat_hash == txt_hash and text[:len(pat)] == pat:
            return 0
        
        power = 1
        for _ in range(len(pat) - 1):
            power = (power * base) % mod
        # power = base ** (len(pat) - 1) % mod
        
        for i in range(1, len(text) - len(pat) + 1):
            txt_hash = txt_hash - power * ord(text[i - 1])
            txt_hash = (txt_hash * base + ord(text[i + len(pat) - 1])) % mod
            if pat_hash == txt_hash and text[i:i+len(pat)] == pat:
                return i
            
        return -1
            
        
        
        
```

### 187. Repeated DNA Sequences

The DNA sequence is composed of a series of nucleotides abbreviated as 'A', 'C', 'G', and 'T'.

For example, "ACGAATTCCG" is a DNA sequence.
When studying DNA, it is useful to identify repeated sequences within the DNA.

Given a string s that represents a DNA sequence, return all the 10-letter-long sequences (substrings) that occur more than once in a DNA molecule. You may return the answer in any order.

```python
class Solution:
    def findRepeatedDnaSequences(self, s: str) -> List[str]:
        
        seen = defaultdict(int)
        
        base = 4
        k = 10
        factor = pow(4, k - 1)
            
        to_int = {'A': 0, 'C': 1, 'G': 2, 'T': 3}
        hash_code = 0
        res = []
        for i in range(len(s)):
            
            if i >= k:
                hash_code = (hash_code - factor * to_int[s[i - k]]) 
                
            hash_code = (hash_code * base + to_int[s[i]]) 
            
            if i >= k - 1:
                if seen[hash_code] == 1:
                    res.append(s[i - k + 1: i + 1])
                    seen[hash_code] += 1
                else:
                    seen[hash_code] += 1
                    
        return res
```

### 1044. Longest Duplicate Substring

Given a string s, consider all duplicated substrings: (contiguous) substrings of s that occur 2 or more times. The occurrences may overlap.

Return any duplicated substring that has the longest possible length. If s does not have a duplicated substring, the answer is ""

```python
class Solution:
    def longestDupSubstring(self, s: str) -> str:
        
        def search(k):
            
            base = 26
            mod = int(1e9 + 7)
            factor = pow(base, k - 1, mod)
            seen = defaultdict(list)
            hash_code = 0
            for i in range(len(s)):
                
                if i >= k:
                    hash_code = (hash_code - factor * ord(s[i - k])) % mod
                    
                hash_code = (hash_code * base + ord(s[i])) % mod
                
                if i >= k - 1:
                    if hash_code not in seen:
                        seen[hash_code].append(i)
                    else:
                        # further verfication
                        curr = s[i - k + 1: i + 1]
                        if any(curr == s[index - k + 1 : index + 1] for index in seen[hash_code]):
                            return curr
            return ""
        
        left, right = 1, len(s)
        while left < right:
            
            mid = (left + right + 1) // 2
            
            res = search(mid)
            if len(res) > 0:
                left = mid
            else:
                right = mid - 1
                
        return search(left)
            
```

### 718. Maximum Length of Repeated Subarray

Given two integer arrays A and B, return the maximum length of an subarray that appears in both arrays.

```python
class Solution:
    def findLength(self, A: List[int], B: List[int]) -> int:
        
        left, right = 0, min(len(A), len(B))
        
        while left < right:
            mid = (left + right + 1) // 2
            
            if self.check(A, B, mid):
                left = mid
            else:
                right = mid - 1
        
        return left
        
        
    def fastPower(self, a, power, mod):
        
        res = 1
        while power:
            if power & 1 == 1:
                res = (res * a) % mod
            a = a * a % mod
            power = power >> 1
        return res
            
        
    def check(self, s, t, k):
        
        base = 233 # use 31 might fail some cases
        mod = int(1e9 + 7)
        hashes = set()
        prePower = self.fastPower(base, k, mod)
        hash_s = 0
        for i in range(len(s)):
            hash_s = (base * hash_s + s[i]) % mod
            if i >= k:
                hash_s = (hash_s - prePower * s[i - k]) % mod
            if i >= k - 1:
                hashes.add(hash_s)
        
        hash_t = 0
        for i in range(len(t)):
            hash_t = (base * hash_t + t[i]) % mod
            if i >= k:
                hash_t = (hash_t - prePower * t[i - k]) % mod
            if i >= k - 1:
                if hash_t in hashes:
                    return True
                
        return False
        
```

### 151. Reverse Words in a String
A word is defined as a sequence of non-space characters.
Input string may contain leading or trailing spaces. However, your reversed string should not contain leading or trailing spaces.
You need to reduce multiple spaces between two words to a single space in the reversed string.


```python
class Solution:
    def reverseWords(self, s: str) -> str:
        
        def split(s):
            
            first = 0
            res = []
            while first < len(s):
                
                while first < len(s) and s[first] == " ":
                    first += 1
                
                second = first
                while second < len(s) and s[second] != " ":
                    second += 1
                
                if second != first:
                    res.append(s[first: second])
                    
                first = second
                
            return res

        res = split(s)
        res.reverse()
        
        return ' '.join(res)
```

```python
647. Palindromic Substrings

Given a string, your task is to count how many palindromic substrings in this string.

The substrings with different start indexes or end indexes are counted as different substrings even they consist of same characters.

class Solution:
    def countSubstrings(self, s: str) -> int:
        
        def count(i, j):
            ans = 0
            while i >= 0 and j < len(s) and s[i] == s[j]:
                i -= 1
                j += 1
                ans += 1
            return ans
        
        res = 0
        for i in range(len(s)):
            res += count(i, i) + count(i, i + 1)
            
        return res
```

```python
345. Reverse Vowels of a String

Write a function that takes a string as input and reverse only the vowels of a string.

class Solution:
    def reverseVowels(self, s: str) -> str:
        vowels = set(list("aeiouAEIOU"))
        s = list(s)
        
        first = 0
        second = len(s) - 1
        
        while first < second:
            
            while first < second and s[first] not in vowels:
                first += 1
            while first < second and s[second] not in vowels:
                second -= 1
                
            s[first], s[second] = s[second], s[first]
            first, second = first + 1, second - 1
            
        return "".join(s)
```

### 758. Bold Words in String
Given a set of keywords words and a string S, make all appearances of all keywords in S bold. Any letters between <b> and </b> tags become bold.

The returned string should use the least number of tags possible, and of course the tags should form a valid combination.

For example, given that words = ["ab", "bc"] and S = "aabcd", we should return "a<b>abc</b>d". Note that returning "a<b>a<b>b</b>c</b>d" would use more tags, so it is incorrect.

```python
class Solution:
    def boldWords(self, words: List[str], S: str) -> str:
        
        bold = set()
        
        for w in words:
            for i in range(len(S) - len(w) + 1):
                if S[i] == w[0] and S[i: i + len(w)] == w:
                    bold.update([i + j for j in range(len(w))])
        
        
        res = ''
        for i in range(len(S)):
            if i not in bold and (i - 1) in bold:
                res += '</b>' + S[i] 
            elif i in bold and (i - 1) not in bold:
                res += '<b>' + S[i]
            else:
                res += S[i]
        
        if len(S) - 1 in bold:
            res += '</b>'
        
        return res
```

### 68. Text Justification

Given an array of strings words and a width maxWidth, format the text such that each line has exactly maxWidth characters and is fully (left and right) justified.

You should pack your words in a greedy approach; that is, pack as many words as you can in each line. Pad extra spaces ' ' when necessary so that each line has exactly maxWidth characters.

Extra spaces between words should be distributed as evenly as possible. If the number of spaces on a line does not divide evenly between words, the empty slots on the left will be assigned more spaces than the slots on the right.

For the last line of text, it should be left-justified and no extra space is inserted between words.

```python
class Solution:
    def fullJustify(self, words: List[str], maxWidth: int) -> List[str]:
        
        res, curr_list, num_of_letters = [], [], 0
        
        for word in words:
            
            # if cannot fill current line
            # len(curr_list) is the minimum need of spaces
            # num_of_letters are the total number of characters
            if num_of_letters + len(word) + len(curr_list) > maxWidth:
                size = max(1, len(curr_list) - 1)
                
                for i in range(maxWidth - num_of_letters):
                    
                    index = i % size
                    curr_list[index] += ' '
                    
                res.append(''.join(curr_list))
                curr_list, num_of_letters = [], 0
                
            curr_list.append(word)
            num_of_letters += len(word)
        
        last_line = ' '.join(curr_list)
        last_line = last_line + ' '*(maxWidth - len(last_line))
        res.append(last_line)
        
        #res.append(' '.join(curr_list).ljust(maxWidth))
        
        return res
```

### 616. Add Bold Tag in String

You are given a string s and an array of strings words. You should add a closed pair of bold tag <b> and </b> to wrap the substrings in s that exist in words. If two such substrings overlap, you should wrap them together with only one pair of closed bold-tag. If two substrings wrapped by bold tags are consecutive, you should combine them.

Return s after adding the bold tags.

```python
class Solution:
    def addBoldTag(self, s: str, words: List[str]) -> str:
        
        
        bold = [False] * len(s)
        
        for w in words:
            start = s.find(w)
            
            while start != -1:
                for i in range(start, start + len(w)):
                    bold[i] = True
                
                start = s.find(w, start + 1)
                
        
        output = []
        
        for i in range(len(s)):
            
            if i == 0 and bold[i]:
                output.append('<b>')
            elif i > 0:
                if not bold[i - 1] and bold[i]:
                    output.append('<b>')
                elif bold[i - 1] and not bold[i]:
                    output.append('</b>')

            output.append(s[i])
                
        if bold[-1]:
            output.append('</b>')
        return ''.join(output)
```
