A string s matches a pattern if there is some bijective mapping of single characters to non-empty strings such that if each character in pattern is replaced by the string it maps to, then the resulting string is s. A bijective mapping means that no two characters map to the same string, and no character maps to two different strings.

 

Example 1:

Input: pattern = "abab", s = "redblueredblue"
Output: true
Explanation: One possible mapping is as follows:
'a' -> "red"
'b' -> "blue"
Example 2:

Input: pattern = "aaaa", s = "asdasdasdasd"
Output: true
Explanation: One possible mapping is as follows:
'a' -> "asd"
Example 3:

Input: pattern = "aabb", s = "xyzabcxzyabc"
Output: false
 

Constraints:

1 <= pattern.length, s.length <= 20
pattern and s consist of only lowercase English letters.

```python
class Solution:
    def wordPatternMatch(self, p: str, s: str) -> bool:
        
        
        def helper(i, j, mappings, reverse_set):
            
            if i == len(p) and j == len(s):
                return True
            elif i < len(p) and j < len(s):
                if p[i] in mappings:
                    word = mappings[p[i]]
                    if s[j: j + len(word)] == word and helper(i + 1, j + len(word), mappings, reverse_set):
                        return True
                else:
                    for k in range(j + 1, len(s) + 1):
                        word = s[j: k]
                        if word not in reverse_set:
                            mappings[p[i]] = word
                            reverse_set.add(word)
                            if helper(i + 1, k, mappings, reverse_set):
                                return True
                            del mappings[p[i]]
                            reverse_set.discard(word)
                    
                    
                    return False
            else:
                return False
            
        return helper(0, 0, {}, set())
```

Given an input string which is the output of a count and say method, return the original number.

For example: If the number if "21", then the count and say method would return
"1211" (one two, one one). In this problem, the input provided to us is "1211" and our goal is to return "21".

I answered this question by approaching it as a variant of the WordBreak problem with a dynamic programming solution. Of course, the conditions to check the validity of the split string will differ. Thoughts?

```python
def say_count(count_str: str) -> str:
    cache = collections.defaultdict(list)
    size = len(count_str)
    def help(idx):
        if idx in cache:
            return cache[idx]
        if idx == size:
            cache[idx] = ['']
            return cache[idx]
        for j in range(idx+2, size+1):
            if j == size -1: # at least 2 chars for the rest
                continue
            cache[j] = help(j)
            cnt = int(count_str[idx:j-1]) 
            word = count_str[j-1:j] * cnt
            for rest in cache[j]:
                cache[idx].append(word+rest)
        return cache[idx]
    help(0)
    return cache[0]
print(say_count('1211'))

# ['21', '1111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111']
```

1 <= wordDict[i].length <= 10

Input is generated in a way that the length of the answer doesn't exceed 105.

```python
class Solution:
    def wordBreak(self, s: str, wordDict: List[str]) -> List[str]:
        
        
        cache = collections.defaultdict(list)
        wordset = set(wordDict)
        
        
        def help(i):
            if i in cache:
                return cache[i]
            
            if i == len(s):
                cache[i] = ['']
                return cache[i]
            else:
                for j in range(i + 1, len(s) + 1):
                    w = s[i:j]
                    if w in wordset:
                        sentences = help(j)
                        for sentence in sentences:
                            if sentence:
                                cache[i].append(w + " " +sentence)
                            else:
                                cache[i].append(w)
            return cache[i]
        
        return help(0)
```

```python
import heapq
class Solution:
    def merge(self, arrays: List[List[int]]) -> List:
        min_heap = []
        for i, array in enumerate(arrays):
            if array:
                min_heap.append((array[0], i, 0))
        heapq.heapify(min_heap)
        res = []
        while min_heap:
            num, array_idx, idx = heapq.heappop(min_heap)
            res.append(num)
            if len(arrays[array_idx]) - 1 > idx:
                heapq.heappush(min_heap, (arrays[array_idx][idx+1], array_idx, idx+1))
        return res
    
s=Solution()
arrs = [
    [1,5,9,13],
    [3,7,11,15],
    [2,6,10],
    [],
    [0]
]
print(s.merge(arrs))
```

workflow = {procedure1: [], procedure2: [procedure1], procedure3: [procedure1, procedure2]}
time = [2, 2, 1]
其中:
workflow 表示 procedure 的依赖关系. 满足依赖条件的 procedure 可以并行. 找出指定 procedure 的 critical path, 定义为决定总时间的那些procedure.
一开始想用 topological sort 去做, 后来觉得没必要, 直接用递归/回朔.
面试过程中 clarify:


assumption 1: 没有环 (acyclic graph)
assumption 2: ‍‍‌‍‍‌‍‌‌‍‍‍‌‌‍‍‌‌‍如果有两个 procedures 在同一节点耗时相同, 则输出前一个.
follow-up: 输出所有 procedure 的执行顺序 (不考虑时间). 感觉这是强行凑了另一个问题, 然后又回到 topological sort了 [捂脸].


```python
class Solution:
    def minimumTime(self, n: int, relations: List[List[int]], time: List[int]) -> int:
        indgree = collections.defaultdict(int)
        graph = collections.defaultdict(list)
        for pre, next in relations:
            graph[pre-1].append(next-1)
            indgree[next-1] += 1
        q = collections.deque([ i for i in range(n) if indgree[i] == 0]) # candidate without indgree
        start = [0]*n  # earliest start time for each task
        finish = -1
        while q: # process the task, update all children start time and indgree
            cur_task = q.popleft()
            cur_end = start[cur_task] + time[cur_task]
            if cur_end > finish:
                finish = cur_end
            for child in graph[cur_task]:
                indgree[child] -= 1
                if start[child] < cur_end:
                    start[child] = cur_end
                if indgree[child] == 0:
                    q.append(child)
        return finish

print([[0]*5])
```
