```python

Next Greater number

Given an array, find the next greater element on its right. If not, return -1. 

input
nums2 = [1,3,4,2]
output
nums2 = [3,4,-1,-1]


# method 1 scanning from back to front
class Solution:
    def nextGreaterElement(self, nums1: List[int], nums2: List[int]) -> List[int]:
        
        
        stack = []
        
        res = [-1] * len(nums2)
        for i in range(len(nums2) - 1, -1, -1):
            
            while stack and stack[-1] <= nums2[i]: ## get all the previous smaller or equal elements out 
                stack.pop()
            
            res[i] = -1 if not stack else stack[-1]  ## the next greater element will be on the stack top
            
            stack.append(nums2[i])
            

        return res
    
    
# method II scanning from from front to back

class Solution:
    def nextGreaterElement(self, nums1: List[int], nums2: List[int]) -> List[int]:
        
        
        stack = []
        
        res = {}
        
        for i in range(len(nums2)):
            
            while stack and stack[-1] < nums2[i]:
                res[stack.pop()] = nums2[i]   # for each pop out element, we just find the next greater element
            stack.append(nums2[i])
            
        
        resAll = []
        
        
        for e in nums2:
            resAll.append(res.get(e, -1))   

        return resAll






503. Next Greater Element II

Given a circular array (the next element of the last element is the first element of the array), print 
he Next Greater Number for every element. The Next Greater Number of a number x is the first greater 
number to its traversing-order next in the array, which means you could search circularly to find its 
next greater number. If it doesn't exist, output -1 for this number.
    
class Solution:
    def nextGreaterElements(self, nums: List[int]) -> List[int]:
        
        res = {}
        
        stack = []
        
        for i in range(2 * len(nums) - 1, - 1, -1):
            
            while stack and stack[-1] <= nums[i % len(nums)]:
                stack.pop()
                
            res[i % len(nums)] = -1 if not stack else stack[-1]
            stack.append(nums[i % len(nums)])
            
        ans = []
        
        for i in range(len(nums)):
            ans.append(res.get(i, -1))
            
        return ans

# from front to back
class Solution:
    def nextGreaterElements(self, nums: List[int]) -> List[int]:
        
        res = {}
        
        stack = []
        
        for i in range(2 * len(nums)):
            
            while stack and stack[-1][0] < nums[i % len(nums)]:
                res[stack.pop()[1]] = nums[i % len(nums)]
                
            stack.append((nums[i % len(nums)], i % len(nums)) )
            
        ans = []
        
        for i in range(len(nums)):
            ans.append(res.get(i, -1))
            
        return ans
```

```python
739. Daily Temperatures
Given a list of daily temperatures T, return a list such that, for each day in the input, tells you how many days you would have to wait until a warmer temperature. If there is no future day for which this is possible, put 0 instead.

For example, given the list of temperatures T = [73, 74, 75, 71, 69, 72, 76, 73], your output should be [1, 1, 4, 2, 1, 1, 0, 0].

Note: The length of temperatures will be in the range [1, 30000]. Each temperature will be an integer in the range [30, 100].
class Solution:
    def dailyTemperatures(self, T: List[int]) -> List[int]:
        
        stack = []
        
        outputDict = {}
        output = [0] * len(T) 
        
        
        for i in range(len(T)):
            
            while stack and stack[-1][0] < T[i]:
                idx = stack[-1][1]
                output[stack.pop()[1]] = i - idx
                
            stack.append((T[i], i))
        
        return output
```

```python
84. Largest Rectangle in Histogram
Given n non-negative integers representing the histogram's bar height where the width of each bar 
is 1, find the area of largest rectangle in the histogram.


class Solution:
    def largestRectangleArea(self, heights: List[int]) -> int:
        
        left = [-1] * len(heights)
        right = [-1] * len(heights)
        
        sLeft = []
        sRight = []
        
        
        for i in range(len(heights)):
            
            while sRight and sRight[-1][0] > heights[i]:
                right[sRight[-1][1]] = i - sRight[-1][1] - 1
                sRight.pop()
                
            sRight.append((heights[i], i))
            
            
            while sLeft and sLeft[-1][0] >= heights[i]:
                sLeft.pop()
            left[i] = -1 if not sLeft else i - sLeft[-1][1] - 1
            
            sLeft.append((heights[i], i))
        
        for i in range(len(heights)):
            if left[i] == -1:
                left[i] = i
            if right[i] == -1:
                right[i] = len(heights) - i - 1
            
        maxRes = 0
        
        for i in range(len(heights)):
            
            maxRes = max(maxRes, heights[i] * (right[i] + left[i] + 1))
            
        return maxRes
    
    
def largestRectangleArea(self, height):
    height.append(0)
    stack = [-1]
    ans = 0
    for i in xrange(len(height)):
        while height[i] < height[stack[-1]]:
            h = height[stack.pop()]
            w = i - stack[-1] - 1
            ans = max(ans, h * w)
        stack.append(i)
    height.pop()
    return ans
```
