## Basic facts

Time Complexity: a generic hash map insert/lookup takes O(N) in the worst case. For hashmaps storing numbers, we can use a hash map that is based on a Binary Search Tree, and reduces inserts and lookups to O(log(N)). For the average case these operations take O(1).

### Lintcode128. Hash Function

```python
128. Hash Function
中文English
In data structure Hash, hash function is used to convert a string(or any other type) into an integer smaller than hash size and bigger or equal to zero. The objective of designing a hash function is to "hash" the key as unreasonable as possible. A good hash function can avoid collision as less as possible. A widely used hash function algorithm is using a magic number 33, consider any string as a 33 based big integer like follow:

hashcode("abcd") = (ascii(a) * 333 + ascii(b) * 332 + ascii(c) *33 + ascii(d)) % HASH_SIZE 

                              = (97* 333 + 98 * 332 + 99 * 33 +100) % HASH_SIZE

                              = 3595978 % HASH_SIZE

here HASH_SIZE is the capacity of the hash table (you can assume a hash table is like an array with index 0 ~ HASH_SIZE-1).

Given a string as a key and the size of hash table, return the hash value of this key.


use some properties like
(a * b ) % MOD = ((a % MOD) * (b % MOD)) % MOD
(a + b) % MOD = ((a % MOD) + (b % MOD)) % MOD
basically means that if we will take mod finally, we can take it at early steps.

class Solution:
    """
    @param key: A string you should hash
    @param HASH_SIZE: An integer
    @return: An integer
    """
    def hashCode(self, key, HASH_SIZE):
        # write your code here


        code = 0
        
        for c in key:
            
            code *= 33 
            code += ord(c)
            code %= HASH_SIZE # important, for long key, this can simplify the computation of power
            
        return code % HASH_SIZE
```

### 705 Design HashSet 
Design a HashSet without using any built-in hash table libraries.

To be specific, your design should include these functions:

add(value): Insert a value into the HashSet. 
contains(value) : Return whether the value exists in the HashSet or not.
remove(value): Remove a value in the HashSet. If the value does not exist in the HashSet, do nothing.


```python
Here I use list as the linkedList
add, contains, remove: O(N/K):  N is the number of all possible values and K is the number buckets
N/K is average number of linkedList length
class MyHashSet:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.capacity = 1024
        self.data = [[] for _ in range(self.capacity)]
        
        

    def add(self, key: int) -> None:
        if not self.contains(key):
            self.data[key % self.capacity].append(key)
        

    def remove(self, key: int) -> None:
        
        if self.contains(key):
            self.data[key % self.capacity].remove(key)
        
        

    def contains(self, key: int) -> bool:
        """
        Returns true if this set contains the specified element
        """
        idx = key % self.capacity   
        
        if key in self.data[idx]:
            return True
        else:
            return False
        
        

# Your MyHashSet object will be instantiated and called as such:
# obj = MyHashSet()
# obj.add(key)
# obj.remove(key)
# param_3 = obj.contains(key)
```

### 706. Design HashMap (using custom linkedList)

```python
706. Design HashMap

Design a HashMap without using any built-in hash table libraries.

To be specific, your design should include these functions:

put(key, value) : Insert a (key, value) pair into the HashMap. If the value already exists in the HashMap, update the value.
get(key): Returns the value to which the specified key is mapped, or -1 if this map contains no mapping for the key.
remove(key) : Remove the mapping for the value key if this map contains the mapping for the key.
    
class ListNode:
    
    def __init__(self, key, value, next=None):
        self.key = key
        self.val = value
        self.next = next

class MyHashMap:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.capacity = 10000
        self.data = [ListNode(-1, -1) for _ in range(self.capacity)]        

    def put(self, key: int, value: int) -> None:
        """
        value will always be non-negative.
        """
        _, node = self._find(key)
        if node:
            node.value = value
        else:
            address = key % self.capacity
            dummy = self.data[address]
            newNode = ListNode(key, value)
            newNode.next = dummy.next
            dummy.next = newNode
        
    def _find(self, key):
        address = key % self.capacity
        curr = self.data[address].next
        prev = self.data[address]
        while curr:
            if curr.key == key:
                return prev, curr
            prev = curr
            curr = curr.next
        return prev, curr
        
    def get(self, key: int) -> int:
        """
        Returns the value to which the specified key is mapped, or -1 if this map contains no mapping for the key
        """
        _, node = self._find(key)
        if node:
            return node.value
        else:
            return -1

    def remove(self, key: int) -> None:
        """
        Removes the mapping of the specified value key if this map contains a mapping for the key
        """
        prev, node = self._find(key)
        if node:
            prev.next = prev.next.next

            

# Your MyHashMap object will be instantiated and called as such:
# obj = MyHashMap()
# obj.put(key,value)
# param_2 = obj.get(key)
# obj.remove(key)    
```

### 129. Rehashing

```python
129. Rehashing
中文English
The size of the hash table is not determinate at the very beginning. 
If the total size of keys is too large (e.g. size >= capacity / 10),
we should double the size of the hash table and rehash every keys. Say you have a hash table looks like below:
    
    
"""
Definition of ListNode
class ListNode(object):

    def __init__(self, val, next=None):
        self.val = val
        self.next = next
"""
class Solution:
    """
    @param hashTable: A list of The first node of linked list
    @return: A list of The first node of linked list which have twice size
    """
    def rehashing(self, hashTable):
        # write your code here
        capacity = len(hashTable) * 2
        
        newTable = [None] * capacity
        
        for head in hashTable:
            while head:
                temp = head.next
                newKey = head.val % capacity
                newHead = newTable[newKey]
                if not newHead:
                    newTable[newKey] = head
                    head.next = None
                else:
                    while newHead.next:# add to the listEnd (depending on the requirement) Add to the head will be more efficient
                        newHead = newHead.next
                    newHead.next = head
                    head.next = None
                head = temp
                
        return newTable
                
            
```

### 535. Encode and Decode TinyURL

Note: This is a companion problem to the System Design problem: Design TinyURL.
TinyURL is a URL shortening service where you enter a URL such as https://leetcode.com/problems/design-tinyurl and it returns a short URL such as http://tinyurl.com/4e9iAk.

Design the encode and decode me

```python
class Codec:
    def __init__(self):
        self.letters = '0123456789abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
        self.count = 0
        self.map = {}
        self.map2 = {}
    def toString(self, count):
        
        res = []
        while count:
            res.append(self.letters[count % 62])
            count //= 62
        return ''.join(res)
        
    def encode(self, longUrl: str) -> str:
        """Encodes a URL to a shortened URL.
        """
        if longUrl in self.map2:
            return "http://tinyurl.com/" + self.map2[longUrl]
        
        self.count += 1
        short = self.toString(self.count)
        self.map[short] = longUrl
        self.map2[longUrl] = short
        return "http://tinyurl.com/" + short

    def decode(self, shortUrl: str) -> str:
        """Decodes a shortened URL to its original URL.
        """
        return self.map[shortUrl.replace('http://tinyurl.com/', "")]
```

### 460. LFU
Design and implement a data structure for Least Frequently Used (LFU) cache. It should support the following operations: get and put.

get(key) - Get the value (will always be positive) of the key if the key exists in the cache, otherwise return -1.
put(key, value) - Set or insert the value if the key is not already present. When the cache reaches its capacity, it should invalidate the least frequently used item before inserting a new item. For the purpose of this problem, when there is a tie (i.e., two or more keys that have the same frequency), the least recently used key would be evicted.

Note that the number of times an item is used is the number of calls to the get and put functions for that item since it was inserted. This number is set to zero when the item is removed.

```python
//minFreq is the smallest frequency so far
//The main idea is to put all keys with the same frequency to a linked list so the most recent one can be evicted;
//mIter stored the key's position in the linked list;
typedef pair<int, int> pii;
class LFUCache {
    int cap;
    int size;
    int minFreq = 0;
    unordered_map<int, pii> m;//key to {value,freq};
    unordered_map<int, list<int>::iterator> mIter; //key to list iterator;
    unordered_map<int, list<int>> freqM; //freq to key list;
public:
    LFUCache(int capacity) {
        size = 0;
        cap = capacity;
    }
    
    int get(int key) {
        if (m.count(key) == 0) return -1;
        
        freqM[m[key].second].erase(mIter[key]);
        m[key].second++;
        int newFreq = m[key].second;
        freqM[newFreq].push_back(key);
        mIter[key] = prev(end(freqM[newFreq]));
        
        if (freqM[minFreq].size() == 0)
            minFreq++;
        return m[key].first;
    }
    
    void put(int key, int value) {
        if (cap <= 0)
            return;
        int storeValue = get(key);
        if (storeValue != -1) {
            m[key].first = value;
            return;
        }
        
        if (size >= cap) {
            m.erase(freqM[minFreq].front());
            mIter.erase(freqM[minFreq].front()); // front item is least recently used
            freqM[minFreq].pop_front();
            size--;
        }
        
        m[key] = {value, 1};
        freqM[1].push_back(key);
        mIter[key] = prev(freqM[1].end());
        minFreq = 1;
        size++;
    }
};

/**
 * Your LFUCache object will be instantiated and called as such:
 * LFUCache* obj = new LFUCache(capacity);
 * int param_1 = obj->get(key);
 * obj->put(key,value);
 */
```

### 146. LRU Cache
Design and implement a data structure for Least Recently Used (LRU) cache. It should support the following operations: get and put.

get(key) - Get the value (will always be positive) of the key if the key exists in the cache, otherwise return -1.
put(key, value) - Set or insert the value if the key is not already present. When the cache reached its capacity, it should invalidate the least recently used item before inserting a new item.

The cache is initialized with a positive capacity.

```python
# doubly linked list solution
class DListNode:
    def __init__(self, key, val):
        self.key = key
        self.val = val
        self.prev, self.next = None, None
        
class LRUCache:

    def __init__(self, capacity: int):
        
        self.capacity = capacity
        self.head, self.tail = DListNode(-1, -1), DListNode(-1, -1)
        self.head.next, self.tail.prev = self.tail, self.head
        self.keyMap = {}
        
    def _add_node(self, node):
        
        
        node.prev = self.head
        node.next = self.head.next
        
        self.head.next.prev, self.head.next = node, node
        
    def _remove_node(self, node):
        
        node.prev.next, node.next.prev = node.next, node.prev
        
    def _move_to_front(self, node):
        
        self._remove_node(node)
        self._add_node(node)
        
        
    def get(self, key: int) -> int:
        
        if key not in self.keyMap:
            return -1
        
        node = self.keyMap[key]
        
        self._move_to_front(node)
        
        return node.val
        
        
    def put(self, key: int, value: int) -> None:
        
        if key not in self.keyMap:
            node = DListNode(key, value)

            self.keyMap[key] = node
            self._add_node(node)

            if len(self.keyMap) > self.capacity:

                tailNode = self.tail.prev
                del self.keyMap[tailNode.key]
                self._remove_node(tailNode)
        else:
            self.keyMap[key].val = value
            self._move_to_front(self.keyMap[key])


# single linked list solution
class ListNode:
    def __init__(self, key, value):
        self.next = None
        self.key, self.val = key, value


class LRUCache:

    def __init__(self, capacity: int):
        
        self.key_to_prev = {}
        self.dummy = ListNode(-1, -1)
        self.tail = self.dummy
        self.capacity = capacity
        
    def _push_back(self, key, value):
        
        self.tail.next = ListNode(key, value)
        self.key_to_prev[key] = self.tail
        self.tail = self.tail.next

    def _pop_front(self):
        
        del self.key_to_prev[self.dummy.next.key]
        self.dummy.next = self.dummy.next.next
        if not self.dummy.next:
            self.tail = self.dummy
        if self.dummy.next:    
            self.key_to_prev[self.dummy.next.key] = self.dummy
           
    def _kick(self, prev):
        
        node = prev.next
        if node == self.tail:
            return
        
        prev.next = node.next
        self.key_to_prev[node.next.key] = prev
        node.next = None
        self.key_to_prev[node.key] = self.tail
        self.tail.next = node
        self.tail = self.tail.next
        
    
    def get(self, key: int) -> int:
        
        if key not in self.key_to_prev:
            return -1
        
        node = self.key_to_prev[key].next
        
        self._kick(self.key_to_prev[key])
        
        return node.val
        
    def put(self, key: int, value: int) -> None:
        
        if key in self.key_to_prev:
            self.key_to_prev[key].next.val = value
            self._kick(self.key_to_prev[key])
        else:
            self._push_back(key, value)
            if len(self.key_to_prev) > self.capacity:
                self._pop_front()
                
        
        


# Your LRUCache object will be instantiated and called as such:
# obj = LRUCache(capacity)
# param_1 = obj.get(key)
# obj.put(key,value)

# C++ implementation
class LRUCache {
public:
    LRUCache(int cap) {
        capacity = cap;
        size = 0;
    }
    
    int get(int key) {
        if (m.find(key) != m.end()) {
            int val = m[key];
            keyList.erase(mIter[key]);
            keyList.push_back(key);
            mIter[key] = prev(end(keyList));
            return val;
        } else {
            return -1;
        }
    }
    
    void put(int key, int value) {
        
        int stored = get(key);
        if (stored != -1) {
            m[key] = value;
            return;
        } 
        
        m[key] = value;
        keyList.push_back(key);
        mIter[key] = prev(end(keyList));
        size++;
        
        if (size > capacity) {
            
            int removeKey = keyList.front();
            keyList.pop_front();
            mIter.erase(removeKey);
            m.erase(removeKey);
            size--;
        }
    }
    
    int capacity;
    int size;
    list<int> keyList;
    unordered_map<int, int> m;
    unordered_map<int, list<int>::iterator> mIter;
    
};

/**
 * Your LRUCache object will be instantiated and called as such:
 * LRUCache* obj = new LRUCache(capacity);
 * int param_1 = obj->get(key);
 * obj->put(key,value);
 */
```

```python
Given an array of integers and an integer k, you need to find the number of unique k-diff pairs in the array. Here a k-diff pair is defined as an integer pair (i, j),
where i and j are both numbers in the array and their absolute difference is k.

class Solution:
    def findPairs(self, nums: List[int], k: int) -> int:
        
        if k < 0:
            return 0
        
        count = collections.Counter(nums)
        ans = 0
        # note that we need to iterate over dictionary key instaed of nums
        for e in count:
            if k == 0:
                ans += 1 if count[e] > 1 else 0
            else:
                # The pairs (i, j) and (j, i) count as the same pair.
                # for each e, we search for e + k > e to avoid duplicates
                ans += 1 if e + k in count else 0
        
        return ans
```

### 957. Prison Cells After N Days

```python
class Solution:
    """
    @param ages: 
    @return: nothing
    """
    def numFriendRequests(self, ages):
        # 
        ageCount = collections.Counter(ages)
        ans = 0
        for i in ageCount:
            for j in ageCount:
                if self.satisfy(i, j):
                    if i == j:
                        ans += ageCount[i] * (ageCount[i] - 1)
                    else:
                        ans += ageCount[j] * ageCount[i]
        return ans
    def satisfy(self, a, b):
        return not (b <= 0.5 * a + 7 or b > a or b > 100 and a < 100)
```

### 554. Brick Wall

There is a brick wall in front of you. The wall is rectangular and has several rows of bricks. The bricks have the same height but different width. You want to draw a vertical line from the top to the bottom and cross the least bricks.

The brick wall is represented by a list of rows. Each row is a list of integers representing the width of each brick in this row from left to right.

If your line go through the edge of a brick, then the brick is not considered as crossed. You need to find out how to draw the line to cross the least bricks and return the number of crossed bricks.

You cannot draw a line just along one of the two vertical edges of the wall, in which case the line will obviously cross no bricks.

```python
class Solution:
    def leastBricks(self, wall: List[List[int]]) -> int:
        width = sum(wall[0])
        m = {}
        for row in wall:
            start = 0
            for b in row:
                m[start] = m.get(start, 0) + 1
                start += b
        ans = len(wall)
        for k in m:
            if 0 < k < width:
                ans = min(ans, len(wall) - m[k])
        return ans
```
