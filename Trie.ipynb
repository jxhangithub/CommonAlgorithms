### Trie implementation (array based)

```python
https://leetcode.com/problems/implement-trie-prefix-tree/

Implementation version one (faster than 23%)
    
class TrieNode:
    
    def __init__(self):
        
        self.children = [None] * 26
        self.isEndOfWord = False
        
class Trie:
    
    def __init__(self):
        
        self.root = TrieNode()
        
    def insert(self, key):
        
        curr = self.root
        for i in range(len(key)):
            idx = ord(key[i]) - ord('a')
            if not curr.children[idx]:
                curr.children[idx] = TrieNode()
                curr = curr.children[idx]
            else:
                curr = curr.children[idx]
        
        curr.isEndOfWord = True
        
    def startsWith(self, prefix):
        
        curr = self.root
        for c in prefix:
            idx = ord(c) - ord('a')
            if curr.children[idx]:
                curr = curr.children[idx]
            else:
                return False
        return True
        
    def search(self, key):
        curr = self.root
        for c in key:
            idx = ord(c) - ord('a')
            if curr.children[idx]:
                curr = curr.children[idx]
            else:
                return False
            
            
        return curr and curr.isEndOfWord
    
    
    

```

### Trie implementation (DefaultDict based)

```python
# implementation version two, faster than 56%)
class TrieNode:
    def __init__(self):
        self.childrens = collections.defaultdict(TrieNode)
        self.is_word = False



class Trie:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.root = TrieNode()

    def insert(self, word: str) -> None:
        """
        Inserts a word into the trie.
        """
        curr = self.root
        for l in word:
            curr = curr.childrens[l]
            
        curr.is_word = True
        

    def search(self, word: str) -> bool:
        """
        Returns if the word is in the trie.
        """
        
        curr = self.root
        
        for l in word:
            if l in curr.childrens: # note that we cannot use if curr.children[l], this will creates a key-value pair
                curr = curr.childrens[l]
            else:
                return False
        return curr.is_word

    def startsWith(self, prefix: str) -> bool:
        """
        Returns if there is any word in the trie that starts with the given prefix.
        """
        curr = self.root
        
        for l in prefix:
            if l in curr.childrens: # note that we cannot use if curr.children[l], this will creates a key-value pair
                curr = curr.childrens[l]
            else:
                return False
        return True
```

### 1065. Index Pairs of a String
Given a text string and words (a list of strings), return all index pairs [i, j] so that the substring text[i]...text[j] is in the list of words.

```python
class Solution:
    def indexPairs(self, text: str, words: List[str]) -> List[List[int]]:
        
        
        root = {}
        
        for word in words:
            curr = root
            for c in word:
                if c not in curr:
                    curr[c] = {}
                curr = curr[c]
            curr['#'] = True
        
        result = []
        for i, c in enumerate(text):
            j = i
            curr = root
            while j < len(text):
                c = text[j]
                if c not in curr:
                    break
                else:
                    curr = curr[c]
                    if '#' in curr:
                        result.append([i, j])
                j += 1
                
        return result
```

```python

# standard implementation
class TrieNode:
    def __init__(self):
        self.children = {}
        self.isWord = False


class WordDictionary:

    def __init__(self):
        
        self.root = TrieNode()

    def addWord(self, word):
        # write your code here
        
        curr = self.root
        
        for c in word:
            if c not in curr.children:
                curr.children[c] = TrieNode() 
            curr = curr.children[c]
        curr.isWord = True


    def search(self, word):
        
        stack = [(self.root, 0)]
        
        while stack:
            
            node, idx = stack.pop()
            
            if idx == len(word):
                if node.isWord:
                    return True
            else:
                for c in node.children:
                    if word[idx] == '.' or word[idx] == c:
                        stack.append((node.children[c], idx + 1))
                        
        return False
            


# alternative
class WordDictionary:
    def __init__(self):
        """
        Initialize your data structure here.
        """
        
        self.root = {}
        
    def addWord(self, word: str) -> None:
        """
        Adds a word into the data structure.
        """
        curr = self.root
        
        for w in word:
            if w not in curr:
                curr[w] = {}
            curr = curr[w]
            
        curr["#"] = True
            
    
    def search(self, word):
        # write your code here
        
        def help(word, node, idx):
            
            if idx == len(word):
                if '#' in node:
                    return True
                else:
                    return False
            
            c = word[idx]
  
            if c == '.':
                for child in node:
                    if child != '#' and help(word, node[child], idx + 1):
                        return True
            else:
                if c not in node:
                    return False
            
                if c in node:
                    return help(word, node[c], idx + 1)

            return False
        
        
        return help(word, self.root, 0)
```

### 212. Word Search II

```python
212. Word Search II

Given a 2D board and a list of words from the dictionary, find all words in the board.

Each word must be constructed from letters of sequentially adjacent cell, where "adjacent" cells are those horizontally
or vertically neighboring. The same letter cell may not be used more than once in a word.

class TrieNode:
    def __init__(self):
        self.children = {}
        self.isWord = False
        
        
class Solution:
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        root = TrieNode()
        for word in words:
            curr = root
            for c in word:
                if c not in curr.children:
                    curr.children[c] = TrieNode()
                curr = curr.children[c]
            curr.isWord = True
                
        
        def dfs(i, j, board, temp, node):
            
            c = board[i][j]
            if c in node.children:
                temp.append(c)
                board[i][j] = '#'
                node = node.children[c]
            
                if node.isWord:
                    res.append(''.join(temp))
                    node.isWord = False # set to false to avoid duplicates
                    
                for dx, dy in (-1, 0), (1, 0), (0, -1), (0, 1):
                    newX = i + dx
                    newY = j + dy

                    if 0 <= newX < H and 0 <= newY < W and board[newX][newY] != '#':
                        dfs(newX, newY, board, temp, node)
                board[i][j] = temp[-1]
                temp.pop()
                        
        res = []
        H = len(board)
        W = len(board[0])
        for i in range(H):
            for j in range(W):
                dfs(i, j, board, [], root)
        return res        
```

### 677. Map Sum Pairs

```python
Implement a MapSum class with insert, and sum methods.

For the method insert, you'll be given a pair of (string, integer). 
The string represents the key and the integer represents the value. If the key already existed, then the original key-value pair will be overridden to the new one.

For the method sum, you'll be given a string representing the prefix,
and you need to return the sum of all the pairs' value whose key starts with the prefix.

# dfs search approach, consume less memory
class MapSum:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.root = {}

    def insert(self, key: str, val: int) -> None:
        
        curr = self.root
        for c in key:
            if c not in curr:
                curr[c] = {}
            curr = curr[c]
            
        curr['#'] = val
        
    def sum(self, prefix: str) -> int:
        
        res= 0

        stack = [(self.root, 0)]
        
        while stack:
            
            node, idx = stack.pop()
            
            if '#' in node and idx >= len(prefix):
                res += node['#']
            
            if idx < len(prefix):
                c = prefix[idx]
                if c in node:
                    stack.append((node[c], idx + 1))
                else:
                    break
            else:
                
                for child in node:
                    if child == '#':
                        continue
                    else:
                        stack.append((node[child], idx + 1))
        
        return res
                
# using hashmap to speed up and make clean code
class MapSum:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.root = {}
        self.exist = {}

    def insert(self, key: str, val: int) -> None:
        
        delta = val - self.exist.get(key, 0)
        self.exist[key] = val
        
        curr = self.root
        # root node has the total count
        curr['score'] = curr.get('score', 0) + delta
        for c in key:
            if c not in curr:
                curr[c] = {}
            curr = curr[c]
            curr['score'] = curr.get('score', 0) + delta 
         
    def sum(self, prefix: str) -> int:
        curr = self.root
        for c in prefix:
            if c not in curr:
                return 0
            else:
                curr = curr[c]
                
        return curr['score']
            
            
        

# Your MapSum object will be instantiated and called as such:
# obj = MapSum()
# obj.insert(key,val)
# param_2 = obj.sum(prefix)
```

### 648. Replace Words

```python
In English, we have a concept called root, which can be followed by some other words to form another longer word - let's call this word successor. For example, the root an, followed by other, which can form another word another.

Now, given a dictionary consisting of many roots and a sentence. You need to replace all the successor in the sentence with the root forming it. If a successor has many roots can form it, replace it with the root with the shortest length.

You need to output the sentence after the replacement.

class Solution:
    def replaceWords(self, dict: List[str], sentence: str) -> str:
        
        self.root = {}
        
        for w in dict:
            curr = self.root
            for c in w:
                if c not in curr:
                    curr[c] = {}
                curr = curr[c]
                
            curr['#'] = w
        
        def replace(w):
            curr = self.root
            for c in w:
                if c not in curr:
                    break
                else:
                    curr = curr[c]
                    if '#' in curr:
                        return curr['#']
            return w            
        
        res = []
        sentence = sentence.split()
        for w in sentence:
            res.append(replace(w))

        return ' '.join(res)
```

### 745. Prefix and Suffix Search (vanilla prefix tree method)

Given many words, words[i] has weight i.

Design a class WordFilter that supports one function, WordFilter.f(String prefix, String suffix). It will return the word with given prefix and suffix with maximum weight. If no word exists, return -1.

Examples:

Input:
WordFilter(["apple"])
WordFilter.f("a", "e") // returns 0
WordFilter.f("b", "") // returns -1

```python

# vanilla solution
class WordFilter:

    def __init__(self, words: List[str]):
        
        self.root = {}
        for i, w in enumerate(words):
            self._add(w, i)
    
    def _add(self, word, i):
        curr = self.root
        for c in word:
            if c not in curr:
                curr[c] = {}
            curr = curr[c]
        curr['#'] = i
        

    def f(self, prefix: str, suffix: str) -> int:
        
        
        self.ans = -1
        def help(node, idx, wordSoFar):
            
            if idx >= len(prefix):
                if '#' in node:
                    word = ''.join(wordSoFar)
                    suffixLen = len(suffix)
                    
                    if suffixLen == 0:
                        self.ans = max(self.ans, node['#'])
                    elif len(word) >= suffixLen and word[-suffixLen:] == suffix:
                        self.ans = max(self.ans, node['#'])
            if idx < len(prefix):
                c = prefix[idx]
                if c in node:
                    wordSoFar.append(c)
                    help(node[c], idx + 1, wordSoFar)
                    wordSoFar.pop()
            else:
                for c in node:
                    if c == '#':
                        continue
                    wordSoFar.append(c)
                    help(node[c], idx + 1, wordSoFar)
                    wordSoFar.pop()
        
        
        
        help(self.root, 0, [])
        
        return self.ans
        
# Your WordFilter object will be instantiated and called as such:
# obj = WordFilter(words)
# param_1 = obj.f(prefix,suffix)
```

### 720. Longest Word in Dictionary
Given a list of strings words representing an English Dictionary, find the longest word in words that can be built one character at a time by other words in words. If there is more than one possible answer, return the longest word with the smallest lexicographical order.

If there is no answer, return the empty string.

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.isWord = False

class Solution:
    def longestWord(self, words: List[str]) -> str:
        
        root = TrieNode()
        
        for word in words:
            curr = root
            for c in word:
                if c not in curr.children:
                    curr.children[c] = TrieNode()
                curr = curr.children[c]
            curr.isWord = True
            
        
        stack = [(root.children[c], [c]) for c in root.children if root.children[c].isWord]
        
        maxDepth = 0
        answer = ''
        while stack:
            
            node, temp = stack.pop()

            if len(temp) > maxDepth:
                maxDepth = len(temp)
                answer = ''.join(temp)
            elif len(temp) == maxDepth:
                tempAnswer = ''.join(temp)
                if tempAnswer < answer:
                    answer = tempAnswer

            for c in node.children:
                if node.children[c].isWord:
                    stack.append((node.children[c], temp+[c]))
        return answer
            
```

### 676. Implement Magic Dictionary

Implement a magic directory with buildDict, and search methods.

For the method buildDict, you'll be given a list of non-repetitive words to build a dictionary.

For the method search, you'll be given a word, and judge whether if you modify exactly one character into another character in this word, the modified word is in the dictionary you just built.






```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.isWord = False

class MagicDictionary:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.root = TrieNode()

    def buildDict(self, dict: List[str]) -> None:
        """
        Build a dictionary through a list of words
        """
        for word in dict:
            curr = self.root
            for c in word:
                if c not in curr.children:
                    curr.children[c] = TrieNode()
                curr = curr.children[c]
            curr.isWord = True
        

    def search(self, word: str) -> bool:
        """
        Returns if there is any word in the trie that equals to the given word after modifying exactly one character
        """
        if not word:
            return False
        
        curr = self.root
        stack = [(curr, 0, 0)]
        
        while stack:
            node, idx, diffCount = stack.pop()
            
            if idx == len(word):
                if node.isWord and diffCount == 1:
                    return True
            else:
                for c in node.children:
                    tempCount = diffCount
                    if word[idx] != c:
                        tempCount += 1
                    if tempCount > 1:
                        continue
                    stack.append((node.children[c], idx + 1, tempCount))
                           
        return False
    
    # alternative implementation of search
    def search(self, word: str) -> bool:
        """
        Returns if there is any word in the trie that equals to the given word after modifying exactly one character
        """
        if not word:
            return False
        
        curr = self.root
        stack = [(curr.children[c], c, 0, 0) for c in curr.children]
        
        while stack:
            node, ch, idx, diffCount = stack.pop()
            
            if idx == len(word):
                continue
            
            if word[idx] != ch:
                diffCount += 1
                if diffCount > 1:
                    continue
            
            if idx == len(word) - 1 and node.isWord and diffCount == 1:
                return True
            
            for c in node.children:
                stack.append((node.children[c], c, idx + 1, diffCount))

        return False    
```

### 425. Word Squares
Given a set of words (without duplicates), find all word squares you can build from them.

A sequence of words forms a valid word square if the kth row and column read the exact same string, where 0 ≤ k < max(numRows, numColumns).

For example, the word sequence ["ball","area","lead","lady"] forms a word square because each word reads the same both horizontally and vertically.


```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.prefixIdx = []

class Solution:
    def wordSquares(self, words: List[str]) -> List[List[str]]:
        
        root = TrieNode()
        
        for idx, word in enumerate(words):
            curr = root
            curr.prefixIdx.append(idx)
            for c in word:
                if c not in curr.children:
                    curr.children[c] = TrieNode()
                curr = curr.children[c]
                curr.prefixIdx.append(idx)
        
        def help(temp, k):
            
            if len(temp) == k:
                result.append([e for e in temp])
            else:
                Len = len(temp)
                prefix = ''
                for j in range(Len):
                    prefix += temp[j][Len]
                
                curr = root
                prefixList = curr.prefixIdx
                for c in prefix:
                    if c in curr.children:
                        curr = curr.children[c]
                        prefixList = curr.prefixIdx
                    else:
                        prefixList = []
                        break
 
                for idx in prefixList:
                    temp.append(words[idx])
                    help(temp, k)
                    temp.pop()

        result = []
        visited = [False] * len(words)
        k = len(words[0])
        help([], k)
        
        return result
        
        
```

### 421. Maximum XOR of Two Numbers in an Array
Given a non-empty array of numbers, a0, a1, a2, … , an-1, where 0 ≤ ai < 231.

Find the maximum result of ai XOR aj, where 0 ≤ i, j < n.

Could you do this in O(n) runtime?

```python
class Solution:
    def findMaximumXOR(self, nums: List[int]) -> int:
        
        # get the maximum length of binary reprenetation
        L = len(bin(max(nums))) - 2
        old = nums
        nums = []
        for x in old:
            temp = [(x >> i) & 1 for i in range(L)]
            nums.append(temp[::-1])
            
        
        
        root = {}
        for num in nums:
            curr = root
            for b in num:
                if b not in curr:
                    curr[b] = {}
                curr = curr[b]
        
        max_xor = 0
        for num in nums:
            curr = root
            curr_xor = 0
            for b in num:
                opposite_bit = 1 - b
                if opposite_bit in curr:
                    curr = curr[opposite_bit]
                    curr_xor = (curr_xor << 1) + 1
                else:
                    curr = curr[b]
                    curr_xor <<= 1
            max_xor = max(max_xor, curr_xor)
            
        return max_xor
            
            
        
```

### 820. Short Encoding of Words
Given a list of words, we may encode it by writing a reference string S and a list of indexes A.

For example, if the list of words is ["time", "me", "bell"], we can write it as S = "time#bell#" and indexes = [0, 2, 5].

Then for each index, we will recover the word by reading from the reference string from that index until we reach a "#" character.

What is the length of the shortest reference string S possible that encodes the given words?

```python



# store all suffix in hashset and count those not found in suffix set
class Solution:
    def minimumLengthEncoding(self, words: List[str]) -> int:
        
        subfix = set()
        
        for w in words:
            for i in range(1, len(w)):
                subfix.add(w[i:])
        count = 0    
        seen = set()
        for w in words:
            if w not in seen and w not in subfix:
                count += len(w) + 1
                seen.add(w)
        return count
    
# an optimized solution
class Solution:
    def minimumLengthEncoding(self, words: List[str]) -> int:
        
        good = set(words)
        for word in words:
            for i in range(1, len(word)):
                good.discard(word[i:])
        return sum([len(w) + 1 for w in good])
```
