```python
class UF:
    
    def __init__(self, size):
        
        self.parent = []
        self.rank = []
        
        self.count = size
        
        for i in range(self.size):
            self.parent.append[i]
            self.rank.append[0]
            
            
    # recursive find parent with path compression
    def find1(self, i):
        if self.parent[i] != i:
            self.parent[i] = self.find(self.parent[i])
            
        return self.parent[i]
    
    
    def find2(self, x):
        j = x
        while self.parent[j] != j:
            j = self.parent[j]
        
        # path compression
        while x != j:
            fx = self.parent[x]
            self.parent[x] = j
            x = fx
            
        return j
    
    def find3(self, x):
        path = []
        while self.parent[x] != x:
            path.append(x)
            x = self.parent[x]
        for n in path:
            self.parent[n] = x
    
    
    def union1(self, i, j):
        
        p = self.find(i)
        q = self.find(j)
        
        if p == q: return
        
        if self.rank[p] < self.rank[q]:
            self.parent[p] = q
        elif self.rank[p] > self.rank[q]:
            self.parent[q] = p
        else:
            self.parent[p] = q
            self.rank[q] += 1
            
        self.count -= 1
        
    def union2(self. i, j):
        fx = self.find[i]
        fy = self.find[j]
        
        if fx == fy:
            return
        
        self.parent[fx] = fy
        self.count -= 1 
        
        # if not counting components
    def union3(self, i, j):
        self.parent[self.find(i)] = self.find(j)
```

```python
There are N students in a class. Some of them are friends, while some are not. Their friendship is transitive in nature. For example, if A is a direct friend of B, and B is a direct friend of C, then A is an indirect friend of C. And we defined a friend circle is a group of students who are direct or indirect friends.

Given a N*N matrix M representing the friend relationship between students in the class. If M[i][j] = 1, then the ith and jth students are direct friends with each other, otherwise not. And you have to output the total number of friend circles among all the students.

Example 1:
Input: 
[[1,1,0],
 [1,1,0],
 [0,0,1]]
Output: 2
Explanation:The 0th and 1st students are direct friends, so they are in a friend circle. 
The 2nd student himself is in a friend circle. So return 2.

class Solution:
    def findCircleNum(self, M: List[List[int]]) -> int:
        
        if not M:
            return 0
        
        count = len(M)
        
        
        parent = [i for i in range(len(M))]
        rank = [0] * len(M)
        def find(i):
            
            if parent[i] != i:
                parent[i] = find(parent[i])
                
            return parent[i]
        
        def union(i, j):
            p = find(i)
            q = find(j)
            
            if p == q:
                return
            
            if rank[p] < rank[q]:
                parent[p] = q
            elif rank[p] > rank[q]:
                parent[q] = p
            else:
                parent[p] = q
                rank[q] += 1
                
            nonlocal count
            count -= 1
            
        for i in range(len(M)):
            for j in range(len(M[0])):
                if M[i][j] == 1:
                    union(i, j)
                
                
        return count

```

```python
Given an unsorted array of integers, find the length of the longest consecutive elements sequence.

Your algorithm should run in O(n) complexity.

Example:

Input: [100, 4, 200, 1, 3, 2]
Output: 4
Explanation: The longest consecutive elements sequence is [1, 2, 3, 4]. Therefore its length is 4.


class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        
        if not nums:
            return 0
        
        parent = {}
        rank = {}
        size = {}
        for e in nums:
            parent[e] = e
            rank[e] = 0
            size[e] = 1
            
        def find(e):
            
            if parent[e] != e:
                parent[e] = find(parent[e])
                
            return parent[e]
        
        def union(i, j):
            p = find(i)
            q = find(j)
            
            if p == q:
                return
            
            if rank[p] < rank[q]:
                parent[p] = q
                size[q] += size[p]
            elif rank[q] < rank[p]:
                parent[q] = p
                size[p] += size[q]
            else:
                parent[q] = p
                rank[p] += 1
                size[p] += size[q]
                
        maxSize = 0
        for e in nums:
            if e - 1 in parent:
                union(e - 1, e)
            if e + 1 in parent:
                union(e + 1, e)
            maxSize = max(maxSize, size[find(e)])

        return maxSize
```

### 178. Graph Valid Tree

```python
178. Graph Valid Tree
中文English
Given n nodes labeled from 0 to n - 1 and a list of undirected edges (each edge is a pair of nodes), 
write a function to check whether these edges make up a valid tree.

class Solution:
    """
    @param n: An integer
    @param edges: a list of undirected edges
    @return: true if it's a valid tree, or false
    """
    def validTree(self, n, edges):
        # write your code here
        
        parent = {k:k for k in range(n)}
        self.count = n
        def find(x):
            if parent[x] != x:
                parent[x] = find(parent[x]) 
            return parent[x] 
        def union(x, y):
            p = find(x)
            q = find(y)
            
            if p != q:
                parent[p] = q
                self.count -= 1       
        for e in edges:
            union(e[0], e[1])
            
        return self.count == 1 and len(edges) == n - 1
```

### 839. Similar String Groups

Two strings X and Y are similar if we can swap two letters (in different positions) of X, so that it equals Y. Also two strings X and Y are similar if they are equal.

For example, "tars" and "rats" are similar (swapping at positions 0 and 2), and "rats" and "arts" are similar, but "star" is not similar to "tars", "rats", or "arts".

Together, these form two connected groups by similarity: {"tars", "rats", "arts"} and {"star"}.  Notice that "tars" and "arts" are in the same group even though they are not similar.  Formally, each group is such that a word is in the group if and only if it is similar to at least one other word in the group.

We are given a list A of strings.  Every string in A is an anagram of every other string in A.  How many groups are there?

```python
class Solution:
    def numSimilarGroups(self, A: List[str]) -> int:
        
        def is_similar(a, b):
            return len(list(filter(lambda i: a[i] != b[i], range(len(a))))) in (0, 2)
        
        
        parent = {s: s for s in A}
        def find(s):
            if parent[s] != s:
                parent[s] = find(parent[s])
            return parent[s]
        
        nGroups = len(parent)
        
        keys = list(parent.keys())
        for i in range(len(parent)):
            for j in range(i + 1, len(parent)):
                s1, s2 = keys[i], keys[j]
                if is_similar(s1, s2):
                    q = find(s1)
                    p = find(s2)
                    if p != q:
                        nGroups -= 1
                        parent[q] = p
        
        return nGroups

# C++ implemenetation
class Solution {
public:
    int numSimilarGroups(vector<string>& A) {
        
        unordered_map<string, string> parent;
        for (auto s: A){
            parent[s] = s;
        }
        int nGroup = parent.size();
        
        for(int i = 0; i < A.size(); i++ ){
            for (int j = i + 1; j < A.size(); j++){
                if(isSimilar(A[i], A[j])){
                    string p = find(parent, A[i]);
                    string q = find(parent, A[j]);
                    if (p != q) {
                        nGroup -= 1;
                        parent[q] = p;
                    }
                }       
            }
        }
        return nGroup;
        
    }
    string find(unordered_map<string, string>& parent, const string& p){
        if (parent[p] != p) {
            parent[p] = find(parent, parent[p]);
        }
        return parent[p];
    }
    
    bool isSimilar(const string& A, const string& B){
        int count = 0;
        for (int i = 0; i < A.size(); i++){
            if (A[i] != B[i]){
                count += 1;
                if (count > 2){
                    return false;
                }
            }
            
        }
        return true;
        
    }
};
```

### 684. Redundant Connection

```python
In this problem, a tree is an undirected graph that is connected and has no cycles.

The given input is a graph that started as a tree with N nodes (with distinct values 1, 2, ..., N), with one additional edge added. The added edge has two different vertices chosen from 1 to N, and was not an edge that already existed.

The resulting graph is given as a 2D-array of edges. Each element of edges is a pair [u, v] with u < v, that represents an undirected edge connecting nodes u and v.

Return an edge that can be removed so that the resulting graph is a tree of N nodes. If there are multiple answers, return the answer that occurs last in the given 2D-array. The answer edge [u, v] should be in the same format, with u < v.
class Solution:
    def findRedundantConnection(self, edges: List[List[int]]) -> List[int]:
        
        self.parent = {}
        self.count = 0
        
        for e in edges:
            self.parent[e[0]] = e[0]
            self.parent[e[1]] = e[1]
        
        self.count = len(self.parent)
        
        if self.count <= 1:
            return []
        
       
        prevCount = self.count
        for e in edges:
            self._union(e[0], e[1])
            
            if self.count == prevCount:
                return e
            prevCount = self.count
    
    def _find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self._find(self.parent[x])
       
        return self.parent[x]
        
    def _union(self, x, y):
        p = self._find(x)
        q = self._find(y)
        if p != q:
            self.parent[p] = q
            self.count -= 1
```

### 1319. Number of Operations to Make Network Connected
There are n computers numbered from 0 to n-1 connected by ethernet cables connections forming a network where connections[i] = [a, b] represents a connection between computers a and b. Any computer can reach any other computer directly or indirectly through the network.

Given an initial computer network connections. You can extract certain cables between two directly connected computers, and place them between any pair of disconnected computers to make them directly connected. Return the minimum number of times you need to do this in order to make all the computers connected. If it's not possible, return -1. 

```python
class Solution:
    def makeConnected(self, n: int, connections: List[List[int]]) -> int:
        
        parent = [i for i in range(n)]
        numComponents = n
        numRedundant = 0
        for e in connections:
            p = self.find(e[0], parent)
            q = self.find(e[1], parent)
            if p != q:
                parent[p] = q
                numComponents -= 1
            else:
                numRedundant += 1
                
        if numRedundant >= (numComponents - 1):
            return (numComponents - 1)
        else:
            return -1
        

    def find(self, x, parent):
        if x != parent[x]:
            parent[x] = self.find(parent[x], parent)
        return parent[x]
    
    
    
```

### 990. Satisfiability of Equality Equations
Given an array equations of strings that represent relationships between variables, each string equations[i] has length 4 and takes one of two different forms: "a==b" or "a!=b".  Here, a and b are lowercase letters (not necessarily different) that represent one-letter variable names.

Return true if and only if it is possible to assign integers to variable names so as to satisfy all the given equations.

```python
class Solution:
    def equationsPossible(self, equations: List[str]) -> bool:
        
        variables = set()
        equalties = []
        inEqualities = []
        for eq in equations:
            variables.add(eq[0])
            variables.add(eq[-1])
            if eq[1:3] == '==':
                equalties.append([eq[0], eq[-1]])
            else:
                inEqualities.append([eq[0], eq[-1]])
        
        parents = {e:e for e in variables}
    
        for e in equalties:
            p = self.find(e[0], parents)
            q = self.find(e[1], parents)
            parents[p] = q
        for e in inEqualities:
            if self.find(e[0], parents) == self.find(e[1], parents):
                return False
        return True
    
    def find(self, x, parent):
        while x != parent[x]:
            x = parent[x]
        
        label = x
        while x != label:
            temp = parent[x]
            parent[x] = label
            x = temp
        return parent[x]
        
```

```python
629. Minimum Spanning Tree
中文English
Given a list of Connections, which is the Connection class (the city name at both ends of the edge and a cost between them), find edges that can connect all the cities and spend the least amount.
Return the connects if can connect all the cities, otherwise return empty list.

Example

'''
Definition for a Connection
class Connection:

    def __init__(self, city1, city2, cost):
        self.city1, self.city2, self.cost = city1, city2, cost
'''
import heapq
class Node:
    
    def __init__(self, val):
        self.val = val
    
    def __lt__(self, other):
        return (self.val.cost, self.val.city1, self.val.city2) < (other.val.cost, other.val.city1, other.val.city2)

class Solution:
    # @param {Connection[]} connections given a list of connections
    # include two cities and cost
    # @return {Connection[]} a list of connections from results
    def lowestCost(self, connections):
        # Write your code here
        
        def find(x):
            if parent[x] != x:
                parent[x] = find(parent[x])
                
            return parent[x]
        
        def union(x, y):
            p = find(x)
            q = find(y)
            if p != q:
                parent[p] = q
                self.count -= 1
                
        
        h = []
        parent = {}
        for c in connections:
            heapq.heappush(h, Node(c))
            parent[c.city1] = c.city1
            parent[c.city2] = c.city2


        self.count = len(parent)
        res = []
        while h:
            
            node = heapq.heappop(h)
            c = node.val
            
            if find(c.city1) == find(c.city2):
                continue
            
            union(c.city1, c.city2)

            res.append(c)
            
        if self.count == 1:
            return res
        else:
            return []
            
```

### number of connected components (Union Find)

```python
## Union find
class Solution:
    def countComponents(self, n: int, edges: List[List[int]]) -> int:
        
        
        parent = [i for i in range(n)]
        
        self.count = n
        
        def find(i):
            
            if parent[i] != i:
                parent[i] = find(parent[i])
            return parent[i]
        
        def union(i, j):
            
            p = find(i)
            q = find(j)
            if p == q:
                return
            parent[p] = parent[q]
            self.count -= 1
            
        for e in edges:
            union(e[0], e[1])
        
        return self.count
```
