```python
## num of islands

class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        
        if not grid:
            return 0
        
        self.H = len(grid)
        self.W = len(grid[0])
        
        count = 0
        for i in range(self.H):
            for j in range(self.W):
                if grid[i][j] == '1':
                    self.dfs(grid, i, j)
                    count += 1
                    
        return count
    
    # recursive method
    def dfs(self, grid, i, j):
        
        grid[i][j] = '0'
        
        for a, b in (-1, 0), (1, 0), (0, -1), (0, 1):
            newX = i + a
            newY = j + b
            if 0<= newX < self.H and 0 <= newY < self.W and grid[newX][newY] == '1':
                self.dfs(grid, newX, newY)
                
    # iterative method
    def dfs_stack(self, grid, i, j):
        
        stack = [(i, j)]

        while stack:
            
            node = stack.pop()
            grid[node[0]][node[1]] = '0'
            for a, b in (-1, 0), (1, 0), (0, -1), (0, 1):
                newX = node[0] + a
                newY = node[1] + b
                if 0<= newX < self.H and 0 <= newY < self.W and grid[newX][newY] == '1':
                    stack.append((newX, newY))
```

```python
# graph valid tree

## Given n nodes labeled from 0 to n-1 and a list of undirected edges (each edge is a pair of nodes),
## write a function to check whether these edges make up a valid tree.

## Input: n = 5, and edges = [[0,1], [0,2], [0,3], [1,4]]
## Output: true

class Solution:
    def validTree(self, n: int, edges: List[List[int]]) -> bool:
        if len(edges) != n - 1:
            return False
        
        nbs= collections.defaultdict(list)
        
        for a, b in edges:
            nbs[a].append(b)
            nbs[b].append(a)
            
            
        ## naive implementation O(n) extra space to check if all nodes are in a graph
        visited = [False] * n
        
        stack = [0]
        
        while stack:
            node = stack.pop()
            visited[node] = True
            for nb in nbs[node]:
                if not visited[nb] :
                    stack.append(nb)
                
        return all(visited)
    

    # another DFS solution
    def validTree(self, n: int, edges: List[List[int]]) -> bool:
        if len(edges) != n - 1:
            return False
        
        
        # must construct nb list for all nodes
        nbs = {i: [] for i in range(n)}
        
        for a, b in edges:
            nbs[a].append(b)
            nbs[b].append(a)
            
        
        def visit_dfs(node):
                # pop out nb lists by key
            nbList = nbs.pop(node, [])
            for nb in nbList:
                visit_dfs(nb)
                
        # or iteratively
        def visit_dfs_stack(node):
            stack = [node]
            while stack:
                n = stack.pop()
                stack += nbs.pop(n, [])
                
                
        visit(0)
                # check if all keys have been visited.
        return not nbs
```

```python
## DFS for permutation and combination generation

# permutation
class Solution:
    """
    @param: nums: A list of integers.
    @return: A list of permutations.
    """
    def permute(self, nums):
        # write your code here
        results = []
        visited = set()      
        self.dfs(nums, visited, [], results)
        return results

    def dfs(self, nums, visited, permutation, results):
        if len(visited) == len(nums):
            results.append(list(permutation))
            
        for i in range(len(nums)):
            if i not in visited:
                visited.add(i)
                permutation.append(nums[i])
                self.dfs(nums, visited, permutation, results)
                permutation.pop()
                visited.remove(i)
                
    def combine(self, n, k):
        # write your code here
        
        results = []
        self.dfs(n, k, 1, [], results)
        return results

    def dfs(self, n, k, startIdx, combination, results):
        
        if len(combination) == k:
            results.append(list(combination))
            return
        
        if startIdx > n:
            return
        
        for i in range(startIdx, n + 1):
            combination.append(i)
            self.dfs(n, k, i + 1, combination, results)
            combination.pop()
            
    def combine(self, n, k):
        # write your code here
        
        results = []
        self.dfs(n, k, 1, [], results)
        return results

    def dfs(self, n, k, startIdx, combination, results):
        
        if len(combination) == k:
            results.append(list(combination))
            return
        
        if startIdx > n:
            return
        
        combination.append(startIdx)
        self.dfs(n, k, startIdx + 1, combination, results)
        combination.pop()
        
        self.dfs(n, k, startIdx + 1, combination, results)
        
    def subsets(self, nums):
        # write your code here
        
        results = []
        nums.sort()
        self.dfs(nums, 0, [], results)
        
        return results
        
    def dfs(self, nums, startIdx, subset, results):
        results.append(list(subset))

        for i in range(startIdx, len(nums)):
            subset.append(nums[i])
            self.dfs(nums, i + 1, subset, results)
            subset.pop()
```

## DFS with memo
### 140. Word Break II
### 10. Regular Expression Matching
### 44. Wildcard Matching


```python
class Solution:
    def wordBreak(self, s: str, wordDict: List[str]) -> List[str]:
        
        
        if not wordDict:
            return []
            
            
        memo = {}
        
        
        def help(word):
            
            if not word:
                return [[]]
                
            
            if word in memo:
                return memo[word]
                
            memo[word] = []
            
            for w in wordDict:
                
                wordLen = len(w)
                
                if word[:wordLen] == w:
                    
                    resList = help(word[wordLen:])
                    
                    for res in resList:
                        
                        memo[word].append([w] + res)
                        
            return memo[word]
                        
        help(s)
        
        if not memo[s]:
            return []
        
        return [" ".join(seq) for seq in memo[s]]
```

## count number of components



```python
There are N students in a class. Some of them are friends, while some are not. Their friendship is transitive in nature. For example, if A is a direct friend of B, and B is a direct friend of C, then A is an indirect friend of C. And we defined a friend circle is a group of students who are direct or indirect friends.

Given a N*N matrix M representing the friend relationship between students in the class. If M[i][j] = 1, then the ith and jth students are direct friends with each other, otherwise not. And you have to output the total number of friend circles among all the students.

Example 1:
Input: 
[[1,1,0],
 [1,1,0],
 [0,0,1]]
Output: 2
Explanation:The 0th and 1st students are direct friends, so they are in a friend circle. 
The 2nd student himself is in a friend circle. So return 2.




class Solution:
    def findCircleNum(self, M: List[List[int]]) -> int:
        count = 0
        visited = [False] * len(M)
        
        
        def dfs(node):
            stack = [node]
            
            while stack:
                n = stack.pop()
                visited[n] = True
                
                for i in range(len(M)):
                    if M[n][i] == 1 and not visited[i]:
                        stack.append(i)
        
        
        
        
        for i in range(len(M)):
            if not visited[i]:
                dfs(i)
                count += 1
                
        return count
        
```

```python
"""
135. Combination Sum
中文English
Given a set of candidtate numbers candidates and a target number target. Find all unique combinations in candidates where the numbers sums to target.

The same repeated number may be chosen from candidates unlimited number of times.

Example
Example 1:

Input: candidates = [2, 3, 6, 7], target = 7
Output: [[7], [2, 2, 3]]
Example 2:

Input: candidates = [1], target = 3
Output: [[1, 1, 1]]
Notice
All numbers (including target) will be positive integers.
Numbers in a combination a1, a2, … , ak must be in non-descending order. (ie, a1 ≤ a2 ≤ … ≤ ak)
Different combinations can be in any order.
The solution set must not contain duplicate combinations.
"""


class Solution:
    """
    @param candidates: A list of integers
    @param target: An integer
    @return: A list of lists of integers
    """
    def combinationSum(self, candidates, target):
        # write your code here
        candidates = list(set(candidates))
        candidates.sort()
        results = []
        self.dfs(candidates, target, 0, [], results)
        return results
        
    def dfs(self, candidates, target, startIdx, temp, results):
        
        if target == 0:
            results.append(list(temp))
            return
        
        if target < 0:
            return
        
        for i in range(startIdx, len(candidates)):
            temp.append(candidates[i])
            self.dfs(candidates, target - candidates[i], i, temp, results)
            temp.pop()
```
