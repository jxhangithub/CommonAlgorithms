```python
#




# input =  [2,  3, 4,  5]
# output = [60, 40, 30, 24]


from typing import List
def multi(input: List[int]) -> List[int]:
   L = len(input)
   if L == 0:
       return None
   prefix = [1] * L
   suffix = [1] * L
   pre = 1
   for i in range(L):
       prefix[i] = pre
       pre *= input[i]
      
   suf = 1
   for i in range(L-1, -1, -1):
       suffix[i] = suf
       suf *= input[i]
      
      
   res = []
   for i in range(L):
       res.append(prefix[i] * suffix[i])
      
   return res
      
# input = [2, 3, 4, 5]
input = [2, 3, 0, 5]
print(multi(input))
# [60, 40, 30, 24]
# [0, 0, 30, 0]







# def addNumbers(a,b):
#     sum = a + b
#     return sum


# num1 = int(input())
# num2 = int(input())


# print("The sum is", addNumbers(num1, num2))


# course schedule


# numberofcourses 4
# [[1,0],[3,2]]


from typing import List
def topo(numCourse: int, requirement: List[List[int]]) -> bool:
   graph = [[] for _ in range(numCourse)]
   indegree = [0] * numCourse
   for second, first in requirement:
       graph[first].append(second)
       indegree[second] += 1
   cnt = 0
   q = []
   for i in range(numCourse):
       if indegree[i] == 0:
           q.append(i)
   while q:
       newQ = []
       for i in q:
           cnt += 1
           for n in graph[i]:
               indegree[n] -= 1
               if indegree[n] == 0:
                   newQ.append(n)
                  
       q = newQ
  
   return cnt == numCourse
  
  
            
            
print(topo(4, [[0, 1], [0, 2], [2, 3]]))


print(topo(4, [[0, 1], [1, 2], [2, 3], [3, 0]]))
```
