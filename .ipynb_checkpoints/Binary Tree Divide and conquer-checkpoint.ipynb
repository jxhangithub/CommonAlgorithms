```python
Preorder
class Solution:
    def preorderTraversal(self, root: TreeNode) -> List[int]:
        if not root:
            return []
        return [root.val] + self.preorderTraversal(root.left) + self.preorderTraversal(root.right)

# similar to dfs
class Solution:
    def preorderTraversal(self, root: TreeNode) -> List[int]:
        
        if not root:
            return []
        
        stack = [root]
        res = []
        while stack:
            node = stack.pop()
            if node:
                res.append(node.val)
                stack.append(node.right)
                stack.append(node.left)
                
        return res

# dfs
class Solution:
    def preorderTraversal(self, root: TreeNode) -> List[int]:
        
        if not root:
            return []
        
        stack = [root]
        res = []
        while stack:
            node = stack.pop()
            res.append(node.val)
            if node.right:
                stack.append(node.right)
            if node.left:
                stack.append(node.left)
                
        return res
    
    
# an alternative solution, can be adapted to inorder and post order
class Solution:
    def preorderTraversal(self, root: TreeNode) -> List[int]:
        
        stack = []
        res = []
        while stack or root:
            
            if root: # record the results and push left children into the stack
                stack.append(root)
                res.append(root.val)
                root = root.left
            else: # go to right for one step
                root = stack.pop()
                root = root.right
                
        return res


# iterative solution using additional Flag
class Solution:
    def preorderTraversal(self, root: TreeNode) -> List[int]:
        
        stack = [(root, False)]
        res = []
        while stack:
            
            node, visited = stack.pop()
            if not node:
                continue
            
            if visited:
                res.append(node.val)
            else:
                stack.extend([ (node.right, False), (node.left, False), (node, True)])
                
        return res
```

```python
Inorder

# recursive
class Solution:
    def inorderTraversal(self, root: TreeNode) -> List[int]:
        if not root:
            return []
        return self.inorderTraversal(root.left) + [root.val] + self.inorderTraversal(root.right)

# iteratively
class Solution:
    def inorderTraversal(self, root: TreeNode) -> List[int]:
        
        stack = []
        res = []     
        curr = root
        while stack or curr:
            while curr:
                stack.append(curr)
                curr = curr.left
            curr = stack.pop()
            res.append(curr.val)
            curr = curr.right
            
        return res
# iteratively, alternative    
class Solution:
    def inorderTraversal(self, root: TreeNode) -> List[int]:
        res = []
        stack = []
        curr = root
        
        while curr or stack:
            
            if curr:
                stack.append(curr)
                curr = curr.left
            else:  
                # since at least of them curr and stack has to be true, we do not need to check if stack:
                curr = stack.pop()
                res.append(curr.val)
                curr = curr.right

        return res
    
# iterative using additional flags
class Solution:
    def inorderTraversal(self, root: TreeNode) -> List[int]:
        res = []
        stack = [(root, False)]

        while stack:
            
            node, visited = stack.pop()
            if not node:
                continue
            if visited:
                res.append(node.val)
            else:
                stack.extend([(node.right, False), (node, True), (node.left, False)])
            
                
        return res
            
```

```python
Post order
class Solution:
    def postorderTraversal(self, root: TreeNode) -> List[int]:
        if not root:
            return []
        return self.postorderTraversal(root.left) + self.postorderTraversal(root.right) + [root.val]
    
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

# iterative solution using three whiles
class Solution:
    def postorderTraversal(self, root: TreeNode) -> List[int]:
        
        stack = []
        res = []
        while stack or root:
            
            # add itself and all the left children
            while root:
                stack.append(root)
                root = root.left
            else:  
                prev = None
                while stack:
                    root = stack.pop()

                    if not root.right or root.right == prev: # if no right children, or right children has been visited
                        res.append(root.val)
                        prev = root
                        root = None # this is critical, set to None

                    else: # if has right child, need to push the root back into stack again and go to the right     
                        stack.append(root)
                        root = root.right
                        break # we break the while because we need to add the left children again.
                    
        return res
    
    
Alterantive solution using two if

class Solution:
    def postorderTraversal(self, root: TreeNode) -> List[int]:
        
        stack = []
        res = []
        prev = None
        while stack or root:          
            # add itself and all the left children
            if root:
                stack.append(root)
                root = root.left
            else:  
                # if no right children, or right children has been visited
                root = stack.pop()
                if not root.right or root.right == prev: # if no right children, or right children has been visited (equivalent to no right child)
                    res.append(root.val)
                    prev = root
                    root = None # critical
                else: # if has right child, need to push the root back into stack again and go to the right  
                    #elif root.right and root.right != prev:
                    stack.append(root)
                    root = root.right

        return res

    
# iterative solution using additional Flag
class Solution:
    def postorderTraversal(self, root: TreeNode) -> List[int]:

        stack = [(root, False)]
        res = []
        while stack:
            
            node, visited = stack.pop()
            if not node:
                continue
            
            if visited:
                res.append(node.val)
            else:
                stack.extend([(node, True), (node.right, False), (node.left, False)])
                
        return res
```

```python
590. N-ary Tree Postorder Traversal

Given an n-ary tree, return the postorder traversal of its nodes' values.

Nary-Tree input serialization is represented in their level order traversal, each group of children is separated by 
the null value (See examples).

"""
# Definition for a Node.
class Node:
    def __init__(self, val=None, children=None):
        self.val = val
        self.children = children
"""

# recursive
class Solution:
    def postorder(self, root: 'Node') -> List[int]:
        
        if not root:
            return []
        
        res = []
        for c in root.children:
            res.extend(self.postorder(c))
        res.append(root.val)
        
        return res
    
# iterative first version, mimicing the recursive process
class Solution:
    def postorder(self, root: 'Node') -> List[int]:
        
        stack = []
        res = []
        nextIdx = 0        
        
        while stack or root:
            
            if not root:
                root, nextIdx = stack.pop()
            
            if nextIdx < len(root.children): 
                stack.append((root, nextIdx + 1))
                root = root.children[nextIdx]
                nextIdx = 0
            else:# if it is a leaf or all childrens are done
                res.append(root.val)
                root = None
        
        return res
    
# iterative second version, using additional flag
class Solution:
    def postorder(self, root: 'Node') -> List[int]:
        
        stack = [(root, False)]
        res = []
        
        while stack:
            
            node, visited = stack.pop()
            if not node:
                continue
                
            if visited:
                res.append(node.val)
            else:
                stack.append((node, True))
                for c in reversed(node.children):
                    stack.append((c, False))
                    
        return res
```

```python
102. Binary Tree Level Order Traversal

Given a binary tree, return the level order traversal of its nodes' values. (ie, from left to right, level by level).

For example:
Given binary tree [3,9,20,null,null,15,7],

# BFS 
class Solution:
    def levelOrder(self, root: TreeNode) -> List[List[int]]:
        
        if not root:
            return []

        res = []        
        queue = [root]
        while queue:
            levelOrder = []            
            nextQueue = []
            for node in queue:
                levelOrder.append(node.val)
                if node.left:
                    nextQueue.append(node.left)
                if node.right:
                    nextQueue.append(node.right)
                    
            res.append(levelOrder)
            queue = nextQueue
            
        return res
    
DFS solution
class Solution:
    def levelOrder(self, root: TreeNode) -> List[List[int]]:
        
        if not root:
            return []

        res = []        
        stack = [(root, 0)]
        while stack:
            node, level = stack.pop()
            if len(res) == level:
                res.append([])
            res[level].append(node.val)
            # right first
            if node.right:
                stack.append((node.right, level + 1))
            
            if node.left:
                stack.append((node.left, level + 1))
```

```python
103. Binary Tree Zigzag Level Order Traversal
Given a binary tree, return the zigzag level order traversal of its nodes' values. (ie, from left to right, then right to left for the next level and alternate between).

For example:
Given binary tree [3,9,20,null,null,15,7],
    3
   / \
  9  20
    /  \
   15   7
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def zigzagLevelOrder(self, root: TreeNode) -> List[List[int]]:
        
        if not root:
            return []
        
        queue = [root]
        res = []
        
        while queue:
            
            newQ = []
            level = []
            
            for node in queue:
                level.append(node.val)
                if node.left:
                    newQ.append(node.left)
                if node.right:
                    newQ.append(node.right)
                    
            if len(res) % 2 == 0:
                res.append(level)
            else:
                res.append(level[::-1])
            queue = newQ
            
        return res
            
```

```python
314. Binary Tree Vertical Order Traversal

Given a binary tree, return the vertical order traversal of its nodes' values. (ie, from top to bottom, column by column).

If two nodes are in the same row and column, the order should be from left to right.

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

# BFS plus hashmap
class Solution:
    def verticalOrder(self, root: TreeNode) -> List[List[int]]:
        
        if not root:
            return []
        
        nodes = collections.defaultdict(list)
        
        queue = collections.deque([(0, root)])
        
        minCol = 0
        maxCol = 0
        while queue:
            
            col,  node = queue.popleft()
            nodes[col].append(node.val)
            if node.left:
                queue.append((col - 1, node.left))
                minCol = min(minCol, col - 1)            
            
            if node.right:
                queue.append((col + 1, node.right))
                maxCol = max(maxCol, col + 1)

        
        res = []
        for i in range(minCol, maxCol + 1):
            res.append(nodes[i])
        
        return res
            
            
            
                
```

```python
987. Vertical Order Traversal of a Binary Tree
Given a binary tree, return the vertical order traversal of its nodes values.

For each node at position (X, Y), its left and right children respectively will be at positions (X-1, Y-1) and (X+1, Y-1).

Running a vertical line from X = -infinity to X = +infinity, whenever the vertical line touches some nodes, we report the values of the nodes in order from top to bottom (decreasing Y coordinates).

If two nodes have the same position, then the value of the node that is reported first is the value that is smaller.

Return an list of non-empty reports in order of X coordinate.  Every report will have a list of values of nodes.

class Solution:
    def verticalTraversal(self, root: TreeNode) -> List[List[int]]:
        if not root:
            return []
        stack = [(0, 0, root)]
        h = []
        while stack: # 
            col, row,  node = stack.pop()
            h.append((col, row, node.val))
            if node.right:
                stack.append((col + 1, row + 1, node.right))
            if node.left:
                stack.append((col - 1, row + 1, node.left))    
                
        h.sort()
        res = []
        preCol = None
        for col, row, val in h:
            if preCol is None or col > preCol:
                preCol = col
                res.append([])
                
            res[-1].append(val)
            
        return res        
```

```python
104. Maximum Depth of Binary Tree
Given a binary tree, find its maximum depth.

The maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.

Note: A leaf is a node with no children.

Example:

Given binary tree [3,9,20,null,null,15,7],

    3
   / \
  9  20
    /  \
   15   7
return its depth = 3.

# maximum depth of a binary tree
def maxDepth(self, root):
    # write your code here
    if root is None:
        return 0

    return max(self.maxDepth(root.left), self.maxDepth(root.right)) + 1

class Solution:
    def maxDepth(self, root: TreeNode) -> int:
        
        
        
        if not root:
            return 0
        
        queue = [root]
        depth = 0
        while queue:
            
            newQ = []
            for i in range(len(queue)):
                node = queue.pop()
                if node.left:
                    newQ.append(node.left)
                if node.right:
                    newQ.append(node.right)
            depth += 1        
            queue = newQ
        
        return depth
```

```python
111. Minimum Depth of Binary Tree
Given a binary tree, find its minimum depth.

The minimum depth is the number of nodes along the shortest path from the root node down to the nearest leaf node.

Note: A leaf is a node with no children.

Example:

Given binary tree [3,9,20,null,null,15,7],

    3
   / \
  9  20
    /  \
   15   7

    
# consider four situations
class Solution:
    def minDepth(self, root: TreeNode) -> int:
        
        if not root:
            return 0
        
        if root.left and not root.right:
            return 1 + self.minDepth(root.left)
        
        if root.right and not root.left:
            return 1 + self.minDepth(root.right)
        
        return 1 + min(self.minDepth(root.left), self.minDepth(root.right))    

# alternative BFS method
def minDepth(self, root):
    # write your code here
    if root is None:
        return 0

    queue = collections.deque([root])
    count = 1

    while queue:

        next_queue = []
        for n in queue:
            if not n.left and not n.right:
                return count
            if n.left:
                next_queue.append(n.left)
            if n.right:
                next_queue.append(n.right)
        count += 1
        queue = next_queue
```

```python
100. Same Tree

Given two binary trees, write a function to check if they are the same or not.

Two binary trees are considered the same if they are structurally identical and the nodes have the same value.

# identifical tree
    """
    @param a: the root of binary tree a.
    @param b: the root of binary tree b.
    @return: true if they are identical, or false.
    
    """
# recusive solution
def isIdentical(self, a, b):
    # write your code here
    if not a and not b:
        return True

    if a and b:
        return (a.val == b.val) and self.isIdentical(a.left, b.left) and self.isIdentical(a.right, b.right)
    return False

# we can also use iterative solution
class Solution:
    def isSameTree(self, p: TreeNode, q: TreeNode) -> bool:
        
        def check(p, q):
            
            if not q and not p:
                return True
            
            if p and q and p.val == q.val:
                return True
            
            return False
        

        queue = collections.deque([(p, q)])
        
        while queue:
            
            p, q = queue.popleft()
            
            if not check(p, q):
                return False
            if p and q:
                queue.append((p.left, q.left))
                queue.append((p.right, q.right))
            
        return True 
```

```python
101. Symmetric Tree
Given a binary tree, check whether it is a mirror of itself (ie, symmetric around its center).

For example, this binary tree [1,2,2,3,4,4,3] is symmetric:

    1
   / \
  2   2
 / \ / \
3  4 4  3
    
# symmetric tree

class Solution:
    def isSymmetric(self, root: TreeNode) -> bool:
        
        
        def isMirror(left, right):
            if not left and not right:
                return True

            if left and right:
                
                return left.val == right.val and isMirror(left.right, right.left) and isMirror(left.left, right.right)
            
            return False
        
        if not root:
            return True
        
        return isMirror(root.left, root.right)
```

```python
110. Balanced Binary Tree

Given a binary tree, determine if it is height-balanced.

For this problem, a height-balanced binary tree is defined as:

a binary tree in which the left and right subtrees of every node differ in height by no more than 1.

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def isBalanced(self, root: TreeNode) -> bool:
        
        
        def help(root):
            if not root:
                return True, 0
            
            leftFlag, leftHeight = help(root.left)
            
            # early return
            if not leftFlag:
                return False, -1
            
            rightFlag, rightHeight = help(root.right)
            
            if leftFlag and rightFlag and abs(leftHeight - rightHeight) <= 1:
                    return True, max(leftHeight, rightHeight) + 1
            
            return False, -1
        
        res, _ = help(root)
        return res
                
```

```python
617. Merge Two Binary Trees

Given two binary trees and imagine that when you put one of them to cover the other, some nodes of the two trees are overlapped while the others are not.

You need to merge them into a new binary tree. The merge rule is that if two nodes overlap, then sum node values up as the new value of the merged node. Otherwise, the NOT null node will be used as the node of new tree.


class Solution:
    def mergeTrees(self, t1: TreeNode, t2: TreeNode) -> TreeNode:
        
        if not t1 and t2:
            return t2
        if not t2 and t1:
            return t1
        
        if not (t1 and t2):
            return None
        
        t1.left = self.mergeTrees(t1.left, t2.left)
        t1.right = self.mergeTrees(t1.right, t2.right)
        
        t1.val += t2.val
        
        return t1
```

```python
114. Flatten Binary Tree to Linked List
Given a binary tree, flatten it to a linked list in-place.

For example, given the following tree:
    
    # Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def flatten(self, root: TreeNode) -> None:
        """
        Do not return anything, modify root in-place instead.
        """
        def help(root):
            
            if not root:
                return root
            
            end = root
            
            leftEnd = help(root.left)
            rightEnd = help(root.right)
            
            if leftEnd:
                leftEnd.right = root.right
                root.right = root.left
                
                end = leftEnd
            
            if rightEnd:
                end = rightEnd

            
            root.left = None
            return end
        
        help(root)
        
        return
```

```python
596. Minimum Subtree
Given a binary tree, find the subtree with minimum sum. Return the root of the subtree.
class Solution:
    """
    @param root: the root of binary tree
    @return: the root of the minimum subtree
    """
    def findSubtree(self, root):
        # write your code here
        
        self.minNode = None
        self.minSum = 0
        
        def help(node):
            
            if not node:
                return 0
            leftSum = help(node.left)
            rightSum = help(node.right)
            totalSum = node.val + leftSum + rightSum
            
            if not self.minNode or self.minSum >= totalSum:
                self.minNode = node
                self.minSum = totalSum
            
            return totalSum
            
        help(root)
        
        return self.minNode
```

```python
### Lintcode597. Subtree with Maximum Average
class Solution:
    """
    @param root: the root of binary tree
    @return: the root of the maximum average of subtree
    """
    def findSubtree2(self, root):
        # write your code here

        self.maxNode = None
        self.maxVal = -float('inf')
        
        
        def help(node):
            if not node:
                return 0, 0
                
            leftSum, leftCount = help(node.left)
            rightSum, rightCount = help(node.right)
            
            average = (leftSum + rightSum + node.val) / (leftCount + rightCount + 1)
            
            if average > self.maxVal:
                self.maxNode = node
                self.maxVal = average
                
            return (leftSum + rightSum + node.val), (leftCount + rightCount + 1)
            
            
        help(root)
        return self.maxNode
```

```python
257. Binary Tree Paths
Given a binary tree, return all root-to-leaf paths.

Note: A leaf is a node with no children.
    Input:

   1
 /   \
2     3
 \
  5

Output: ["1->2->5", "1->3"]
    
post-order solution recursive (bottom up)
class Solution:
    def binaryTreePaths(self, root: TreeNode) -> List[str]:
        if not root:
            return []
        
        if not root.left and not root.right:
            return [str(root.val)]
        
            
        res = []
        combinedPath = self.binaryTreePaths(root.left) + self.binaryTreePaths(root.right)
        for e in combinedPath:
            res.append(str(root.val) + '->' + e)
            
        return res
    
dfs/backtracking recursive (top down)
    def binaryTreePaths(self, root: TreeNode) -> List[str]:
        def help(node, temp):
            
            if not node:
                return
             
            if not temp:
                temp = str(node.val)
            else:
                temp = temp + "->" + str(node.val)
                
            if not node.left and not node.right:
                res.append(temp)                
                
            if node.left:
                help(node.left, temp)
            if node.right:
                help(node.right, temp)
                
        res = []
        help(root, [])
        return res
    
dfs iterative (top down)
class Solution:
    def binaryTreePaths(self, root: TreeNode) -> List[str]:
        
        if not root:
            return []
        
        res = []
        stack = [(root, "")]
        
        while stack:  
            node, temp = stack.pop()  
            if not temp:
                temp = str(node.val)
            else:
                temp += '->' + str(node.val)
                  
            if not node.left and not node.right:
                res.append(temp)        
            if node.left:
                stack.append((node.left, temp))
            if node.right:
                stack.append((node.right, temp))
     
        return res
```

```python
298. Binary Tree Longest Consecutive Sequence

Given a binary tree, find the length of the longest consecutive sequence path.

The path refers to any sequence of nodes from some starting node to any node in the tree along the parent-child connections. 
The longest consecutive path need to be from parent to child (cannot be the reverse).


# top down
class Solution:
    def longestConsecutive(self, root: TreeNode) -> int:
        
        self.ans = 0
        
        def help(node, cumRes, prev):
            
            if not node:
                return
            
            if prev is None or node.val == prev + 1:
                cumRes += 1
                self.ans = max(self.ans, cumRes)
            else:
                cumRes = 1
                
            help(node.left, cumRes, node.val)
            help(node.right, cumRes, node.val)
            
        
        help(root, 0, None)
        
        return self.ans
    
# bottom up

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def longestConsecutive(self, root: TreeNode) -> int:
        
        self.ans = 0
        
        def help(node):
            
            if not node:
                return 0
            
            cumLeft = help(node.left)
            cumRight = help(node.right)
            
            if not node.left or node.val != node.left.val - 1:
                cumLeft = 0
            if not node.right or node.val != node.right.val - 1:
                cumRight = 0
                
            self.ans = max(self.ans, max(cumLeft, cumRight) + 1)
            
            return max(cumLeft, cumRight) + 1

        
        help(root)
        
        return self.ans
        
```

### 1372. Longest ZigZag Path in a Binary Tree

```python
Given a binary tree root, a ZigZag path for a binary tree is defined as follow:

Choose any node in the binary tree and a direction (right or left).
If the current direction is right then move to the right child of the current node otherwise move to the left child.
Change the direction from right to left or right to left.
Repeat the second and third step until you can't move in the tree.
Zigzag length is defined as the number of nodes visited - 1. (A single node has a length of 0).

Return the longest ZigZag path contained in that tree.

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

# recursive bottom up
class Solution:
    def longestZigZag(self, root: TreeNode) -> int:
        
        self.ans = 0
        
        def help(node):
            
            if not node:
                return 0, 0
            
            if node.left:
                leftLeft, leftRight = help(node.left)
                left = 1 + leftRight
            else:
                left = 1
            
            if node.right:
                rightLeft, rightRight = help(node.right)
                right = 1 + rightLeft
            else:
                right = 1
                
            self.ans = max(self.ans, right, left)
            
            return left, right
        
        help(root)
        
        return self.ans - 1 if self.ans != 0 else 0
    
# iterative top down
class Solution:
    def longestZigZag(self, root: TreeNode) -> int:
        
        if not root:
            return 0
        
        stack = [(root, 1, True), (root, 1, False)]
        maxVal = 1
        while stack:
                        # right Flag indicates we are searching for the right children
            node, count, rightFlag = stack.pop()
            
            maxVal = max(maxVal, count)
            if node.left:
                if rightFlag:
                    # if we have left child but we are searching for the right, we restart and search the left
                    # there is no harm to append another stack.append((node.left, 1, True)), but it is duplicate
                    # since stack = [(root, 1, True), (root, 1, False)]
                    stack.append((node.left, 1, False))
                else:
                    stack.append((node.left, count + 1, True))
                    
            if node.right:
                if rightFlag:
                    stack.append((node.right, count + 1, False))
                else:
                    stack.append((node.right, 1, True))
                    
            
        return maxVal - 1
```

```python
426. Convert Binary Search Tree to Sorted Doubly Linked List

Convert a Binary Search Tree to a sorted Circular Doubly-Linked List in place.

You can think of the left and right pointers as synonymous to the predecessor and successor 
pointers in a doubly-linked list. For a circular doubly linked list, the predecessor of the first
element is the last element, and the successor of the last element is the first element.

We want to do the transformation in place. After the transformation, the left pointer of the tree 
node should point to its predecessor, and the right pointer should point to its successor. You should 
return the pointer to the smallest element of the linked list.

"""
# Definition for a Node.
class Node:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
"""

class Solution:
    def treeToDoublyList(self, root: 'Node') -> 'Node':
        
        if not root:
            return None
        
        stack = []
        curr = root
        
        prev = None
        head = None
        while curr or stack:
            
            if curr:
                stack.append(curr)
                curr = curr.left
            else:
                
                node = stack.pop()
                if prev:
                    prev.right = node
                    node.left = prev
                else:
                    head = node
                
                prev = node
                curr = node.right
                
        prev.right = head
        head.left = prev
        
        return head
        
```

```python
297. Serialize and Deserialize Binary Tree

Serialization is the process of converting a data structure or object into a sequence of bits
so that it can be stored in a file or memory buffer, or transmitted across a network connection link to be reconstructed later in the same or another computer environment.

Design an algorithm to serialize and deserialize a binary tree. There is no restriction 
on how your serialization/deserialization algorithm should work. You just need to ensure that
a binary tree can be serialized to a string and this string can be deserialized to the original tree structure.

# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


# BFS approach
class Codec:

    def serialize(self, root):
        """Encodes a tree to a single string.
        
        :type root: TreeNode
        :rtype: str
        """
        if not root:
            return ''
        
        queue = collections.deque()
        queue.append(root)
        res = []
        while queue:
            
            node = queue.popleft()
            if node:
                res.append(str(node.val))
            else:
                res.append('#')
                
            if node:
                queue.append(node.left)
                queue.append(node.right)
                
        return ' '.join(res)
        

    def deserialize(self, data):
        """Decodes your encoded data to tree.
        
        :type data: str
        :rtype: TreeNode
        """
        data = data.split()
        
        if data:
            root = TreeNode(int(data[0]))
        else:
            return None
        queue = collections.deque([root])
        idx = 1
        while queue:
            
            node = queue.popleft()
            
            if data[idx] != '#':
                node.left = TreeNode(int(data[idx]))
                queue.append(node.left)
            if data[idx + 1] != '#':
                node.right = TreeNode(int(data[idx + 1]))
                queue.append(node.right)
            idx += 2
             
        return root        

# Your Codec object will be instantiated and called as such:
# codec = Codec()
# codec.deserialize(codec.serialize(root))

# DFS approach

class Codec:

    def serialize(self, root):
        """Encodes a tree to a single string.
        
        :type root: TreeNode
        :rtype: str
        """
        if not root:
            return '#'
        
        stack = [root]
        res = []
        while stack:
            
            node = stack.pop()
            if node:
                res.append(str(node.val))
            else:
                res.append('#')
                
            if node:
                stack.append(node.right)
                stack.append(node.left)
                
                
        return ' '.join(res)
        

    def deserialize(self, data):
        """Decodes your encoded data to tree.
        
        :type data: str
        :rtype: TreeNode
        """
        def help(data):
            
            if data[0] == '#':
                data.pop(0)
                return None
            
            root = TreeNode(int(data[0]))
            data.pop(0)
            root.left = help(data)
            root.right = help(data)
            
            return root
        
        data = data.split()
           
        
        root = help(data)
        
        return root
```

### 105. Construct Binary Tree from Preorder and Inorder Traversal

```python
105. Construct Binary Tree from Preorder and Inorder Traversal

Given preorder and inorder traversal of a tree, construct the binary tree.

Note:
You may assume that duplicates do not exist in the tree.

For example, given

preorder = [3,9,20,15,7]
inorder = [9,3,15,20,7]
Return the following binary tree:

    3
   / \
  9  20
    /  \
   15   7

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

# recursive solution
class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> TreeNode:
        
        if not preorder:
            return None
        
        rootVal = preorder[0]
        idx = inorder.index(rootVal)
        
        root = TreeNode(rootVal)
        root.left = self.buildTree(preorder[1:1 + idx], inorder[:idx])
        root.right = self.buildTree(preorder[1 + idx:], inorder[idx + 1:])
        
        return root

# iterative solution
class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> TreeNode:
        
        if not preorder:
            return None
        
        left, right = 0, len(preorder) - 1
        
        
        
        root = TreeNode(0)
        stack = [(root, left, left, right)]
        
        while stack:
            
            node, preleft, inleft, inright = stack.pop()
            rootVal = preorder[preleft]
            node.val = rootVal
            idx = inorder.index(rootVal)
            
            if idx - 1 >= inleft:
                node.left = TreeNode(0)
                stack.append((node.left, preleft + 1, inleft, idx - 1))
                
            if idx + 1 <= inright:
                node.right = TreeNode(0)
                # preleft + idx - inleft + 1 is the starting position of preorder for the right subtree
                stack.append((node.right, preleft + idx - inleft + 1, idx + 1, inright))
        
        return root
```

### 106. Construct Binary Tree from Inorder and Postorder Traversal

```python
Given inorder and postorder traversal of a tree, construct the binary tree.

Note:
You may assume that duplicates do not exist in the tree.

For example, given

inorder = [9,3,15,20,7]
postorder = [9,15,7,20,3]
Return the following binary tree:

    3
   / \
  9  20
    /  \
   15   7
    
    
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def buildTree(self, inorder: List[int], postorder: List[int]) -> TreeNode:
        
        
        if not inorder:
            return None
        
        idx = inorder.index(postorder[-1])
        root = TreeNode(postorder[-1])
        
        root.left = self.buildTree(inorder[:idx], postorder[:idx])
        root.right = self.buildTree(inorder[1 + idx:], postorder[idx: - 1])
        
        
        return root    
```

### 742. Closest Leaf in a Binary Tree

```python
742. Closest Leaf in a Binary Tree

Given a binary tree where every node has a unique value, and a target key k, find the value of the nearest leaf node to target k in the tree.

Here, nearest to a leaf means the least number of edges travelled on the binary tree to reach any leaf of the tree. Also, a node is called a leaf if it has no children.

In the following examples, the input tree is represented in flattened form row by row. The actual root tree given will be a TreeNode object.


class Solution:
    def findClosestLeaf(self, root: TreeNode, k: int) -> int:
        
        
        G = collections.defaultdict(list)
        leafSet = set()
        def constructG(root):
            
            if not root:
                return
            
            if root.left:
                G[root.val].append(root.left.val)
                G[root.left.val].append(root.val)
                
                constructG(root.left)
            if root.right:
                G[root.val].append(root.right.val)
                G[root.right.val].append(root.val)
                constructG(root.right)
                
            if not root.left and not root.right:
                leafSet.add(root.val)
                
        constructG(root)
        
        
        queue = collections.deque([k])
        visited = set([k])
        
        while queue:
            node = queue.popleft()
            if node in leafSet:
                return node
            
            for nb in G[node]:
                if nb not in visited:
                    visited.add(nb)
                    queue.append(nb)
```

### 287. Find the Duplicate Number

```python
287. Find the Duplicate Number

Given an array nums containing n + 1 integers where each integer is between 1 and n (inclusive), 
prove that at least one duplicate number must exist. Assume that there is only one duplicate number, find the duplicate one.

class Solution:
    def findDuplicate(self, nums: List[int]) -> int:
        
        left, right = 1, len(nums) - 1
        
        while left < right:
            
            mid = (left + right) // 2
            count = 0
            for i in nums:
                if i <= mid:
                    count+=1           
            # if the # of numbers <= mid, duplicates must in [mid + 1, ]
            if count <= mid:
                left = mid + 1
            elif count > mid:
                right = mid
                
        return left
```
