```python
class UF:
    
    def __init__(self, size):
        
        self.parent = []
        self.rank = [] #the “rank” refers to the height of each vertex
        
        self.count = size
        
        for i in range(self.size):
            self.parent.append[i]
            self.rank.append[0]
            
            
    # recursive find parent with path compression
    def find1_1(self, i):
        if self.parent[i] != i:
            self.parent[i] = self.find(self.parent[i])
            
        return self.parent[i]
    
    def find1_2(self, x):
        if x == self.root[x]:
            return x
        self.root[x] = self.find(self.root[x])
        return self.root[x]
    # iterative find parent without path compression
    def find1_3(self, x):
        while x != self.parent[x]:
            x = self.parent[x]
        return self.parent[x]
    
    
    def find2(self, x):
        j = x
        while self.parent[j] != j:
            j = self.parent[j]
        
        # path compression
        while x != j:
            fx = self.parent[x]
            self.parent[x] = j
            x = fx
            
        return j
    
    def find3(self, x):
        path = []
        while self.parent[x] != x:
            path.append(x)
            x = self.parent[x]
        for n in path:
            self.parent[n] = x
    
    
    def union1(self, i, j):
        
        p = self.find(i)
        q = self.find(j)
        
        if p == q: return
        
        if self.rank[p] < self.rank[q]:
            self.parent[p] = q
        elif self.rank[p] > self.rank[q]:
            self.parent[q] = p
        else:
            self.parent[p] = q
            self.rank[q] += 1
            
        self.count -= 1
        
    def union2(self. i, j):
        fx = self.find[i]
        fy = self.find[j]
        
        if fx == fy:
            return
        
        self.parent[fx] = fy
        self.count -= 1 
        
        # if not counting components
    def union3(self, i, j):
        self.parent[self.find(i)] = self.find(j)
```

```python
There are N students in a class. Some of them are friends, while some are not. Their friendship is transitive in nature. For example, if A is a direct friend of B, and B is a direct friend of C, then A is an indirect friend of C. And we defined a friend circle is a group of students who are direct or indirect friends.

Given a N*N matrix M representing the friend relationship between students in the class. If M[i][j] = 1, then the ith and jth students are direct friends with each other, otherwise not. And you have to output the total number of friend circles among all the students.

Example 1:
Input: 
[[1,1,0],
 [1,1,0],
 [0,0,1]]
Output: 2
Explanation:The 0th and 1st students are direct friends, so they are in a friend circle. 
The 2nd student himself is in a friend circle. So return 2.

class Solution:
    def findCircleNum(self, M: List[List[int]]) -> int:
        
        if not M:
            return 0
        
        count = len(M)
        
        
        parent = [i for i in range(len(M))]
        rank = [0] * len(M)
        def find(i):
            
            if parent[i] != i:
                parent[i] = find(parent[i])
                
            return parent[i]
        
        def union(i, j):
            p = find(i)
            q = find(j)
            
            if p == q:
                return
            
            if rank[p] < rank[q]:
                parent[p] = q
            elif rank[p] > rank[q]:
                parent[q] = p
            else:
                parent[p] = q
                rank[q] += 1
                
            nonlocal count
            count -= 1
            
        for i in range(len(M)):
            for j in range(len(M[0])):
                if M[i][j] == 1:
                    union(i, j)
                
                
        return count

```

```python
Given an unsorted array of integers, find the length of the longest consecutive elements sequence.

Your algorithm should run in O(n) complexity.

Example:

Input: [100, 4, 200, 1, 3, 2]
Output: 4
Explanation: The longest consecutive elements sequence is [1, 2, 3, 4]. Therefore its length is 4.


class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        
        if not nums:
            return 0
        
        parent = {}
        rank = {}
        size = {}
        for e in nums:
            parent[e] = e
            rank[e] = 0
            size[e] = 1
            
        def find(e):
            
            if parent[e] != e:
                parent[e] = find(parent[e])
                
            return parent[e]
        
        def union(i, j):
            p = find(i)
            q = find(j)
            
            if p == q:
                return
            
            if rank[p] < rank[q]:
                parent[p] = q
                size[q] += size[p]
            elif rank[q] < rank[p]:
                parent[q] = p
                size[p] += size[q]
            else:
                parent[q] = p
                rank[p] += 1
                size[p] += size[q]
                
        maxSize = 0
        for e in nums:
            if e - 1 in parent:
                union(e - 1, e)
            if e + 1 in parent:
                union(e + 1, e)
            maxSize = max(maxSize, size[find(e)])

        return maxSize
```

### 261. Graph Valid Tree

```python
class Solution:
    def validTree(self, n: int, edges: List[List[int]]) -> bool:
        if len(edges) != n - 1:
            return False
        
        count = n        
        parent = [i for i in range(n)]

        def find(i):
            if parent[i] != i:
                parent[i] = find(parent[i])
            return parent[i]
        
        
        for e in edges:
            p = find(e[0])
            q = find(e[1])
            if p != q:
                parent[p] = q
                count -= 1
        return count == 1

# just check cycle    
class Solution:
    def validTree(self, n: int, edges: List[List[int]]) -> bool:
        if len(edges) != n - 1:
            return False
      
        parent = [i for i in range(n)]
        def find(i):
            if parent[i] != i:
                parent[i] = find(parent[i])
            return parent[i]
        
        
        for e in edges:
            p = find(e[0])
            q = find(e[1])
            if p != q:
                parent[p] = q
            elif p == q: # in p, q is already conntected somehow
                return False
        return True
```

###  721. Accounts Merge
Given a list accounts, each element accounts[i] is a list of strings, where the first element accounts[i][0] is a name, and the rest of the elements are emails representing emails of the account.

Now, we would like to merge these accounts. Two accounts definitely belong to the same person if there is some email that is common to both accounts. Note that even if two accounts have the same name, they may belong to different people as people could have the same name. A person can have any number of accounts initially, but all of their accounts definitely have the same name.

After merging the accounts, return the accounts in the following format: the first element of each account is the name, and the rest of the elements are emails in sorted order. The accounts themselves can be returned in any order.

Explanation: 
The first and third John's are the same person as they have the common email "johnsmith@mail.com".
The second John and Mary are different people as none of their email addresses are used by other accounts.
We could return these lists in any order, for example the answer [['Mary', 'mary@mail.com'], ['John', 'johnnybravo@mail.com'], 
['John', 'john00@mail.com', 'john_newyork@mail.com', 'johnsmith@mail.com']] would still be accepted.


```python

    
# an alternative cleaner implementation
class Solution:
    def accountsMerge(self, accounts: List[List[str]]) -> List[List[str]]:
        
        email_2_name = {}
        parents = {}
        for account in accounts:
            name = account[0]
            emails = account[1:]
            for email in emails:
                parents[email] = email
                email_2_name[email] = name
        
        def find(x):
            if parents[x] != x:
                parents[x] = find(parents[x])
            return parents[x]
        
        
        for account in accounts:
            first_email = account[1]
            emails = account[2:]
            for email in emails:
                p = find(first_email)
                q = find(email)
                if p != q:
                    parents[q] = p
                    
        
        unique_key_2_emails = collections.defaultdict(list)
        
        # must use find(email) rather than parents[email], because find gives the ultimate class label
        for email in parents:
            unique_key_2_emails[find(email)].append(email)
            
        
        res = []
        for key, emails in unique_key_2_emails.items():
            res.append([email_2_name[key]] + sorted(emails))
        return res
```

### 305. Number of Islands II

```python
305. Number of Islands II

A 2d grid map of m rows and n columns is initially filled with water. We may perform an addLand operation which turns the water at position (row, col) into a land. Given a list of positions to operate, count the number of islands after each addLand operation. An island is surrounded by water and is formed by connecting
adjacent lands horizontally or vertically. You may assume all four edges of the grid are all surrounded by water.


# union find with the mapping from tuples to parents
class Solution:
    def numIslands2(self, m: int, n: int, positions: List[List[int]]) -> List[int]:
        
        def find(x):
            if island[x] != x:
                island[x] = find(island[x])
        
            return island[x]
        
        def union(x, y):
            p = find(x)
            q = find(y)
            
            if p != q:
                island[p] = q
                self.count -= 1
            
        
        self.count = 0
        res = []
        island = {}
        for i, p in enumerate(positions):
            
            idx0 = p[0] * n + p[1]
            
            if idx0 not in island:# critical, because we are union while adding, cannot change island[idx0]
                # map tuple to integers
                island[idx0] = idx0
                self.count += 1

                # union
                for dx, dy in (-1, 0), (1, 0), (0, -1), (0, 1):
                    newX = p[0] + dx
                    newY = p[1] + dy

                    idx = newX * n + newY
                    if idx in island:
                        union(idx0, idx)
                        
            res.append(self.count)
            
        return res
    
# directly use tuple as the label

class Solution:
    def numIslands2(self, m: int, n: int, positions: List[List[int]]) -> List[int]:
        
        def find(x):
            if island[x] != x:
                island[x] = find(island[x])
        
            return island[x]
        
        def union(x, y):
            p = find(x)
            q = find(y)
            
            if p != q:
                island[p] = q
                self.count -= 1
            
        
        self.count = 0
        res = []
        island = {}
        for i, p in enumerate(positions):
            
            if tuple(p) not in island:# critical, because we are union while adding, cannot change island[tuple(p)]

                island[tuple(p)] = tuple(p)
                self.count += 1

                # union
                for dx, dy in (-1, 0), (1, 0), (0, -1), (0, 1):
                    newX = p[0] + dx
                    newY = p[1] + dy

                       
                    if (newX, newY) in island:
                        union(tuple(p), (newX, newY))
                        
            res.append(self.count)
            
        return res
```

### 839. Similar String Groups

Two strings X and Y are similar if we can swap two letters (in different positions) of X, so that it equals Y. Also two strings X and Y are similar if they are equal.

For example, "tars" and "rats" are similar (swapping at positions 0 and 2), and "rats" and "arts" are similar, but "star" is not similar to "tars", "rats", or "arts".

Together, these form two connected groups by similarity: {"tars", "rats", "arts"} and {"star"}.  Notice that "tars" and "arts" are in the same group even though they are not similar.  Formally, each group is such that a word is in the group if and only if it is similar to at least one other word in the group.

We are given a list A of strings.  Every string in A is an anagram of every other string in A.  How many groups are there?

```python
class Solution:
    def numSimilarGroups(self, A: List[str]) -> int:
        
        def is_similar(a, b):
            return len(list(filter(lambda i: a[i] != b[i], range(len(a))))) in (0, 2)
        
        
        parent = {s: s for s in A}
        def find(s):
            if parent[s] != s:
                parent[s] = find(parent[s])
            return parent[s]
        
        nGroups = len(parent)
        
        keys = list(parent.keys())
        for i in range(len(parent)):
            for j in range(i + 1, len(parent)):
                s1, s2 = keys[i], keys[j]
                if is_similar(s1, s2):
                    q = find(s1)
                    p = find(s2)
                    if p != q:
                        nGroups -= 1
                        parent[q] = p
        
        return nGroups

# C++ implemenetation
class Solution {
public:
    int numSimilarGroups(vector<string>& A) {
        
        unordered_map<string, string> parent;
        for (auto s: A){
            parent[s] = s;
        }
        int nGroup = parent.size();
        
        for(int i = 0; i < A.size(); i++ ){
            for (int j = i + 1; j < A.size(); j++){
                if(isSimilar(A[i], A[j])){
                    string p = find(parent, A[i]);
                    string q = find(parent, A[j]);
                    if (p != q) {
                        nGroup -= 1;
                        parent[q] = p;
                    }
                }       
            }
        }
        return nGroup;
        
    }
    string find(unordered_map<string, string>& parent, const string& p){
        if (parent[p] != p) {
            parent[p] = find(parent, parent[p]);
        }
        return parent[p];
    }
    
    bool isSimilar(const string& A, const string& B){
        int count = 0;
        for (int i = 0; i < A.size(); i++){
            if (A[i] != B[i]){
                count += 1;
                if (count > 2){
                    return false;
                }
            }
            
        }
        return true;
        
    }
};
```

### 684. Redundant Connection

```python
In this problem, a tree is an undirected graph that is connected and has no cycles.

The given input is a graph that started as a tree with N nodes (with distinct values 1, 2, ..., N), with one additional edge added. The added edge has two different vertices chosen from 1 to N, and was not an edge that already existed.

The resulting graph is given as a 2D-array of edges. Each element of edges is a pair [u, v] with u < v, that represents an undirected edge connecting nodes u and v.

Return an edge that can be removed so that the resulting graph is a tree of N nodes. If there are multiple answers, return the answer that occurs last in the given 2D-array. The answer edge [u, v] should be in the same format, with u < v.
class Solution:
    def findRedundantConnection(self, edges: List[List[int]]) -> List[int]:
        
        n = len(edges)
        parents = [i for i in range(n)]
        
        def find(x):
            if parents[x] != x:
                parents[x] = find(parents[x])
            return parents[x]
        
        for e in edges:
            p = find(e[0] - 1)
            q = find(e[1] - 1)
            
            if p != q:
                parents[p] = q
            else:# form a cycle, must be redundant
                return e
            
# with path compression            
class Solution {
public:
    unordered_map<int, int> parents;
    
    vector<int> findRedundantConnection(vector<vector<int>>& edges) {
        for (auto e: edges) {
            if (parents.count(e[0]) == 0)
                parents.insert({e[0], e[0]});
            if (parents.count(e[1]) == 0)
                parents.insert({e[1], e[1]});
            int p = find(e[0]);
            int q = find(e[1]);
            if (p != q)
                parents[p] = q;
            else
                return e;
        }
        return {};
    }
    
    int find(int x) {
        if (parents[x] != x) {
            parents[x] = find(parents[x]);
        }
        return parents[x];
    }
};
#without compression
class Solution {
public:
    unordered_map<int, int> parents;
    
    vector<int> findRedundantConnection(vector<vector<int>>& edges) {
        for (auto e: edges) {
            if (parents.count(e[0]) == 0)
                parents.insert({e[0], e[0]});
            if (parents.count(e[1]) == 0)
                parents.insert({e[1], e[1]});
            int p = parents[e[0]], q = parents[e[1]];
            while (parents[p] != p) p = parents[p];
            while (parents[q] != q) q = parents[q];
            if (p != q)
                parents[p] = q;
            else
                return e;
        }
        return {};
    }
};
```

### 1319. Number of Operations to Make Network Connected
There are n computers numbered from 0 to n-1 connected by ethernet cables connections forming a network where connections[i] = [a, b] represents a connection between computers a and b. Any computer can reach any other computer directly or indirectly through the network.

Given an initial computer network connections. You can extract certain cables between two directly connected computers, and place them between any pair of disconnected computers to make them directly connected. Return the minimum number of times you need to do this in order to make all the computers connected. If it's not possible, return -1. 

```python
class Solution:
    def makeConnected(self, n: int, connections: List[List[int]]) -> int:
        
        parent = [i for i in range(n)]
        numComponents = n
        numRedundant = 0
        for e in connections:
            p = self.find(e[0], parent)
            q = self.find(e[1], parent)
            if p != q:
                parent[p] = q
                numComponents -= 1
            else:
                numRedundant += 1
                
        if numRedundant >= (numComponents - 1):
            return (numComponents - 1)
        else:
            return -1
        

    def find(self, x, parent):
        if x != parent[x]:
            parent[x] = self.find(parent[x], parent)
        return parent[x]
    
    
    
```

### 990. Satisfiability of Equality Equations
Given an array equations of strings that represent relationships between variables, each string equations[i] has length 4 and takes one of two different forms: "a==b" or "a!=b".  Here, a and b are lowercase letters (not necessarily different) that represent one-letter variable names.

Return true if and only if it is possible to assign integers to variable names so as to satisfy all the given equations.

```python
class Solution:
    def equationsPossible(self, equations: List[str]) -> bool:
        
        variables = set()
        equalties = []
        inEqualities = []
        for eq in equations:
            variables.add(eq[0])
            variables.add(eq[-1])
            if eq[1:3] == '==':
                equalties.append([eq[0], eq[-1]])
            else:
                inEqualities.append([eq[0], eq[-1]])
        
        parents = {e:e for e in variables}
    
        for e in equalties:
            p = self.find(e[0], parents)
            q = self.find(e[1], parents)
            parents[p] = q
        for e in inEqualities:
            if self.find(e[0], parents) == self.find(e[1], parents):
                return False
        return True
    
    def find(self, x, parent):
        while x != parent[x]:
            x = parent[x]
        
        label = x
        while x != label:
            temp = parent[x]
            parent[x] = label
            x = temp
        return parent[x]
        
```

### 827. Making A Large Island

In a 2D grid of 0s and 1s, we change at most one 0 to a 1.

After, what is the size of the largest island? (An island is a 4-directionally connected group of 1s).

```python
class Solution:
    def largestIsland(self, grid: List[List[int]]) -> int:
        
        H = len(grid)
        W = len(grid[0])
        parents = {}
        size = {}
        for i in range(H):
            for j in range(W):
                if grid[i][j]:
                    parents[i, j] = (i, j)
                    size[i, j] = 1
    
        for i in range(H):
            for j in range(W):
                if grid[i][j]:
                    for dx, dy in (-1, 0), (1, 0), (0, -1), (0, 1):
                        newX = i + dx
                        newY = j + dy
                        if (newX, newY) in parents:
                            self.union((i, j), (newX, newY), parents, size)
        
        if not size:
            return 1
        maxSize = max(size.values())
        for i in range(H):
            for j in range(W):
                if grid[i][j] == 0:
                    connected = set()
                    for dx, dy in (-1, 0), (1, 0), (0, -1), (0, 1):
                        newX = i + dx
                        newY = j + dy
                        x = (newX, newY)
                        if x in parents:
                            p = self.find(x, parents)
                            connected.add(p)
                    tempSize = 1
                    for e in connected:
                        tempSize += size[e]
                    maxSize = max(maxSize, tempSize)
        return maxSize

    def union(self, x, y, parents, size):
        p = self.find(x, parents)
        q = self.find(y, parents)
        if p != q:
            parents[p] = q
            size[q] += size[p]
            
    
    def find(self, x, parents):
        if parents[x] != x:
            parents[x] = self.find(parents[x], parents)
        return parents[x]
    
```

### Lintcode 1396  Set Union
There is a list composed by sets. If two sets have the same elements, merge them. Returns the last remaining collection.

```python
class Solution:
    """
    @param sets: Initial set list
    @return: The final number of sets
    """
    def setUnion(self, sets):
        # Write your code here
        parents = {}
        for s in sets:
            for c in s:
                parents[c] = c

        def find(x):
            if parents[x] != x:
                parents[x] = find(parents[x])
            return parents[x]

        count = len(parents)
        for s in sets:
            for i in range(1, len(s)):
                p = find(s[i - 1])
                q = find(s[i])

                if p != q:
                    parents[p] = q
                    count -= 1

        return count
```

### 2076. Process Restricted Friend Requests

You are given an integer n indicating the number of people in a network. Each person is labeled from 0 to n - 1.

You are also given a 0-indexed 2D integer array restrictions, where restrictions[i] = [xi, yi] means that person xi and person yi cannot become friends, either directly or indirectly through other people.

Initially, no one is friends with each other. You are given a list of friend requests as a 0-indexed 2D integer array requests, where requests[j] = [uj, vj] is a friend request between person uj and person vj.

A friend request is successful if uj and vj can be friends. Each friend request is processed in the given order (i.e., requests[j] occurs before requests[j + 1]), and upon a successful request, uj and vj become direct friends for all future friend requests.

Return a boolean array result, where each result[j] is true if the jth friend request is successful or false if it is not.

Note: If uj and vj are already direct friends, the request is still successful.

```python
class Solution:
    def friendRequests(self, n: int, restrictions: List[List[int]], requests: List[List[int]]) -> List[bool]:
        
        parents = {i:i for i in range(n)}
        
        def find(x):
            if parents[x] != x:
                parents[x] = find(parents[x])
            return parents[x]
        
        # check if restriction can by satisfied
        def check_req(i, j):
            p = find(i)
            q = find(j)
            for e in restrictions:
                if find(e[0]) == p and find(e[1]) == q:
                    return False
                if find(e[0]) == q and find(e[1]) == p:
                    return False
            return True
        
        res = []
        for req in requests:
            p = find(req[0])
            q = find(req[1])
            if p != q:
                if check_req(req[0], req[1]):
                    parents[p] = q
                    res.append(True)
                else:
                    res.append(False)
            else:
                res.append(True)
                
                
        return res
```

### 1202. Smallest String With Swaps

You are given a string s, and an array of pairs of indices in the string pairs where pairs[i] = [a, b] indicates 2 indices(0-indexed) of the string.

You can swap the characters at any pair of indices in the given pairs any number of times.

Return the lexicographically smallest string that s can be changed to after using the swaps.

```python
class Solution:
    def smallestStringWithSwaps(self, s: str, pairs: List[List[int]]) -> str:
        
        parents = {i: i for i in range(len(s))}
        
        def find(x):
            if parents[x] != x:
                parents[x] = find(parents[x])
            return parents[x]
        
        for pair in pairs:
            p = find(pair[0])
            q = find(pair[1])
            if p != q:
                parents[p] = q
        
        d = defaultdict(list)
        for i in range(len(s)):
            d[find(i)].append(s[i])
        
        for k in d:
            d[k].sort(reverse=True)
            
        out = []
        for i in range(len(s)):
            out.append(d[find(i)].pop())
            
        return ''.join(out)
```

### number of connected components (Union Find)

```python
## Union find
class Solution:
    def countComponents(self, n: int, edges: List[List[int]]) -> int:
        
        
        parent = [i for i in range(n)]
        
        self.count = n
        
        def find(i):
            
            if parent[i] != i:
                parent[i] = find(parent[i])
            return parent[i]
        
        def union(i, j):
            
            p = find(i)
            q = find(j)
            if p == q:
                return
            parent[p] = parent[q]
            self.count -= 1
            
        for e in edges:
            union(e[0], e[1])
        
        return self.count
```
