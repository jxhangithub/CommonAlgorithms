### quick sort

```python
class Solution:
    def sortArray(self, nums: List[int]) -> List[int]:
        
        def partition(left, right, idx):
            
            nums[idx], nums[right] = nums[right], nums[idx]
            slow = left
            for i in range(left, right):
                
                if nums[i] < nums[right]:
                    nums[i], nums[slow] = nums[slow], nums[i]
                    slow += 1
                    
            nums[slow], nums[right] = nums[right], nums[slow]
            
            return slow
        
        def quickSort(start, end):
            
            if start >= end:
                return
            
            idx = random.randint(start, end)
            
            k = partition(start, end, idx)
            
            quickSort(start, k - 1)
            quickSort(k + 1, end)
            
            
        quickSort(0, len(nums) - 1)
        
        return nums
            
                
```

### Merge sort

```python
class Solution:
    def sortArray(self, nums: List[int]) -> List[int]:
        
        def mergeSort(left, right):
            
            if left >= right:
                return
            
            mid = (left + right) // 2
            
            mergeSort(left, mid)
            mergeSort(mid + 1, right)
            
            mergeSortedArray(left, mid, mid + 1, right)
            
        
        def mergeSortedArray(leftS, leftE, rightS, rightE):
            start = leftS
            buffer = [0] * (rightE - leftS + 1)
            
            p = 0
            while leftS <= leftE and rightS <= rightE:
                
                if nums[leftS] <= nums[rightS]:
                    buffer[p] = nums[leftS]
                    leftS += 1
                else:
                    buffer[p] = nums[rightS]
                    rightS += 1
                p += 1
            while leftS <= leftE:
                buffer[p] = nums[leftS]
                p += 1
                leftS += 1
            while rightS <= rightE:
                buffer[p] = nums[rightS]
                p += 1
                rightS += 1
            
            for i in range(len(buffer)):
                nums[start + i] = buffer[i]
                
        mergeSort(0, len(nums) - 1)
        return nums
```

## selection sort

```python
## selection sort
The selection sort algorithm sorts an array by repeatedly finding the minimum element (considering ascending order) 
from unsorted part and putting it at the beginning. The algorithm maintains two subarrays in a given array.

# recursive approach
def selectionSort1(A, left, right):
    
    if left >= right:
        return
    minIdx = left
    min = A[left]
    for j in range(left + 1, right + 1):
        if A[j] < min:
            min = A[j]
            minIdx = j
    A[left], A[minIdx] = A[minIdx], A[left]
    selectionSort(A, left + 1, right)
    
# @selectionSort
def selectionSort2(nums):
    for i in range(len(nums)):
        _min = min(nums[i:])
        min_index = nums[i:].index(_min)
        nums[i + min_index] = nums[i]
        nums[i] = _min
    return nums
    
    
    
```

### Counting Sort
Counting sort is a sorting technique based on keys between a specific range. It works by counting the number of objects having distinct key values (kind of hashing). Then doing some arithmetic to calculate the position of each object in the output sequence.

Counting sort is efficient if the range of input data is not significantly greater than the number of objects to be sorted. 


Time Complexity: O(n+k) where n is the number of elements in input array and k is the range of input.
Auxiliary Space: O(n+k)

```python
# counting sort for strings
def countingSort(s: str) -> str:
    
    N = 256
    
    count = [0] * N
    
    for c in s:
        count[ord(c)] += 1
        
    for i in range(N - 1):
        count[i + 1] += count[i]
    
    print(count)
    
    output = [0] * len(s)
    
    for i in range(len(s)):
        output[count[ord(s[i])] - 1] = s[i]
        count[ord(s[i])] -= 1

    return output

arr = "geeksforgeeks"
ans = countingSort(arr) 
print("Sorted character array is %s"  %("".join(ans)))

# counting sort for integers
def countSortIntArray(arr):
    
    maxVal = max(arr)
    minVal = min(arr)
    N = (maxVal - minVal + 1)
    count = [0] * N
    
    for e in arr:
        count[e] += 1
        
    output = []
    for e in range(minVal, maxVal + 1):
        if count[e] > 0:
            output.extend([e] * count[e])
    
    return output


arr = [-5, -10, 0, -3, 8, 5, -1, 10, 3, 10, -3]
print(countSortIntArray(arr))
```

### 75. Sort Colors (counting sort approach)

```python
Given an array with n objects colored red, white or blue, sort them in-place so that objects of the same color are adjacent, with the colors in the order red, white and blue.

Here, we will use the integers 0, 1, and 2 to represent the color red, white, and blue respectively.

class Solution:
    def sortColors(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        count = [0] * 3
        
        for e in nums:
            count[e] += 1
            
        for i in range(count[0]):
            nums[i] = 0
        for j in range(count[0], count[0] + count[1]):
            nums[j] = 1
        for k in range(count[0] + count[1], sum(count)):
            nums[k] = 2
            
        return nums
```

### 49. Group Anagrams (count sort idea)

```python
Given an array of strings, group anagrams together.
Time Complexity: O(NK), where NN is the length of strs, and KK is the maximum length of a string in strs. Counting each string is linear in the size of the string, and we count every string.

Space Complexity: O(NK), the total information content stored in ans.
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        
        def count(s):
            countResult = [0] * 26
            for c in s:
                countResult[ord(c) - ord('a')] += 1
            return countResult

        store = collections.defaultdict(list)
        for s in strs:
            key = tuple(count(s))
            store[key].append(s)
        
        return store.values()
```

### 1122. Relative Sort Array (counting sort and its generalization via hash)

```python
1122. Relative Sort Array
Given two arrays arr1 and arr2, the elements of arr2 are distinct, and all elements in arr2 are also in arr1.

Sort the elements of arr1 such that the relative ordering of items in arr1 are the same as in arr2. 
Elements that don't appear in arr2 should be placed at the end of arr1 in ascending order.
Constraints:

arr1.length, arr2.length <= 1000
0 <= arr1[i], arr2[i] <= 1000
Each arr2[i] is distinct.

# counting sort
class Solution:
    def relativeSortArray(self, arr1: List[int], arr2: List[int]) -> List[int]:
        
        N = 1001
        count = [0] * N
        for e in arr1:
            count[e] += 1
        
        output = []
        for e in arr2:
            output.extend([e] * count[e])
            count[e] = 0
        for e in range(N):
            if count[e] > 0:
                output.extend([e] * count[e])
                count[e] = 0
        
        return output


# generalize via hash when the sorted Array does not have limited range
class Solution:
    def relativeSortArray(self, arr1: List[int], arr2: List[int]) -> List[int]:
        
        count = {k:0  for k in arr2}
        
        tail = []
        for e in arr1:
            if e in count:
                count[e] += 1
            else:
                tail.append(e)
        tail.sort()
        output = []
        for k in arr2:
            output.extend([k] * count[k])
        output.extend(tail)
        
        return output
```

### Bucket sort

Counting sort can not be applied here as we use keys as index in counting sort. Here keys are floating point numbers. 

bucketSort(arr[], n)
1) Create n empty buckets (Or lists).
2) Do following for every array element arr[i].
.......a) Insert arr[i] into bucket[n*array[i]]
3) Sort individual buckets using insertion sort.
4) Concatenate all sorted buckets.

The complexity of the Bucket Sort Technique
Time Complexity: O(n + k) for best case and average case and O(n^2) for the worst case.

Space Complexity: O(nk) for worst case

桶排序和归并排序有那么点点类似，也使用了归并的思想。大致步骤如下：

设置一个定量的数组当作空桶。
Divide - 从待排序数组中取出元素，将元素按照一定的规则塞进对应的桶子去。
对每个非空桶进行排序，通常可在塞元素入桶时进行插入排序。
Conquer - 从非空桶把元素再放回原来的数组中。

```python
def bucketSort(arr, N):
    
    maxVal = max(arr)
    minVal = min(arr)
    
    
    count = [[] for _ in range(N + 1)]
    
    for e in arr:
        bucketIdx = int((e - minVal) / (maxVal - minVal) * N) # the bucketIdx ranges from 0 to N
        count[bucketIdx].append(e)
        
    output = []    
    for i in range(N + 1):
        count[i].sort()
        output.extend(count[i])
    
    
    
    return output


arr = [-5, -10, 0, -3, 8, 5, -1, 10, 3, 10, -3]
print(bucketSort(arr, 10))

import random

random.seed(0)
arr = []
for i in range(100):
    arr.append(random.uniform(0, 5))
print(bucketSort(arr, 10))
```

### 164. Maximum Gap (bucket count)

```python
Given an unsorted array, find the maximum difference between the successive elements in its sorted form.

Return 0 if the array contains less than 2 elements.

Example 1:

Input: [3,6,9,1]
Output: 3
Explanation: The sorted form of the array is [1,3,6,9], either
             (3,6) or (6,9) has the maximum difference 3.
 
# naive bucketsort
class Solution:
    def maximumGap(self, nums: List[int]) -> int:
        
        if len(nums) < 2:
            return 0
        N = len(nums)
        minVal, maxVal = min(nums), max(nums)
        if maxVal == minVal:
            return 0
        
        count = [[] for _ in range(N + 1)]
        for e in nums:
            idx = int((e - minVal) / (maxVal - minVal) * N)
            count[idx].append(e)

        output = []
        for i in range(N + 1):
            count[i].sort()
            output.extend(count[i])
        maxDiff = 0
        for i in range(N - 1):
            maxDiff = max(maxDiff, output[i + 1] - output[i])
        return maxDiff
    
# two pointer and store bucket max and min    
class Solution:
    def maximumGap(self, nums: List[int]) -> int:
        
        if len(nums) < 2:
            return 0
        N = len(nums)
        minVal, maxVal = min(nums), max(nums)
        if maxVal == minVal:
            return 0
        
        count = [[] for _ in range(N + 1)]
        bucketMin = [float('inf')] * (N + 1)
        bucketMax = [-float('inf')] * (N + 1)
        
        for e in nums:
            idx = int((e - minVal) / (maxVal - minVal) * N)
            count[idx].append(e)
            bucketMin[idx] = min(bucketMin[idx], e)
            bucketMax[idx] = max(bucketMax[idx], e)
            
        prevMax = None
        maxDiff = 0
        for i in range(N + 1):
            if count[i]:
                if prevMax is None:
                    prevMax = bucketMax[i]
                else:
                    maxDiff = max(maxDiff, bucketMin[i] - prevMax)
                    prevMax = bucketMax[i]
                
        return maxDiff
```

### 347. Top K Frequent Elements (bucket sort idea)

```python
Given a non-empty array of integers, return the k most frequent elements.
You can return the answer in any order.
# bucket sort idea O(n)
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        count = collections.Counter(nums)
        
        bucket = [[] for _ in range(len(nums) + 1)]
        
        for key in count:
            bucket[count[key]].append(key)
            
        output = []
        for i in range(len(nums), -1, -1):
            
            if not bucket[i]:
                continue
            
            numE = len(bucket[i])
            addE = bucket[i][:k]
            output.extend(addE)
            
            k -= len(addE)
            
            if k == 0:
                return output
            
        
```

### Stable sorting and application 1451. Rearrange Words in a Sentence

```python
Given a sentence text (A sentence is a string of space-separated words) in the following format:

First letter is in upper case.
Each word in text are separated by a single space.
Your task is to rearrange the words in text such that all words are rearranged in an increasing order of their lengths. If two words have the same length, arrange them in their original order.

Return the new text following the format shown above.

# note that python library sort() is stable sort
# even if it is not stable, we can create tuple (key, idx) to sort

class Solution:
    def arrangeWords(self, text: str) -> str:
        
        words = text.split()
        
        words = [w.lower() for w in words]
        
        toSort = []
        for i, w in enumerate(words):
            toSort.append((len(w), i, w))
        
        toSort.sort()

        firstWord = toSort[0][2]
        
        result = [firstWord[0].upper() + firstWord[1:]]

        for i in range(1, len(toSort)):
            result.append(toSort[i][2])

        return ' '.join(result)
```

## select K largest from array (quick select)


```python
215. Kth Largest Element in an Array

Example 1:

Input: [3,2,1,5,6,4] and k = 2
Output: 5
Example 2:

Input: [3,2,3,1,2,4,5,5,6] and k = 4
Output: 4

class Solution:
    def findKthLargest(self, nums: List[int], k: int) -> int:
        
        def partition(left, right, pivotIdx):
            
            nums[right], nums[pivotIdx] = nums[pivotIdx], nums[right]
            
            
            slowIdx = left
            
            for i in range(left, right):
                
                if nums[i] < nums[right]:
                    nums[slowIdx], nums[i] = nums[i], nums[slowIdx]
                    slowIdx += 1
                    
            nums[slowIdx], nums[right] = nums[right], nums[slowIdx]
            
            # return the rank of the nums[pivotIdx], 0 base in the whole nums
            return slowIdx 
        
        
        def select(left, right, kSmallest):
            # kSmallest is 0 base
            
            randomIdx = random.randint(left, right)
            
            pivotIdx = partition(left, right, randomIdx)
            
            if pivotIdx == kSmallest:
                return nums[pivotIdx]
            elif pivotIdx > kSmallest:
                return select(left, pivotIdx - 1, kSmallest)
            else:
                return select(pivotIdx + 1, right, kSmallest)
            
            
        return select(0, len(nums) - 1, len(nums) - k)
            
            
                    
```

```python
# partition algorithm
"""
Given an array nums of integers and an int k, partition the array (i.e move the elements in "nums") such that:

All elements < k are moved to the left
All elements >= k are moved to the right
Return the partitioning index, i.e the first index i nums[i] >= k.
https://www.lintcode.com/problem/partition-array/description
"""

def partitionArray(self, nums, k):
    # write your code here
    start, end = 0, len(nums) - 1

    while start <= end:

        while start <= end and nums[start] < k:
            start += 1
        while start <= end and nums[end] >= k:
            end -= 1

        if start <= end: # this is critical only when start <= end, we have nums[end] < k and nums[start] >=0; if start > end, we will not have this
            nums[start], nums[end] = nums[end], nums[start]
            start += 1
            end -= 1
    # if all numbers >= k, then start = 0, end = -1
    # if all numbers <k, then start = len(numbers), end = len(number) - 1
    
    return start


# choose the right-most element as pivot   
def partition(self, nums, l, r):
    low = l
    while l < r:
        if nums[l] < nums[r]:
            nums[l], nums[low] = nums[low], nums[l]
            low += 1
        l += 1
    nums[low], nums[r] = nums[r], nums[low]
    return low

def partition(left, right, pivotIdx):
    pivot = nums[pivotIdx]
    nums[pivotIdx], nums[right] = nums[right], nums[pivotIdx]

    storeIdx = left

    for i in range(left, right):
        if nums[i] < pivot:
            nums[storeIdx], nums[i] = nums[i], nums[storeIdx]
            storeIdx += 1


    nums[right], nums[storeIdx] = nums[storeIdx], nums[right]

    return storeIdx
```

```python
"""
quick sort
"""
    def sortIntegers2(self, A):
        # write your code here
        self.quickSort(A, 0, len(A) - 1)
        
    def quickSort(self, A, start, end):
        
        if start >= end:
            return
        
        left, right = start, end
        
        pivot = A[(start + end) // 2]
        
        while left <= right:
            # note that for A[left] < pivot, we can also use A[left] <= pivot. 
            # but use = can lead to unbalance partition. For example, consider all numbers in A are equal.
            # when left or right hid pivot, it will stop and wait for the other to echange
            # finally, the location of pivot is unclear
            while left <= right and A[left] < pivot:
                left += 1
            while left <= right and A[right] > pivot:
                right -= 1
            
            if left <= right:
                A[left], A[right] = A[right], A[left]
                left += 1
                right -= 1
        # finally, all numbers 
        
        self.quickSort(A, start, right)
        self.quickSort(A, left, end)
```

```python
# sort three colors

def sortColors(a):
    if not a:
        return
    
    pl, pr = 0, len(a)-1
    i = 0
    while i <= pr:
        if a[i] == 0:
            a[pl], a[i] = a[i], a[pl]
            pl += 1
            i += 1
        elif a[i] == 1:
            i += 1
        else:
            a[pr], a[i] = a[i], a[pr]
            pr -= 1
pl 和 pr 是传统的双指针，分别代表 0~pl-1 都已经是 0 了，pr+1~a.length - 1 都已经是 2 了。
另一个角度说就是，如果你发现了一个 0 ，就可以和 pl 上的数交换，pl 就可以 ++；如果你发现了一个 2 就可以和 pr 上的数交换 pr 就可以 --。

这样，我们用第三根指针 i 来循环整个数组。如果发现 0，就丢到左边（和 pl 交换，pl++），如果发现 2，就丢到右边（和 pr 交换，pr--），如果发现 1，就不管（i++）

这就是三根指针的算法，两根指针在两边，一根指针扫描所有的数。

这里有一个实现上的小细节，当发现一个 0 丢到左边的时候，i需要++，但是发现一个2 丢到右边的时候，i不用++。原因是，从pr 换过来的数有可能是0, 1或者2，需要继续判断丢到左边还是右边。而从 pl 换过来的数，要么是0要么是1，不需要再往右边丢了。因此这里 i 指针还有一个角度可以理解为，i指针的左侧，都是0和1。
```

```python
## merge two sorted arrays and produce a new one
class Solution:
    """
    @param A: sorted integer array A
    @param B: sorted integer array B
    @return: A new sorted integer array
    """
    def mergeSortedArray(self, A, B):
        # write your code here
        res = []
        first = 0
        second = 0
        while first < len(A) and second < len(B):
            if A[first] <= B[second]:
                res.append(A[first])
                first += 1
            else:
                res.append(B[second])
                second += 1
        
        while first < len(A):
            res.append(A[first])
            first += 1
            
        while second < len(B):
            res.append(B[second])
            second += 1
            
        return res
    
```

# Number theory

Prime number is a natural number that can be and can only be divided by 1 and itself.
1 is not prime, 
2, 3, 5, 7 are prime numbers

A composite number is a natural number that can be divided by more than 1 and itself.

In number theory, the fundamental theorem of arithmetic, also called the unique factorization theorem or the unique-prime-factorization theorem, states that every integer greater than 1 either is a prime number itself or can be represented as the product of prime numbers and that, moreover, this representation is unique, up to (except for) the order of the factors.

```python
method 1
def isPrime(n): 
      
    # Corner case 
    if (n <= 1): 
        return False
  
    # Check from 2 to n-1 
    for i in range(2, n): 
        if (n % i == 0):  # if can be divided, then must not be prime
            return False
  
    return True

method 2. We only need to check upto sqrt(n). 
The theory behind this is that if a number x is a composite number, then there must be one factor greater than 1 less than sqrt(n).
In other words, if x is prime, then there is no factor between 2 and sqrt(n), inclusively.
Proof: since x is prime number, then it must has a factor besides 1 and itself. Let the factor be a and a*b = x. 
    If both a and b are less than sqrt(x) or are both greater than sqrt(x), then a*b cannot hold. 



def isPrime(n): 
      
    # Corner case 
    if (n <= 1): 
        return False
  
    up = int(math.sqrt(n)) + 1
    # Check from 2 to n-1 
    for i in range(2, up): 
        if (n % i == 0): 
            return False
  
    return True
```

```python
204. Count Primes

Count the number of prime numbers less than a non-negative number, n.
class Solution:
    def countPrimes(self, n: int) -> int:
        isPrime = [True] * n
        
        count = 0
        for i in range(2, n):
            if isPrime[i]:
                count += 1
                for j in range(i * i, n, i):
                    isPrime[j] = False
            
        
        return count
            
```

```python
26. Remove Duplicates from Sorted Array
Given a sorted array nums, remove the duplicates in-place such that each element appear only once and return the new length.

Do not allocate extra space for another array, you must do this by modifying the input array in-place with O(1) extra memory.

class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        
        slow = -1
        for fast in range(len(nums)):
            if slow < 0 or nums[fast] != nums[slow]:
                slow += 1
                nums[slow] = nums[fast]
        return slow + 1
        
```

```python
122. Best Time to Buy and Sell Stock II
Say you have an array prices for which the ith element is the price of a given stock on day i.

Design an algorithm to find the maximum profit. You may complete as many transactions as you like (i.e., buy one and sell one share of the stock multiple times).

Note: You may not engage in multiple transactions at the same time (i.e., you must sell the stock before you buy again).

    
# trade at every possible gain position
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        
        res = 0
        for i in range(len(prices) - 1):
            
            if prices[i + 1] - prices[i] > 0:
                res += prices[i + 1] - prices[i]
            
        return res

# finding sequential velley and peaks and then  trade
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        
        res = 0
        
        i = 0
        valley = prices[0]
        peak = prices[0]
        
        while i < len(prices) - 1:
            
            while i < len(prices) - 1 and prices[i] >= prices[i + 1]:
                i += 1
                
            valley = prices[i]
            
            # stop condition at the final peak is prices[i] > prices[i + 1] or i == len(prices) - 1 and prices[i - 1] <= prices[i]
            while i < len(prices) - 1 and prices[i] <= prices[i + 1]:
                i += 1
            
            peak = prices[i]
            
            res += peak - valley
        return res
```

```python
Given an array, rotate the array to the right by k steps, where k is non-negative.

Example 1:

Input: [1,2,3,4,5,6,7] and k = 3
Output: [5,6,7,1,2,3,4]
    
class Solution:
    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        
        
        k = k % len(nums)
        
        
        def reverse(left, right):
            
            while left < right:
                nums[left], nums[right] = nums[right], nums[left]
                left += 1
                right -= 1
                
                
        reverse(0, len(nums) - 1)
        reverse(0, k - 1)
        reverse(k, len(nums) - 1)
```

```python
121. Best Time to Buy and Sell Stock

Say you have an array for which the ith element is the price of a given stock on day i.

If you were only permitted to complete at most one transaction (i.e., buy one and sell one share of the stock), design an algorithm to find the maximum profit.

Note that you cannot sell a stock before you buy one.


class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        
        maxProfit = 0
        minSoFar = float('inf')
        maxAfter = 0
        for e in prices:
            
            if minSoFar > e:
                minSoFar = e
                maxAfter = e
            
            maxAfter = max(maxAfter, e)
            
            maxProfit = max(maxProfit, maxAfter - minSoFar)
            
        return maxProfit
            
            
            
```

```python
384. Shuffle an Array

Shuffle a set of numbers without duplicates.


import random

class Solution:

    def __init__(self, nums: List[int]):
        
        self.data = nums
        self.origin = self.data.copy() # shallow copy

    def reset(self) -> List[int]:
        """
        Resets the array to its original configuration and return it.
        """
        self.data = self.origin.copy()
        
        return self.data

    def shuffle(self) -> List[int]:
        """
        Returns a random shuffling of the array.
        """
        for i in range(len(self.data)):
            swapIdx = random.randint(i, len(self.data) - 1)
            self.data[i], self.data[swapIdx] = self.data[swapIdx], self.data[i]
        
        return self.data
        
        


# Your Solution object will be instantiated and called as such:
# obj = Solution(nums)
# param_1 = obj.reset()
# param_2 = obj.shuffle()
```

```python
169. Majority Element

Given an array of size n, find the majority element. The majority element is the element that appears more than ⌊ n/2 ⌋ times.
Boyer-Moore Voting Algorithm
class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        
        count = 1
        e = nums[0]
        
        for i in range(1, len(nums)):
            
            if nums[i] == e:
                count += 1
            else:
                count -= 1
                if count == 0:
                    e = nums[i]
                    count = 1
        return e
        
```

### KMP algorithm

KMP algorithm preprocesses pat[] and constructs an auxiliary lps[] of size m (same as size of pattern) which is used to skip characters while matching.
name lps indicates longest proper prefix which is also suffix.. A proper prefix is prefix with whole string not allowed. For example, prefixes of “ABC” are “”, “A”, “AB” and “ABC”. Proper prefixes are “”, “A” and “AB”. Suffixes of the string are “”, “C”, “BC” and “ABC”.
We search for lps in sub-patterns. More clearly we focus on sub-strings of patterns that are either prefix and suffix.
For each sub-pattern pat[0..i] where i = 0 to m-1, lps[i] stores length of the maximum matching proper prefix which is also a suffix of the sub-pattern pat[0..i].

lps[i] = the longest proper prefix of pat[0..i] which is also a suffix of pat[0..i].


```python

def KMPSearch(pat, txt):
    
    M = len(pat)
    N = len(txt)
    
    # lps[i] = the longest proper prefix of pat[0..i] which is also a suffix of pat[0..i]. 
    lps = computeLPS(pat)
    
    i = 0
    j = 0
    while i < N:
        if pat[j] == txt[i]:
            i += 1
            j += 1
        if j == M:
            print("Found pattern at index ", i - j)
            j = lps[j - 1]
        elif i < N and pat[j] != txt[i]:
            if j > 0:
                j = lps[j - 1]
            else:# if at j = 0 we do not match, just move to next i
                i += 1
                
    
def computeLPS(pat):
    lps = [0] * len(pat)
    
    i = 1
    L = 0 # length of the previous longest prefix suffix 
    # the loop calculates lps[i] for i = 1 to M-1 
    while i < len(pat):
        if pat[i] == pat[L]:
            L += 1
            lps[i] = L
            i += 1
        else:
            
            if L > 0:
                L = lps[L - 1]
            else:
                lps[i] = 0
                i += 1
    print(lps)
    return lps
                
txt = "ABABDABACDABABCABAB"
pat = "ABABCABAB"
pat = "AAACAAAA"
KMPSearch(pat, txt) 
        
    
```

###  50.Pow(x, n)  （fast power)

```python
Implement pow(x, n), which calculates x raised to the power n (xn).

# recursive method
class Solution:
    def myPow(self, x: float, n: int) -> float:
        
        def help(x, n):
            
            if n == 0:
                return 1
            if n % 2 == 1:
                res = help(x, n // 2)
                return x * res * res
            else:
                res = help(x, n // 2)
                return res * res
            
            
        if n < 0:
            return 1.0 / help(x, -n)
        return help(x, n)
            
# iterative implemntation

class Solution:
    def myPow(self, x: float, n: int) -> float:
        
        def help(x, n):
            
            ans = 1
            while n:
                if n&1:# if is odd
                    ans *= x
                # square the base
                x *= x
                n = n >> 1
            return ans
            
        if n < 0:
            return 1.0 / help(x, -n)
        return help(x, n)
            
            
```

### fast matrix power

```python
#time complexity is N^3 log K, compared to naive N^3 K

def multiply(mat1, mat2):
    
    result = [[0] * len(mat2[0]) for _ in range(len(mat1))]
    
    for row in range(len(mat1)):
        for col in range(len(mat2[0])):
            for k in range(len(mat2)):
                result[row][col] += mat1[row][k] * mat2[k][col]
    return result

def fastMatrixPower(matrix, exp):
    
    if exp == 1:
        return matrix
    else:
        temp = fastMatrixPower(matrix, exp // 2)
        result = multiply(temp, temp)
        if exp % 2 == 1:
            return multiply(matrix, result)
        else:
            return result
```
