```python
Given a sorted linked list, delete all nodes that have duplicate numbers, leaving only distinct numbers from the original list.

For example,
Given 1->2->3->3->4->4->5, return 1->2->5.
Given 1->1->1->2->3, return 2->3.

class Solution:  
    # @param head, a ListNode  
    # @return a ListNode  
    def deleteDuplicates(self, head):  
        if None == head or None == head.next:  
            return head  
  
        new_head = ListNode(-1)  
        new_head.next = head  
        parent = new_head  
        cur = head  
        while None != cur and None != cur.next:   ### check cur.next None  
            if cur.val == cur.next.val:  
                val = cur.val  
                while None != cur and val == cur.val: ### check cur None  
                    cur = cur.next  
                parent.next = cur  
            else:  
                cur = cur.next  
                parent = parent.next  
  
        return new_head.next 
```

```python
206. Reverse Linked List
# recursive solution
class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        
        if not head or not head.next:
            return head
            
        h = self.reverseList(head.next)
        
        head.next.next = head  # head.next is now is the tail 
        head.next = None 
        
        return h
    
# iterative solution

class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        
        first = None
        second = head
        while second:
            next = second.next
            second.next = first
            first = second
            second = next
                
        return first
    
```

```python
92. Reverse Linked List II

Reverse a linked list from position m to n. Do it in one-pass.

Note: 1 ≤ m ≤ n ≤ length of list.
    
class Solution:
    # @param head, a ListNode
    # @param m, an integer
    # @param n, an integer
    # @return a ListNode
    def reverseBetween(self, head, m, n):
        if m == n:
            return head

        dummyNode = ListNode(0)
        dummyNode.next = head
        pre = dummyNode

        for i in range(m - 1):
            pre = pre.next
        
        # reverse the [m, n] nodes
        reverse = None
        cur = pre.next
        for i in range(n - m + 1):
            next = cur.next
            cur.next = reverse
            reverse = cur
            cur = next

        pre.next.next = cur
        pre.next = reverse

        return dummyNode.next
```

```python
234. Palindrome Linked List
Given a singly linked list, determine if it is a palindrome.
class Solution:
    def isPalindrome(self, head: ListNode) -> bool:
        
        res = []
        
        while head:
            res.append(head.val)
            head = head.next
            
        return res == res[::-1]
    
    
O(1) solution. Reverse the second half, compare, and then reverse it back.
```

## MergeSort

```python
class Solution:
    def sortList(self, head: ListNode) -> ListNode:
        
        
        def mergeSortedLists(a, b):
            
            dummyHead = ListNode(-1)
            head = dummyHead
            
            
            while a and b:
                
                if a.val <= b.val:
                    head.next = a
                    head = head.next
                    a = a.next
                else:
                    head.next = b
                    head = head.next
                    b = b.next
            
            if a:
                head.next = a
            if b:
                head.next = b
                
            return dummyHead.next
        
        def findMiddle(a):
            
            if not a:
                return a
            
            slow = a
            fast = a
            
            while fast.next and fast.next.next:
                slow = slow.next
                fast = fast.next.next
                
            return slow
                
                
            
        
        def help(node):
            
            if not node or not node.next:
                return node
            
            mid = findMiddle(node)
            
            first = node
            second = mid.next
            mid.next = None
            
            first = help(first)
            second = help(second)
            
            sort = mergeSortedLists(first, second)
            
            return sort
        
        res = help(head)
        return res
```
