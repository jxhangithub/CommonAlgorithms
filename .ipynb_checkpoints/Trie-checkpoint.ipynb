## Trie implementation
### Trie implementation (array based)

```python
https://leetcode.com/problems/implement-trie-prefix-tree/

Implementation version one (faster than 23%)
    
class TrieNode:
    
    def __init__(self):
        
        self.children = [None] * 26
        self.isEndOfWord = False
        
class Trie:
    
    def __init__(self):
        
        self.root = TrieNode()
        
    def insert(self, key):
        
        curr = self.root
        for i in range(len(key)):
            idx = ord(key[i]) - ord('a')
            if not curr.children[idx]:
                curr.children[idx] = TrieNode()
                curr = curr.children[idx]
            else:
                curr = curr.children[idx]
        
        curr.isEndOfWord = True
        
    def startsWith(self, prefix):
        
        curr = self.root
        for c in prefix:
            idx = ord(c) - ord('a')
            if curr.children[idx]:
                curr = curr.children[idx]
            else:
                return False
        return True
        
    def search(self, key):
        curr = self.root
        for c in key:
            idx = ord(c) - ord('a')
            if curr.children[idx]:
                curr = curr.children[idx]
            else:
                return False
            
            
        return curr and curr.isEndOfWord
    
    
    

```

### Trie implementation (DefaultDict based)

```python
# implementation version two, faster than 56%)
class TrieNode:
    def __init__(self):
        self.childrens = collections.defaultdict(TrieNode)
        self.is_word = False



class Trie:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.root = TrieNode()

    def insert(self, word: str) -> None:
        """
        Inserts a word into the trie.
        """
        curr = self.root
        for l in word:
            curr = curr.childrens[l]
            
        curr.is_word = True
        

    def search(self, word: str) -> bool:
        """
        Returns if the word is in the trie.
        """
        
        curr = self.root
        
        for l in word:
            if l in curr.childrens: # note that we cannot use if curr.children[l], this will creates a key-value pair
                curr = curr.childrens[l]
            else:
                return False
        return curr.is_word

    def startsWith(self, prefix: str) -> bool:
        """
        Returns if there is any word in the trie that starts with the given prefix.
        """
        curr = self.root
        
        for l in prefix:
            if l in curr.childrens: # note that we cannot use if curr.children[l], this will creates a key-value pair
                curr = curr.childrens[l]
            else:
                return False
        return True
```

### C++ implementation

```python
struct TrieNode{
    TrieNode():isWord(false){}
    unordered_map<char, TrieNode*> children;
    bool isWord;
};


class Trie {
public:
    /** Initialize your data structure here. */
    Trie() {
        root = new TrieNode();
    }
    
    /** Inserts a word into the trie. */
    void insert(string word) {
        TrieNode *curr = root;
        
        for(const auto& c: word) {
            if (curr->children.find(c) == curr->children.end()){
                curr->children[c] = new TrieNode();
            }
            curr = curr->children[c];
        }
        curr->isWord = true;
    }
    
    /** Returns if the word is in the trie. */
    bool search(string word) {
        auto node = find(word);
        return node and node->isWord;
    }
    
    /** Returns if there is any word in the trie that starts with the given prefix. */
    bool startsWith(string prefix) {
        return bool(find(prefix));
    }
private:
    inline TrieNode* find(string word){
        
        TrieNode *curr = root;
        for (const auto& c: word){
            
            if (curr->children.find(c) == curr->children.end()){
                return nullptr;
            }
            else{
                curr = curr->children[c];
            }
        }
        
        return curr;
   
        
    }
 
    TrieNode *root;
};

/**
 * Your Trie object will be instantiated and called as such:
 * Trie* obj = new Trie();
 * obj->insert(word);
 * bool param_2 = obj->search(word);
 * bool param_3 = obj->startsWith(prefix);
 */
```

### 211. Add and Search Word - Data structure design

```python
Design a data structure that supports the following two operations:

void addWord(word)
bool search(word)
search(word) can search a literal word or a regular expression string containing only letters a-z or .. A . means it can represent any one letter.
class WordDictionary:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        
        self.root = {}
        
    def addWord(self, word: str) -> None:
        """
        Adds a word into the data structure.
        """
        curr = self.root
        
        for w in word:
            if w not in curr:
                curr[w] = {}
            curr = curr[w]
            
        curr["#"] = True
            
    
    def search(self, word):
        # write your code here
        
        def help(word, node, idx):
            
            if idx == len(word):
                if '#' in node:
                    return True
                else:
                    return False
            
            c = word[idx]
  
            if c == '.':
                for child in node:
                    if child != '#' and help(word, node[child], idx + 1):
                        return True
            else:
                if c not in node:
                    return False
            
                if c in node:
                    return help(word, node[c], idx + 1)

            return False
        
        
        return help(word, self.root, 0)
```

### Lintcode559. Trie Service

```python
559. Trie Service
中文English
Build tries from a list of <word, freq> pairs. Save top 10 for each node.

Example
Example1

Input:  
 <"abc", 2>
 <"ac", 4>
 <"ab", 9>
Output:<a[9,4,2]<b[9,2]<c[2]<>>c[4]<>>> 
Explanation:
			Root
             / 
           a(9,4,2)
          /    \
        b(9,2) c(4)
       /
     c(2)

"""
Definition of TrieNode:
class TrieNode:
    def __init__(self):
        # <key, value>: <Character, TrieNode>
        self.children = collections.OrderedDict()
        self.top10 = []
"""
class TrieService:

    def __init__(self):
        self.root = TrieNode()

    def get_root(self):
        # Return root of trie root, and 
        # lintcode will print the tree struct.
        return self.root

    # @param {str} word a string
    # @param {int} frequency an integer
    # @return nothing
    def insert(self, word, frequency):
        
        curr = self.root
        
        for c in word:
            if c not in curr.children:
                curr.children[c] = TrieNode()
            curr = curr.children[c]
            curr.top10.append(frequency)
            curr.top10.sort(reverse=True)
            if len(curr.top10) > 10:
                curr.top10.pop
                
    # using binary search (although it is no need since array size is small)            
    def insert(self, word, frequency):
        
        curr = self.root
        
        for c in word:
            if c not in curr.children:
                curr.children[c] = TrieNode()
            curr = curr.children[c]
            self._addFreq(curr, frequency)
            if len(curr.top10) > 10:
                curr.top10.pop
        
    def _addFreq(self, node, freq):
        
        if not node.top10:
            node.top10.append(freq)
            return
        
        
        left, right = 0, len(node.top10)
        
        while left < right:
            mid = (left + right) // 2
            
            if node.top10[mid] > freq:
                left = mid + 1
            else:
                right = mid
                
        if freq < node.top10[-1]:
            node.top10.append(freq)
        else:
            node.top10.insert(left, freq)
```

### 212. Word Search II

```python
212. Word Search II

Given a 2D board and a list of words from the dictionary, find all words in the board.

Each word must be constructed from letters of sequentially adjacent cell, where "adjacent" cells are those horizontally
or vertically neighboring. The same letter cell may not be used more than once in a word.

class Solution:
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        self.root = {}
        
        for w in words:
            curr = self.root
            for c in w:
                if c not in curr:
                    curr[c] = {}
                curr = curr[c]
            curr['#'] = True
            
        def find(word):
            
            curr = self.root
            for c in word:
                if c not in curr:
                    return None
                else:
                    curr = curr[c]
            return curr
        
        def startWith(prefix):
            
            return bool(find(prefix))
        
        def search(word):
            node = find(word)
            return bool(node) and '#' in node
            
        
        H = len(board)
        W = len(board[0])
        
        def dfs(i, j, visited, temp):
            
            temp.append(board[i][j])
            visited.add((i, j))
            if startWith(temp):
                
                if search(temp):
                    res.append(''.join(temp))

                for dx, dy in (-1, 0), (1, 0), (0, -1), (0, 1):
                    newX = i + dx
                    newY = j + dy

                    if 0 <= newX < H and 0 <= newY < W and (newX, newY) not in visited:
                        dfs(newX, newY, visited, temp)

            temp.pop()
            visited.remove((i, j))
        
        res = []    
        for i in range(H):
            for j in range(W):
                dfs(i, j, set(), [])
                
        return list(set(res))        

    
    
# improved one
class Solution:
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        self.root = {}
        
        for w in words:
            curr = self.root
            for c in w:
                if c not in curr:
                    curr[c] = {}
                curr = curr[c]
            curr['#'] = True
            
        def find(word):
            
            curr = self.root
            for c in word:
                if c not in curr:
                    return None
                else:
                    curr = curr[c]
            return curr
        
        def startWith(prefix):
            
            return bool(find(prefix))
        
        def search(word):
            node = find(word)
            flag = bool(node) and '#' in node
            if flag:
                del node['#']
                return True
            return False
            
        
        H = len(board)
        W = len(board[0])
        
        def dfs(i, j, board, temp):
            
            temp.append(board[i][j])
            board[i][j] = '#'
            if startWith(temp):
                
                if search(temp):
                    res.append(''.join(temp))

                for dx, dy in (-1, 0), (1, 0), (0, -1), (0, 1):
                    newX = i + dx
                    newY = j + dy

                    if 0 <= newX < H and 0 <= newY < W and board[newX][newY] != '#':
                        dfs(newX, newY, board, temp)

            board[i][j] = temp[-1]
            temp.pop()
        
        res = []    
        for i in range(H):
            for j in range(W):
                dfs(i, j, board, [])
                
        return res        

# signifance improved two, merge two dfs together

class Solution:
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        self.root = {}
        
        for w in words:
            curr = self.root
            for c in w:
                if c not in curr:
                    curr[c] = {}
                curr = curr[c]
            curr['#'] = True
            
        H = len(board)
        W = len(board[0])
        
        def dfs(i, j, board, temp, node):
            
            
                
            if board[i][j] not in node:
                return 
            c = board[i][j]
            temp.append(c)
            board[i][j] = '#'
            node = node[c]
            if '#' in node:
                res.append(''.join(temp))
                del node['#']
            
            for dx, dy in (-1, 0), (1, 0), (0, -1), (0, 1):
                newX = i + dx
                newY = j + dy

                if 0 <= newX < H and 0 <= newY < W and board[newX][newY] != '#':
                    dfs(newX, newY, board, temp, node)
                    
            board[i][j] = temp.pop()
                        
        res = []    
        for i in range(H):
            for j in range(W):
                dfs(i, j, board, [], self.root)
        return res         
        
```

### 677. Map Sum Pairs

```python
Implement a MapSum class with insert, and sum methods.

For the method insert, you'll be given a pair of (string, integer). 
The string represents the key and the integer represents the value. If the key already existed, then the original key-value pair will be overridden to the new one.

For the method sum, you'll be given a string representing the prefix,
and you need to return the sum of all the pairs' value whose key starts with the prefix.

# dfs search approach, consume less memory
class MapSum:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.root = {}

    def insert(self, key: str, val: int) -> None:
        
        curr = self.root
        for c in key:
            if c not in curr:
                curr[c] = {}
            curr = curr[c]
            
        curr['#'] = val
        
    def sum(self, prefix: str) -> int:
        
        res= 0

        stack = [(self.root, 0)]
        
        while stack:
            
            node, idx = stack.pop()
            
            if '#' in node and idx >= len(prefix):
                res += node['#']
            
            if idx < len(prefix):
                c = prefix[idx]
                if c in node:
                    stack.append((node[c], idx + 1))
                else:
                    break
            else:
                
                for child in node:
                    if child == '#':
                        continue
                    else:
                        stack.append((node[child], idx + 1))
        
        return res
                
# using hashmap to speed up and make clean code
class MapSum:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.root = {}
        self.exist = {}

    def insert(self, key: str, val: int) -> None:
        
        delta = val - self.exist.get(key, 0)
        self.exist[key] = val
        
        curr = self.root
        # root node has the total count
        curr['score'] = curr.get('score', 0) + delta
        for c in key:
            if c not in curr:
                curr[c] = {}
            curr = curr[c]
            curr['score'] = curr.get('score', 0) + delta 
         
    def sum(self, prefix: str) -> int:
        curr = self.root
        for c in prefix:
            if c not in curr:
                return 0
            else:
                curr = curr[c]
                
        return curr['score']
            
            
        

# Your MapSum object will be instantiated and called as such:
# obj = MapSum()
# obj.insert(key,val)
# param_2 = obj.sum(prefix)
```

### 648. Replace Words

```python
In English, we have a concept called root, which can be followed by some other words to form another longer word - let's call this word successor. For example, the root an, followed by other, which can form another word another.

Now, given a dictionary consisting of many roots and a sentence. You need to replace all the successor in the sentence with the root forming it. If a successor has many roots can form it, replace it with the root with the shortest length.

You need to output the sentence after the replacement.

class Solution:
    def replaceWords(self, dict: List[str], sentence: str) -> str:
        
        self.root = {}
        
        for w in dict:
            curr = self.root
            for c in w:
                if c not in curr:
                    curr[c] = {}
                curr = curr[c]
                
            curr['#'] = w
        
        def replace(w):
            curr = self.root
            for c in w:
                if c not in curr:
                    break
                else:
                    curr = curr[c]
                    if '#' in curr:
                        return curr['#']
            return w            
        
        res = []
        sentence = sentence.split()
        for w in sentence:
            res.append(replace(w))

        return ' '.join(res)
```

### 745. Prefix and Suffix Search

```python
Given many words, words[i] has weight i.

Design a class WordFilter that supports one function, WordFilter.f(String prefix, String suffix). It will return the word with given prefix and suffix with maximum weight. If no word exists, return -1.

Examples:

Input:
WordFilter(["apple"])
WordFilter.f("a", "e") // returns 0
WordFilter.f("b", "") // returns -1

class WordFilter:

    def __init__(self, words: List[str]):
        
        self.root = {}
        for i, w in enumerate(words):
            self._add(w, i)
    
    def _add(self, word, i):
        curr = self.root
        for c in word:
            if c not in curr:
                curr[c] = {}
            curr = curr[c]
        curr['#'] = i
        

    def f(self, prefix: str, suffix: str) -> int:
        
        
        self.ans = -1
        def help(node, idx, wordSoFar):
            
            if idx >= len(prefix):
                if '#' in node:
                    word = ''.join(wordSoFar)
                    suffixLen = len(suffix)
                    
                    if suffixLen == 0:
                        self.ans = max(self.ans, node['#'])
                    elif len(word) >= suffixLen and word[-suffixLen:] == suffix:
                        self.ans = max(self.ans, node['#'])
            if idx < len(prefix):
                c = prefix[idx]
                if c in node:
                    wordSoFar.append(c)
                    help(node[c], idx + 1, wordSoFar)
                    wordSoFar.pop()
            else:
                for c in node:
                    if c == '#':
                        continue
                    wordSoFar.append(c)
                    help(node[c], idx + 1, wordSoFar)
                    wordSoFar.pop()
        
        
        
        help(self.root, 0, [])
        
        return self.ans
        
# Your WordFilter object will be instantiated and called as such:
# obj = WordFilter(words)
# param_1 = obj.f(prefix,suffix)
```
